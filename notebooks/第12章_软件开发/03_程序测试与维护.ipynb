## 2. 错误的类型 Types of Errors

在编程中，我们会遇到三种主要的错误：

**好消息：** 编译器/解释器会**直接告诉你**错在哪里！

常见的语法错误：
- 忘记冒号 `:` （if, for, while, def 后面）

### 2.2 逻辑错误 Logic Errors

逻辑错误是**最危险的错误**——程序不会崩溃，也不会报错，
但输出的结果**不正确**。

**类比：** 如果语法错误是「话都说不清楚」，那逻辑错误是「话说清楚了但说的是错的」。
前者容易被发现，后者可能骗过所有人。

常见的逻辑错误：
- 使用了错误的运算符（+ 写成 -）
- 偏差一位错误 (Off-by-one Error): 循环多了一次或少了一次
- 条件判断写反了（> 写成 <）
- 变量初始化错误
- 算法本身有问题

````python
# 逻辑错误示例
print('=== Logic Error Examples (逻辑错误示例) ===')
print()

# Example 1: 计算平均值 - 运算符优先级错误
print('Example 1: Calculate average (计算平均值)')
a, b, c = 80, 90, 70

# Wrong: 除法只对 c 生效
wrong_avg = a + b + c / 3
print(f'  Wrong:   (a + b + c / 3)   = {wrong_avg}')

# Correct: 需要括号
right_avg = (a + b + c) / 3
print(f'  Correct: (a + b + c) / 3   = {right_avg}')
print(f'  Bug: Operator precedence! Division before addition.')
print()

# Example 2: Off-by-one error
print('Example 2: Print 1 to 10 (打印1到10)')

print('  Wrong:   range(10) gives:', list(range(10)))  # 0-9
print('  Correct: range(1, 11) gives:', list(range(1, 11)))  # 1-10
print(f'  Bug: Off-by-one error! range(10) starts at 0, ends at 9.')
print()

# Example 3: 条件写反
print('Example 3: Check if passing (判断是否及格, pass >= 60)')
score = 55

# Wrong
if score > 60:
    result_wrong = 'PASS'
else:
    result_wrong = 'FAIL'

# Correct (包含60分)
if score >= 60:
    result_right = 'PASS'
else:
    result_right = 'FAIL'

print(f'  Score = {score}')
print(f'  Wrong  (> 60):  {result_wrong} (should be FAIL for 55)')
print(f'  Note: Both give FAIL for 55, but for score=60:')
print(f'  Wrong  (> 60):  FAIL  <-- score of 60 should PASS!')
print(f'  Correct (>= 60): PASS')
print(f'  Bug: Boundary condition error! >= not >')
````

### 2.3 运行时错误 Runtime Errors

运行时错误发生在程序**执行过程中**——程序可以启动，
但在某个时刻遇到无法处理的情况而崩溃。

**类比：** 你在开车（程序运行中），突然前面出现一个大坑（异常情况），
如果你没有提前准备，车就翻了（程序崩溃）。

````python
# 常见运行时错误演示
print('=== Runtime Error Examples (运行时错误示例) ===')
print()

# 1. ZeroDivisionError - 除以零
print('1. ZeroDivisionError (除以零错误)')
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print(f'   Error caught: {e}')
    print('   Fix: Check divisor is not zero before dividing')
print()

# 2. IndexError - 索引越界
print('2. IndexError (索引越界错误)')
try:
    my_list = [1, 2, 3]
    value = my_list[5]
except IndexError as e:
    print(f'   Error caught: {e}')
    print(f'   List has {len(my_list)} items (index 0-{len(my_list)-1}), but tried index 5')
print()

# 3. TypeError - 类型错误
print('3. TypeError (类型错误)')
try:
    result = 'hello' + 5
except TypeError as e:
    print(f'   Error caught: {e}')
    print('   Fix: Convert to same type first: "hello" + str(5)')
print()

# 4. FileNotFoundError - 文件未找到
print('4. FileNotFoundError (文件未找到错误)')
try:
    f = open('nonexistent_file.txt', 'r')
except FileNotFoundError as e:
    print(f'   Error caught: {e}')
    print('   Fix: Check if file exists before opening')
print()

# 5. ValueError - 值错误
print('5. ValueError (值错误)')
try:
    number = int('hello')
except ValueError as e:
    print(f'   Error caught: {e}')
    print('   Fix: Validate input before converting')
print()

# 6. RecursionError - 递归深度超限
print('6. RecursionError (递归深度超限)')
def infinite_recursion():
    return infinite_recursion()
try:
    infinite_recursion()
except RecursionError:
    print('   Error caught: maximum recursion depth exceeded')
    print('   Fix: Add a base case to stop recursion')
````

## 3. 测试方法 Testing Methods

软件测试有多种方法，每种方法从不同角度发现不同类型的问题。

### 3.1 干运行与跟踪表 Dry Run and Trace Table

**干运行 (Dry Run)** 是指**不用电脑**，用纸和笔手动模拟程序的执行过程。
**跟踪表 (Trace Table)** 是记录每一步中变量值变化的表格。

**为什么这很重要？**
- 帮助你深入理解程序的执行流程
- 快速定位逻辑错误
- 考试中经常考到！

**类比：** 就像侦探重建犯罪现场——一步一步回溯发生了什么。

````python
# 跟踪表生成器 Trace Table Generator
print('=== Trace Table Example (跟踪表示例) ===')
print()
print('Code to trace (要跟踪的代码):')
print('  x = 0')
print('  for i in range(1, 6):')
print('      x = x + i')
print('  print(x)')
print()

# Generate trace table
print(f'{"Step":<6} {"i":<6} {"x (before)":<14} {"x + i":<10} {"x (after)":<12}')
print('-' * 48)

x = 0
step = 0
print(f'{"Init":<6} {"-":<6} {"-":<14} {"-":<10} {x:<12}')

for i in range(1, 6):
    step += 1
    x_before = x
    x = x + i
    print(f'{step:<6} {i:<6} {x_before:<14} {str(x_before) + "+" + str(i):<10} {x:<12}')

print('-' * 48)
print(f'Final output: x = {x}')
print()
print('Note: The trace table shows how x accumulates:')
print('x = 0 + 1 + 2 + 3 + 4 + 5 = 15')
print('This is the sum of numbers 1 to 5!')
````

````python
# 更复杂的跟踪表示例
print('=== Trace Table: Binary Search (二分查找跟踪表) ===')
print()
print('Searching for target = 7 in [1, 3, 5, 7, 9, 11, 13]')
print()

arr = [1, 3, 5, 7, 9, 11, 13]
target = 7
low = 0
high = len(arr) - 1
found = False
step = 0

print(f'{"Step":<6} {"low":<6} {"high":<6} {"mid":<6} {"arr[mid]":<10} {"Comparison":<20} {"Action":<15}')
print('-' * 70)

while low <= high and not found:
    step += 1
    mid = (low + high) // 2
    mid_val = arr[mid]
    
    if mid_val == target:
        comparison = f'{mid_val} == {target}'
        action = 'FOUND!'
        found = True
    elif mid_val < target:
        comparison = f'{mid_val} < {target}'
        action = 'low = mid + 1'
        print(f'{step:<6} {low:<6} {high:<6} {mid:<6} {mid_val:<10} {comparison:<20} {action:<15}')
        low = mid + 1
        continue
    else:
        comparison = f'{mid_val} > {target}'
        action = 'high = mid - 1'
        print(f'{step:<6} {low:<6} {high:<6} {mid:<6} {mid_val:<10} {comparison:<20} {action:<15}')
        high = mid - 1
        continue
    
    print(f'{step:<6} {low:<6} {high:<6} {mid:<6} {mid_val:<10} {comparison:<20} {action:<15}')

print('-' * 70)
print(f'Result: target {target} found at index {mid}')
````

### 3.2 代码走查 Walkthrough

**代码走查** (Walkthrough) 是一种**团队活动**——开发者向团队成员
展示代码，大家一起逐行检查。

**特点：**
- 不需要运行代码
- 利用「集体智慧」发现问题
- 不仅找 bug，还能改善代码质量
- 是一种**同行评审** (Peer Review) 的形式

> **程序员界的名言：** "Given enough eyeballs, all bugs are shallow." —— Linus' Law
> （如果有足够多的人看代码，所有 bug 都很容易被发现。）

### 3.3 白盒测试 vs 黑盒测试 White-box vs Black-box Testing

这是两种**根本不同的测试哲学**：

| 特性 | White-box 白盒测试 | Black-box 黑盒测试 |
|---|---|---|
| **能看到代码吗？** | 能 (看到内部结构) | 不能 (只看到输入和输出) |
| **测试什么？** | 代码的内部逻辑 | 功能是否符合需求 |
| **谁来做？** | 通常是开发者 | 通常是测试人员或用户 |
| **关注点** | 所有路径都被执行了吗？ | 所有需求都满足了吗？ |
| **类比** | 检查汽车引擎的每个零件 | 试驾看车是否跑得好 |

#### 白盒测试的关键概念：代码覆盖率 (Code Coverage)
- **语句覆盖** (Statement Coverage): 每条语句都执行了吗？
- **分支覆盖** (Branch Coverage): 每个 if/else 分支都走到了吗？
- **路径覆盖** (Path Coverage): 所有可能的执行路径都测试了吗？

````python
# 白盒测试 vs 黑盒测试示例

def calculate_grade(score):
    """Calculate grade based on score."""
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

# White-box testing: 测试每个分支 (branch coverage)
print('=== White-box Testing (白盒测试) ===')
print('Goal: Test every branch in the code')
print()
white_box_tests = [
    (95, 'A', 'Tests score >= 90 branch'),
    (85, 'B', 'Tests score >= 80 branch'),
    (75, 'C', 'Tests score >= 70 branch'),
    (65, 'D', 'Tests score >= 60 branch'),
    (50, 'F', 'Tests else branch'),
]

for score, expected, reason in white_box_tests:
    result = calculate_grade(score)
    status = 'PASS' if result == expected else 'FAIL'
    print(f'  Score={score:<4} Expected={expected}  Got={result}  [{status}]  ({reason})')

print()
print('Branch coverage: 5/5 = 100%')
print()

# Black-box testing: 根据需求测试 (不看代码)
print('=== Black-box Testing (黑盒测试) ===')
print('Goal: Test based on requirements specification')
print()
black_box_tests = [
    (100, 'A', 'Maximum valid score'),
    (0,   'F', 'Minimum valid score'),
    (90,  'A', 'Boundary: exactly 90'),
    (89,  'B', 'Boundary: just below 90'),
    (60,  'D', 'Boundary: exactly 60'),
    (59,  'F', 'Boundary: just below 60'),
]

for score, expected, reason in black_box_tests:
    result = calculate_grade(score)
    status = 'PASS' if result == expected else 'FAIL'
    print(f'  Score={score:<4} Expected={expected}  Got={result}  [{status}]  ({reason})')
````

### 3.4 其他测试类型 Other Testing Types

| 测试类型 | 英文 | 说明 | 何时进行 |
|---|---|---|---|
| **集成测试** | Integration Testing | 测试模块之间能否正确协作 | 模块合并时 |
| **Alpha 测试** | Alpha Testing | 开发团队内部测试 | 开发末期 |
| **Beta 测试** | Beta Testing | 选定的外部用户测试 | 发布前 |
| **验收测试** | Acceptance Testing | 客户确认是否满足需求 | 交付前 |
| **桩测试** | Stub Testing | 用占位符替代未完成的模块 | 开发过程中 |

#### 理解桩测试 Stub Testing

当你的程序依赖一个还没写好的模块时，你可以用一个「桩」(Stub) 来替代它——
一个临时的、简化的模块，只返回固定的测试数据。

**类比：** 拍电影时，危险动作用替身演员（Stub）代替主演。

````python
# Stub Testing 示例
print('=== Stub Testing Example (桩测试示例) ===')
print()

# 假设 get_data_from_database 还没写好
# 我们用一个 stub 来替代它

def get_data_from_database_STUB(student_id):
    """STUB: 返回固定的测试数据，模拟数据库查询"""
    print(f'  [STUB] Pretending to query database for ID: {student_id}')
    # 返回固定的测试数据
    return {'id': student_id, 'name': 'Test Student', 'score': 85}

def generate_report(student_id):
    """主功能：生成学生报告（使用 stub 替代真实数据库）"""
    data = get_data_from_database_STUB(student_id)  # 用 stub 替代
    grade = 'A' if data['score'] >= 90 else 'B' if data['score'] >= 80 else 'C'
    report = f"Student: {data['name']}, Score: {data['score']}, Grade: {grade}"
    return report

# 即使数据库模块没写好，我们也能测试 generate_report 的逻辑
print('Testing generate_report with stub:')
result = generate_report(12345)
print(f'  Result: {result}')
print()
print('Advantage: We can test our logic WITHOUT a real database!')
print('When the real module is ready, just replace the stub.')
````

## 4. 测试策略与测试计划 Test Strategy and Test Plan

### 4.1 测试策略 Test Strategy

测试策略是一份**高层次文档**，定义了：
- 测试的总体方法和目标
- 使用哪些测试方法
- 测试的范围（测什么，不测什么）
- 资源分配（谁来测试）
- 时间安排

### 4.2 测试计划 Test Plan

测试计划是一份**详细的文档**，包含每个具体测试用例 (Test Case)：
- **测试编号** (Test ID)
- **测试描述** (Description)
- **输入数据** (Input Data)
- **预期输出** (Expected Output)
- **实际输出** (Actual Output)
- **结果** (Pass/Fail)

### 4.3 测试数据的三种类型 Three Types of Test Data

| 类型 | 英文 | 说明 | 例子（输入年龄 0-120） |
|---|---|---|---|
| **正常数据** | Normal Data | 典型的有效输入 | 25, 50, 80 |

### 练习 1: Debug Challenge (找Bug挑战)

````python
# DEBUG CHALLENGE - 找出并修复以下代码中的所有错误！
# 这个函数应该返回列表中所有偶数的平均值

print('=== Debug Challenge (找Bug挑战) ===')
print()

# BUGGY VERSION (有 Bug 的版本)
def average_of_evens_buggy(numbers):
    total = 0
    count = 0
    for num in numbers:
        if num % 2 == 0:   # check if even
            total = total + num
            count = count + 1
    average = total / count  # Bug 1: What if count is 0?
    return average

# FIXED VERSION (修复后的版本)
def average_of_evens_fixed(numbers):
    total = 0
    count = 0
    for num in numbers:
        if num % 2 == 0:
            total = total + num
            count = count + 1
    if count == 0:          # Fix 1: Handle division by zero
        return 0
    average = total / count
    return average

# Test cases
test_cases = [
    ([2, 4, 6, 8], 5.0, 'All even numbers'),
    ([1, 2, 3, 4, 5], 3.0, 'Mixed numbers'),
    ([1, 3, 5, 7], 0, 'No even numbers (edge case!)'),
    ([], 0, 'Empty list (edge case!)'),
]

print('Testing fixed version:')
print(f'{"Input":<20} {"Expected":<10} {"Got":<10} {"Status":<8} {"Note"}')
print('-' * 65)
for nums, expected, note in test_cases:
    result = average_of_evens_fixed(nums)
    status = 'PASS' if result == expected else 'FAIL'
    print(f'{str(nums):<20} {expected:<10} {result:<10} {status:<8} {note}')
````
