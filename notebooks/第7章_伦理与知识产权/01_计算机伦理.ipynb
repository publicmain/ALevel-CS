## 1. 什么是计算机伦理？(What is Computer Ethics?)

### 1.1 从哲学到代码：理解"伦理"

**伦理（Ethics）** 是关于"什么是对的、什么是错的"的哲学研究。它不是法律——法律规定你**不能**做什么，而伦理引导你**应该**做什么。

> 🎯 **关键区别：**
> - **法律 (Law)：** 违反会受到法律制裁（罚款、监禁）
> - **伦理 (Ethics)：** 违反可能不犯法，但会伤害他人或社会

**类比：** 想象你在考试中发现同学的答案纸掉在地上。
- **法律上：** 捡起来看可能不犯法
- **伦理上：** 你应该归还而不是偷看

### 1.2 为什么计算机专业人员需要特别注意伦理？

| 原因 | 解释 | 例子 |
|------|------|------|
| **影响范围巨大** | 一行代码可能影响数十亿用户 | Facebook的算法影响30亿用户 |
| **技术不对称性** | 开发者比用户更了解系统 | 用户不知道App在后台收集什么数据 |
| **决策自动化** | 算法可能做出影响人生的决定 | AI决定你能否获得贷款 |
| **难以撤回** | 数据一旦泄露无法"取消泄露" | 个人信息泄露后无法收回 |
| **社会信任** | 社会依赖技术系统正常运作 | 银行系统、医疗系统、交通系统 |

> 💡 **深度思考：** 与传统工程师（如建桥的土木工程师）相比，软件工程师的伦理责任更为复杂——因为软件的影响往往是**看不见的**、**大规模的**、且**持续演变的**。

### 1.3 真实案例分析 (Real-world Cases)

#### 📌 案例一：Facebook-Cambridge Analytica 丑闻 (2018)

**发生了什么？**
1. 一个名为 "This Is Your Digital Life" 的性格测试App获取了27万用户的Facebook数据
2. 但该App不仅收集了这些用户的数据，还**未经同意**收集了他们**所有好友**的数据
3. 最终影响了 **8,700万** 用户的个人数据
4. 这些数据被用于2016年美国总统大选的**精准政治广告投放**

**伦理问题：**
- 用户是否真正理解他们同意了什么？（Informed Consent）
- 一个人能否代替其好友"同意"数据被收集？
- 政治操纵算不算"合理使用"数据？

**后果：**
- Facebook被罚款50亿美元（当时史上最大科技公司罚款）
- CEO Mark Zuckerberg 被传唤至美国国会作证
- 推动了全球数据保护立法（如欧盟GDPR）

---

#### 📌 案例二：自动驾驶汽车的"电车难题" (Trolley Problem)

**经典电车难题：**
一辆失控的电车正驶向5个人。你可以拉一个开关，让电车转向另一条轨道，但那条轨道上有1个人。你会拉开关吗？

**应用到自动驾驶：**
想象一辆自动驾驶汽车面临不可避免的事故：
- **选择A：** 撞向行人（可能导致1人死亡）
- **选择B：** 急转弯撞墙（可能导致车内乘客死亡）

| 问题 | 思考 |
|------|------|
| 谁来做这个决定？ | 程序员？公司？政府？ |
| 算法应该保护谁？ | 乘客优先？还是最小化总伤亡？ |
| 年龄是否应该影响决策？ | 小孩 vs 老人？这是否是歧视？ |
| 责任归谁？ | 程序员？汽车制造商？车主？ |

#### 📌 案例三：AI偏见 (AI Bias)

**Amazon招聘AI (2018)：**
- Amazon开发了一个AI来筛选简历
- 训练数据：过去10年的招聘记录（当时科技行业以男性为主）
- 结果：AI系统**自动降低**包含"女性"相关词汇的简历评分

**为什么会这样？**
AI从历史数据中"学到"了偏见——如果过去招聘偏向男性，AI就会复制这种偏见。

> ⚠️ **核心教训：** AI不是客观的。它反映的是训练数据中的偏见。"垃圾进，垃圾出"（Garbage In, Garbage Out）在伦理领域同样适用。

## 2. 专业伦理机构 (Professional Ethical Bodies)

### 为什么需要专业伦理机构？

想象一下没有医生执照制度的世界——任何人都可以给你做手术。**可怕吧？**

计算机行业也是类似的。虽然目前编程不需要"执照"，但专业机构通过制定**行为准则（Code of Conduct）**来确保从业者遵守伦理标准。

> 🎯 **核心思想：** 专业机构就像一个"道德指南针"——当你不确定某件事是否应该做时，可以参考它们的准则。

---

### 2.1 BCS — 英国计算机学会 (British Computer Society)

**成立时间：** 1957年 | **总部：** 英国

BCS是英国IT行业的**特许专业机构（Chartered Professional Body）**，相当于IT界的"医师协会"。

**BCS 行为准则四大原则：**

| 原则 | 含义 | 例子 |
|------|------|------|
| **1. 公共利益 (Public Interest)** | 保护公众健康、安全和环境 | 发现安全漏洞时必须上报，不能隐瞒 |
| **2. 专业能力与诚信 (Professional Competence & Integrity)** | 不做超出能力范围的事 | 不熟悉加密技术就不应该设计安全系统 |
| **3. 权威义务 (Duty to Relevant Authority)** | 对雇主和客户负责 | 不泄露公司机密，不利用职位谋私利 |
| **4. IT专业义务 (Duty to the Profession)** | 维护行业声誉 | 帮助新人、分享知识、不抄袭他人作品 |

---

### 2.2 IEEE — 电气与电子工程师协会

**成立时间：** 1963年 | **总部：** 美国

IEEE是全球最大的技术专业组织，拥有超过**40万会员**，遍布160多个国家。

**IEEE的伦理准则要求成员承诺：**
1. 将公众安全放在首位
2. 避免利益冲突
3. 基于真实数据做出判断
4. 拒绝贿赂
5. 提升技术理解
6. 公平对待所有人
7. 不伤害他人声誉
8. 帮助同事提升专业能力
9. 正确归因他人贡献
10. 支持同行遵守伦理准则

---

### 2.3 ACM — 美国计算机协会 (Association for Computing Machinery)

**成立时间：** 1947年 | **总部：** 美国（世界上第一个计算机科学学术组织）

**ACM伦理准则（2018年更新）的核心原则：**
1. **贡献社会福祉** — 技术应造福人类
2. **避免伤害** — 评估技术可能造成的损害
3. **诚实可信** — 透明地报告系统的局限性
4. **公平无歧视** — 确保技术不歧视任何群体
5. **尊重知识产权** — 不抄袭、不盗版
6. **尊重隐私** — 只收集必要的数据
7. **保守机密** — 保护用户和客户的信息

---

### 2.4 为什么要加入专业机构？

| 好处 | 说明 |
|------|------|
| **问责制 (Accountability)** | 有明确的标准来判断行为是否恰当 |
| **行业标准 (Standards)** | 统一的最佳实践指南 |
| **公众信任 (Trust)** | 客户知道你遵循专业准则 |
| **职业发展 (Career)** | 获得认证、培训和networking机会 |
| **法律保护** | 遵循准则可以在争议中证明你的职业行为 |
| **知识更新** | 获取最新的行业研究和技术动态 |

> 💡 **考试提示：** 考试中可能会问"为什么计算机从业者应该加入专业机构"，你需要能列出**至少3-4个理由**，并用例子来支持。

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(14, 8))
ax.axis('off')

# Title
ax.text(0.5, 0.97, 'Three Major Computing Professional Bodies Comparison',
        ha='center', va='top', fontsize=16, fontweight='bold',
        transform=ax.transAxes)

# Draw three columns
colors = ['#3498db', '#e74c3c', '#2ecc71']
orgs = ['BCS', 'IEEE', 'ACM']
founded = ['1957', '1963', '1947']
hq = ['UK', 'USA', 'USA']
members = ['68,000+', '400,000+', '100,000+']
focus_areas = [
    'IT professionalism\nChartered status\nUK-focused',
    'Electrical & Computer\nEngineering\nGlobal standards',
    'Computing science\nResearch & education\nACM Awards'
]
key_principles = [
    'Public Interest\nCompetence & Integrity\nDuty to Authority\nDuty to Profession',
    '10 ethical commitments\nSafety first\nHonesty\nFairness',
    '7 core principles\nSocial welfare\nAvoid harm\nRespect privacy'
]

for i, (org, color) in enumerate(zip(orgs, colors)):
    x_center = 0.18 + i * 0.32
    
    # Organization name circle
    circle = plt.Circle((x_center, 0.82), 0.06, color=color, transform=ax.transAxes, zorder=5)
    ax.add_patch(circle)
    ax.text(x_center, 0.82, org, ha='center', va='center', fontsize=14, 
            fontweight='bold', color='white', transform=ax.transAxes, zorder=6)
    
    # Info boxes
    box_y_positions = [0.68, 0.55, 0.38, 0.18]
    box_labels = ['Founded', 'HQ / Members', 'Focus', 'Key Principles']
    box_contents = [
        founded[i],
        f'{hq[i]} / {members[i]}',
        focus_areas[i],
        key_principles[i]
    ]
    
    for j, (y_pos, label, content) in enumerate(zip(box_y_positions, box_labels, box_contents)):
        height = 0.08 if j < 2 else 0.12
        rect = plt.Rectangle((x_center - 0.13, y_pos - height/2), 0.26, height,
                             facecolor=color, alpha=0.1, edgecolor=color,
                             linewidth=2, transform=ax.transAxes, zorder=3)
        ax.add_patch(rect)
        ax.text(x_center, y_pos + height/2 + 0.01, label, ha='center', va='bottom',
                fontsize=9, fontweight='bold', color=color, transform=ax.transAxes)
        ax.text(x_center, y_pos, content, ha='center', va='center',
                fontsize=9, transform=ax.transAxes)

# Common ground box at bottom
rect = plt.Rectangle((0.05, 0.02), 0.9, 0.07, facecolor='#f39c12', alpha=0.2,
                     edgecolor='#f39c12', linewidth=2, transform=ax.transAxes)
ax.add_patch(rect)
ax.text(0.5, 0.055, 'COMMON GOAL: Promote ethical, responsible, and professional computing practice',
        ha='center', va='center', fontsize=12, fontweight='bold', color='#e67e22',
        transform=ax.transAxes)

plt.tight_layout()
plt.savefig('professional_bodies.png', dpi=100, bbox_inches='tight')
plt.show()
print('All three bodies share the goal of promoting ethical computing.')
```

## 3. 伦理与非伦理行为场景分析 (Ethical vs Unethical Scenarios)

### 3.1 场景分析

让我们来分析几个真实的伦理困境场景：

---

#### 🔴 场景一：未经同意收集用户数据

**情况：** 你是一家健身App公司的开发者。产品经理要求你在App中添加隐藏代码，在用户不知情的情况下收集他们的GPS位置数据，然后将这些数据卖给广告公司。

| 分析角度 | 评价 |
|----------|------|
| **法律层面** | 违反GDPR等数据保护法，可能面临巨额罚款 |
| **伦理层面** | 严重违反知情同意原则（Informed Consent） |
| **BCS准则** | 违反"公共利益"和"专业诚信"原则 |
| **正确做法** | 拒绝执行，向上级或伦理委员会报告 |

---

#### 🔴 场景二：故意设计上瘾软件

**情况：** 社交媒体公司的产品团队使用心理学技巧（如无限下拉刷新、随机奖励通知）来最大化用户使用时间，**即使内部研究表明这对青少年心理健康有害**。

**这正是2021年 Facebook "吹哨人" Frances Haugen 揭露的情况：**
- Facebook内部研究发现Instagram让32%的青少年女孩对自己的身体形象感觉更差
- 但公司选择隐瞒这些研究结果
- 继续优化"参与度"算法以增加广告收入

> ⚠️ **伦理判断：** 当利润与用户福祉冲突时，伦理要求将用户福祉放在首位。

---

#### 🔴 场景三：算法偏见在招聘/贷款中的影响

**情况：** 一家银行使用AI系统来决定是否批准贷款申请。该系统发现，居住在某些邮政编码地区的人违约率较高——但这些地区恰好是少数族裔聚居区。

**问题链：**
1. 历史数据反映了历史歧视 →
2. AI"学到"了这种歧视模式 →
3. AI拒绝这些地区的贷款申请 →
4. 这些地区经济继续落后 →
5. 产生新的"违约率高"数据 →
6. 形成**恶性循环（Feedback Loop）**

---

#### 🟡 场景四：数据中心的环境影响

**你知道吗？**
- 全球数据中心消耗的电力约占全球总用电量的**1-2%**
- 训练一个大型AI模型（如GPT-3）产生的碳排放约等于**5辆汽车一生的排放**
- 一次Google搜索消耗约0.3Wh电力，全球每天约85亿次搜索

| 公司 | 环保承诺 |
|------|----------|
| Google | 2030年实现24/7无碳运营 |
| Apple | 整个供应链2030年碳中和 |

## 4. 版权法 (Copyright Legislation)

### 4.1 什么是版权？

**版权（Copyright）** 是一种法律保护，自动授予**原创作品的创作者**对其作品的专有权利。

> 🎯 **关键点：** 版权是**自动获得**的——你不需要注册或申请。当你写了一首诗、画了一幅画、或写了一段代码，你就自动拥有了版权。

**版权保护什么？**

| 类别 | 例子 | 与计算机的关系 |
|------|------|----------------|
| **文学作品** | 书籍、文章、手册 | 软件文档、用户手册 |
| **软件代码** | 源代码、目标代码 | 程序、应用、网站代码 |
| **音乐作品** | 歌曲、音效 | 游戏音乐、App音效 |
| **视觉作品** | 图片、图标、UI设计 | 网站设计、App界面 |
| **数据库** | 结构化数据集合 | 数据库设计和内容 |
| **影视作品** | 视频、动画 | 软件教程视频、宣传片 |

### 4.2 版权期限

| 国家/地区 | 期限 |
|-----------|------|
| 英国 | 作者去世后**70年** |
| 美国 | 作者去世后**70年**（公司作品为发表后95年） |
| 中国 | 作者去世后**50年** |
| 欧盟 | 作者去世后**70年** |

> 💡 **有趣事实：** 这就是为什么莎士比亚的作品是"公共领域"（Public Domain）的——因为他去世已经超过400年了。任何人都可以免费使用。

### 4.3 侵犯版权的后果

| 后果类型 | 具体内容 |
|----------|----------|
| **民事诉讼** | 版权持有者可以起诉你要求赔偿 |
| **罚款** | 可能面临数万到数百万的罚款 |
| **刑事处罚** | 严重情况下可能面临监禁 |
| **声誉损害** | 被列入黑名单，失去行业信任 |
| **产品下架** | 侵权产品必须停止销售和分发 |

### 4.4 合理使用 (Fair Use / Fair Dealing)

**合理使用**允许在特定情况下使用受版权保护的材料，**无需获得许可**：

| 允许的使用 | 条件 |
|-----------|------|
| **教育目的** | 课堂教学中展示少量内容 |
| **研究和学习** | 个人学习中引用少量内容 |
| **评论和批评** | 书评、影评中引用原作 |
| **新闻报道** | 报道中引用相关内容 |
| **讽刺和戏仿** | 对原作的艺术性改编 |

> ⚠️ **注意：** 合理使用不是万能的挡箭牌。如果你的"合理使用"损害了原作者的经济利益，仍然可能构成侵权。

## 5. 软件许可证类型 (Software Licensing Types)

### 为什么需要许可证？

当你写了一段代码，你自动拥有版权。但版权只告诉别人"这是我的"——**许可证（License）** 告诉别人"你可以如何使用它"。

> 🎯 **类比：** 版权就像房产证（证明房子是你的），许可证就像租约（规定租客可以怎么使用房子）。

---

### 5.1 专有/商业许可证 (Proprietary / Commercial License)

**定义：** 源代码不公开，用户付费获得使用权（不是所有权）。

**例子：** Microsoft Office, Adobe Photoshop, Windows

**特点：**
- ❌ 不能查看源代码
- ❌ 不能修改软件
- ❌ 不能重新分发
- ✅ 通常有专业技术支持
- ✅ 定期更新和安全补丁
- 💰 需要付费（买断或订阅）

**你实际拥有什么？** 只是一个**使用许可**——你买的是"使用权"而不是"所有权"。这就是为什么软件的"最终用户许可协议"(EULA) 通常会说"您不得..."。

---

### 5.2 开源许可证 (Open Source Licenses)

**定义：** 源代码公开，任何人都可以查看、修改和分发。

**但"开源"不等于"没有规则"！** 不同的开源许可证有不同的要求：

#### 🔵 Copyleft 许可证 — GPL (GNU General Public License)

**核心规则：** 如果你修改了GPL软件并分发，你的修改版本**也必须**使用GPL许可证开源。

**类比：** 这就像一个"传染性"条款——GPL的"自由"属性会"传染"给所有衍生作品。

**例子：** Linux内核, WordPress, GIMP

**Richard Stallman（自由软件之父）的理念：**
> "Free as in freedom, not free as in free beer."
> （自由如同言论自由，而非免费啤酒。）

#### 🟢 宽松许可证 — MIT License

**核心规则：** 你可以做几乎任何事（使用、修改、卖钱都行），只需保留原始版权声明。

**类比：** 这就像说"随便用，但别忘了说是从我这儿来的"。

**例子：** React, Node.js, jQuery

#### 🟡 Apache License 2.0

**类似MIT但增加了：**
- 专利保护条款（Patent Grant）
- 更明确的贡献者权利

**例子：** Android (部分), Kubernetes, Apache HTTP Server

---

### 5.3 免费软件 (Freeware)

**定义：** 免费使用，但源代码**不公开**。

**例子：** Skype, Adobe Acrobat Reader, Discord

**注意区分：**
- Freeware ≠ Free Software（自由软件）
- Freeware ≠ Open Source
- 你可以免费用，但不能看到代码或修改它

---

### 5.4 共享软件 (Shareware)

**定义：** "先试后买"——可以免费试用一段时间或部分功能，之后需要付费。

**例子：** WinRAR（虽然弹窗催你付费，但很多人一直"试用"...）, 很多手机游戏的"免费试玩"

**常见模式：**
- 时间限制（30天试用期）
- 功能限制（免费版只有基础功能）
- 水印（导出的文件带有软件水印）

---

### 5.5 Creative Commons (CC) 许可证

**定义：** 主要用于**非软件**的创意作品（文章、图片、音乐、视频）。

**模块化设计——像搭积木一样组合：**

| 模块 | 符号 | 含义 |
|------|------|------|
| **BY (Attribution)** | 👤 | 必须标明原作者 |
| **SA (ShareAlike)** | 🔄 | 衍生作品必须使用相同许可证 |
| **NC (NonCommercial)** | 🚫💰 | 不能用于商业目的 |
| **ND (NoDerivatives)** | 🚫✏️ | 不能修改原作品 |

**常见组合：**
- **CC BY:** 最宽松，只需署名
- **CC BY-SA:** 署名 + 衍生作品相同许可证（Wikipedia使用这个！）
- **CC BY-NC:** 署名 + 非商业使用
- **CC BY-NC-ND:** 最严格，署名 + 非商业 + 不能修改

## 6. 人工智能概述 (Artificial Intelligence Overview)

### 6.1 什么是AI？

**人工智能（Artificial Intelligence, AI）** 是让计算机模拟人类智能行为的技术。

> 🎯 **简单定义：** AI是让机器能够"学习"、"推理"和"做决定"的技术。

### 6.2 AI的分类

| 类型 | 定义 | 现状 | 例子 |
|------|------|------|------|
| **窄AI / 弱AI (Narrow AI)** | 只能完成特定任务的AI | ✅ 已经存在 | Siri, AlphaGo, 推荐算法 |
| **通用AI / 强AI (General AI / AGI)** | 能像人类一样完成任何智力任务 | ❌ 尚未实现 | 科幻电影中的AI |
| **超级AI (Super AI)** | 超越人类的智能 | ❌ 纯理论 | 终结者、天网 |

> 💡 **重要认知：** 当前所有AI都是**窄AI**——它们在特定任务上可能超越人类（如下棋），但无法像人类一样灵活地处理各种任务。

### 6.3 AI的主要应用领域

| 领域 | 技术 | 应用 |
|------|------|------|
| **自然语言处理 (NLP)** | 理解和生成人类语言 | ChatGPT, 机器翻译, 语音助手 |
| **计算机视觉 (Computer Vision)** | 理解图像和视频 | 人脸识别, 自动驾驶, 医学影像分析 |
| **推荐系统 (Recommendation)** | 预测用户偏好 | Netflix推荐, 淘宝"猜你喜欢", Spotify |
| **医疗健康 (Healthcare)** | 辅助诊断和药物研发 | AI读CT片, 药物分子设计 |
| **游戏 (Gaming)** | 智能对手和NPC | AlphaGo, 游戏AI |
| **金融 (Finance)** | 风险评估和交易 | 信用评分, 量化交易, 欺诈检测 |

### 6.4 AI的社会影响 (Social Impact)

#### 🔴 就业影响 (Job Displacement)

**可能被自动化的工作：**
- 重复性体力劳动（工厂装配线）
- 数据输入和处理
- 基础客服（聊天机器人）
- 简单的内容审核

**不太可能被完全替代的工作：**
- 需要创造力的工作（艺术、设计）
- 需要情感连接的工作（心理咨询、护理）
- 需要复杂决策的工作（战略规划、研究）
- 需要物理灵活性的工作（水管工、电工）

> 💡 **关键观点：** AI更可能**改变**工作而不是**消除**工作。就像Excel没有消除会计师，而是改变了会计工作的方式。

#### 🔴 监控与隐私 (Surveillance)

- 中国的社会信用系统使用AI面部识别
- 英国有超过600万个CCTV摄像头
- 美国警方使用预测性警务（Predictive Policing）
- **争议：** 面部识别对深色皮肤人群的错误率更高

#### 🔴 深度伪造 (Deepfakes)

AI可以生成几乎无法分辨的假视频：
- 让名人"说"他们从未说过的话
- 制造虚假的新闻视频
- 身份欺诈和诈骗
- **威胁：** 如果任何视频都可能是假的，我们还能信任什么？

### 6.5 AI的经济影响 (Economic Impact)

| 方面 | 积极影响 | 消极影响 |
|------|----------|----------|
| **生产力** | 大幅提高效率和产出 | 减少对人工的需求 |
| **新产业** | 创造全新的行业和工作 | 传统行业被颠覆 |
| **财富分配** | 降低某些商品/服务的成本 | 加剧贫富差距（AI的收益集中在少数公司） |
| **全球竞争** | 国家AI战略成为竞争力关键 | AI"军备竞赛"加剧国际紧张 |

### 6.6 AI的环境影响 (Environmental Impact)

| 阶段 | 能耗 |
|------|------|
| **训练大型模型** | GPT-3训练一次消耗约1,287 MWh电力 |
| **推理/使用** | 每次ChatGPT查询的能耗是Google搜索的约10倍 |

> ⚠️ **矛盾：** AI可以用来优化能源使用、预测气候变化——但训练这些AI本身就在消耗大量能源。

### 6.7 伦理AI原则 (Ethical AI Principles)

负责任的AI开发应遵循以下原则：

| 原则 | 英文 | 含义 | 例子 |
|------|------|------|------|
| **公平性** | Fairness | AI不应歧视任何群体 | 贷款审批不因种族而不同 |
| **透明性** | Transparency | 用户应知道AI在做什么 | 告知用户内容是AI生成的 |
| **可解释性** | Explainability | 能解释AI为何做出某个决定 | 告诉用户为什么贷款被拒绝 |
| **问责制** | Accountability | 明确谁对AI的决定负责 | 自动驾驶事故的法律责任 |
| **隐私** | Privacy | 保护训练数据中的个人信息 | 不用个人照片训练模型而不告知 |
| **安全** | Safety | 确保AI不会造成伤害 | 医疗AI的严格测试 |

> 💡 **欧盟AI法案 (EU AI Act, 2024)** 是全球第一个全面的AI监管法规，将AI系统按风险等级分类：
> - **不可接受的风险：** 禁止（如社会评分系统）
> - **高风险：** 严格监管（如医疗AI、招聘AI）
> - **有限风险：** 透明度要求（如聊天机器人需告知用户）
> - **最低风险：** 无特别要求（如垃圾邮件过滤器）

```python
# Interactive AI Demo: Simple Sentiment Analysis
# This demonstrates how a basic rule-based "AI" works

def simple_sentiment_analyzer(text):
    """A simple rule-based sentiment analyzer to demonstrate AI concepts."""
    
    positive_words = [
        'good', 'great', 'excellent', 'amazing', 'wonderful', 'fantastic',
        'love', 'happy', 'best', 'beautiful', 'awesome', 'perfect',
        'brilliant', 'outstanding', 'superb', 'enjoy', 'like', 'nice',
        'helpful', 'useful', 'impressive', 'recommend', 'fun'
    ]
    
    negative_words = [
        'bad', 'terrible', 'horrible', 'awful', 'worst', 'hate',
        'ugly', 'disgusting', 'poor', 'boring', 'useless', 'waste',
        'annoying', 'disappointing', 'broken', 'fail', 'wrong',
        'slow', 'expensive', 'difficult', 'confusing', 'crash'
    ]
    
    negation_words = ['not', 'no', 'never', 'neither', 'nobody', 'nothing']
    
    words = text.lower().split()
    
    pos_count = 0
    neg_count = 0
    negation_active = False
    
    analysis_details = []
    
    for i, word in enumerate(words):
        # Clean punctuation
        clean_word = word.strip('.,!?;:')
        
        if clean_word in negation_words:
            negation_active = True
            analysis_details.append(f'  [NEGATION] "{clean_word}" detected - flipping next sentiment')
            continue
        
        if clean_word in positive_words:
            if negation_active:
                neg_count += 1
                analysis_details.append(f'  [-] "{clean_word}" (positive BUT negated -> negative)')
                negation_active = False
            else:
                pos_count += 1
                analysis_details.append(f'  [+] "{clean_word}" (positive word)')
        elif clean_word in negative_words:
            if negation_active:
                pos_count += 1
                analysis_details.append(f'  [+] "{clean_word}" (negative BUT negated -> positive)')
                negation_active = False
            else:
                neg_count += 1
                analysis_details.append(f'  [-] "{clean_word}" (negative word)')
        else:
            if negation_active:
                negation_active = False
    
    total = pos_count + neg_count
    if total == 0:
        sentiment = 'NEUTRAL'
        confidence = 0
        emoji = '  ->'
    else:
        score = (pos_count - neg_count) / total
        confidence = abs(score) * 100
        if score > 0.1:
            sentiment = 'POSITIVE'
            emoji = '  :)'
        elif score < -0.1:
            sentiment = 'NEGATIVE'
            emoji = '  :('
        else:
            sentiment = 'MIXED'
            confidence = 50
            emoji = '  :/'
    
    return sentiment, confidence, pos_count, neg_count, analysis_details, emoji

# Demo sentences
test_sentences = [
    'This product is amazing and I love it!',
    'Terrible experience, the software keeps crashing.',
    'The movie was not bad, actually quite enjoyable.',
    'I am not happy with this disappointing result.',
    'The weather is cloudy today.',
    'Great customer service but the product quality is poor.'
]

print('=' * 65)
print('  SIMPLE SENTIMENT ANALYZER DEMO')
print('  Demonstrating basic AI / NLP concepts')
print('=' * 65)

for sentence in test_sentences:
    sentiment, confidence, pos, neg, details, emoji = simple_sentiment_analyzer(sentence)
    print(f'\nInput: "{sentence}"')
    print(f'Analysis:')
    for d in details:
        print(d)
    print(f'Score: +{pos} / -{neg}')
    print(f'Result: {sentiment} {emoji} (confidence: {confidence:.0f}%)')
    print('-' * 50)

print('\n' + '=' * 65)
print('  LIMITATIONS OF THIS APPROACH:')
print('  - Only checks individual words (no context understanding)')
print('  - Cannot understand sarcasm or irony')
print('  - Limited vocabulary')
print('  - Real AI uses neural networks trained on millions of examples')
print('  - Modern NLP (like BERT, GPT) understands context and nuance')
print('=' * 65)
print('\nTry modifying the test_sentences list to test your own inputs!')
```

- **Permissive (如MIT)：** 你可以用它做几乎任何事，包括将修改后的版本变为专有代码。只需保留原始版权声明。
