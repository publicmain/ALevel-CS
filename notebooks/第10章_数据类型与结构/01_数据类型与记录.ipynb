## 1. 为什么数据类型很重要? (Why Data Types Matter)

### 想象一个没有数据类型的世界

假设你走进一家银行, 告诉柜员: "请把 **abc** 元存入我的账户." 柜员会困惑 --- "abc" 不是一个有效的金额!

计算机也面临同样的问题. 在底层, 所有数据都是 **0 和 1 (binary)**. 数据类型告诉计算机:

| 作用 | 解释 |
|------|------|
| **内存分配 (Memory Allocation)** | 不同类型需要不同大小的存储空间. 一个 BOOLEAN 只需要 1 bit, 而一个 REAL 需要 32 或 64 bits |
| **合法操作 (Valid Operations)** | 数字可以做加减乘除, 但对字符串做除法毫无意义 |
| **错误预防 (Error Prevention)** | 编译器/解释器可以在程序运行前就发现类型错误, 避免灾难性后果 |
| **数据解释 (Interpretation)** | 同样的二进制 `01000001` 可以是数字 65, 也可以是字符 'A' --- 类型决定了如何解读 |

### 真实案例: 1996年 Ariane 5 火箭爆炸
欧洲航天局的 Ariane 5 火箭在发射后 37 秒爆炸, 原因是一个 **64-bit 浮点数被转换为 16-bit 整数时溢出 (overflow)**. 这个类型转换错误造成了 **3.7亿美元** 的损失! 数据类型真的很重要.

````python
# 同样的二进制, 不同的解释
binary_value = 0b01000001  # 二进制 01000001

print(f"作为整数 (INTEGER): {binary_value}")
print(f"作为字符 (CHAR):    {chr(binary_value)}")
print(f"作为布尔 (BOOLEAN): {bool(binary_value)}")
print()
print("同样的 0 和 1, 数据类型决定了含义!")
````

## 2. 基本数据类型 (Fundamental Data Types)

### 2.1 INTEGER (整数类型)

**定义**: 没有小数部分的完整数字, 可以是正数、负数或零.

**为什么叫 "整数"?** 数学中的整数集合 Z = {..., -3, -2, -1, 0, 1, 2, 3, ...}, INTEGER 就是计算机对这个概念的实现.

**底层存储**: 计算机用 **二进制补码 (Two's Complement)** 来存储整数:
- 正数直接转换为二进制
- 负数用补码表示 (取反加一)

| 位数 (Bits) | 范围 | 常见用途 |
|------------|------|---------|
| 8-bit | -128 到 127 | 小计数器, ASCII 码 |
| 16-bit | -32,768 到 32,767 | 旧系统整数 |
| 32-bit | -2,147,483,648 到 2,147,483,647 | 现代系统默认 |
| 64-bit | 约 -9.2 x 10^18 到 9.2 x 10^18 | 大数计算 |

> **Python 特殊之处**: Python 的整数没有固定大小限制! 它会自动扩展. 但在 pseudocode 和大多数语言中, 整数有固定范围.

````python
# INTEGER 整数演示
age = 17
temperature = -5
population = 7900000000

print(f"年龄: {age} (类型: {type(age).__name__})")
print(f"温度: {temperature} (类型: {type(temperature).__name__})")
print(f"世界人口: {population:,}")
print()

# 二进制存储演示
num = 42
print(f"十进制 {num} 的二进制表示: {bin(num)}")
print(f"十进制 {-num} 的二进制表示: {bin(-num & 0xFF)} (8-bit 补码)")
print()

# Python 整数无限制
huge = 2 ** 100
print(f"2^100 = {huge}")
print(f"位数: {huge.bit_length()} bits")
````

### 2.2 REAL / FLOAT (实数 / 浮点数类型)

**定义**: 带有小数部分的数字.

**为什么叫 "浮点 (floating point)"?** 因为小数点的位置是 "浮动" 的!
- 固定点: `123.45` (小数点固定在某位)
- 浮点: `1.2345 x 10^2` --- 小数点可以 "浮动" 到任何位置

**IEEE 754 标准** (你不需要记住细节, 但理解概念很有用):
- 一个浮点数由三部分组成: **符号位 (sign)** + **指数 (exponent)** + **尾数 (mantissa)**
- 32-bit (单精度): 约 7 位有效数字
- 64-bit (双精度): 约 15 位有效数字

> **重要警告**: 浮点数有 **精度问题 (precision issues)**! `0.1 + 0.2` 在计算机中并不等于 `0.3`.

````python
# REAL / FLOAT 浮点数演示
pi = 3.14159265358979
price = 9.99
speed_of_light = 3.0e8  # 科学计数法: 3.0 x 10^8

print(f"圆周率: {pi} (类型: {type(pi).__name__})")
print(f"价格: {price}")
print(f"光速: {speed_of_light:,.0f} m/s")
print()

# 经典的浮点精度问题!
result = 0.1 + 0.2
print(f"0.1 + 0.2 = {result}")
print(f"0.1 + 0.2 == 0.3? {result == 0.3}")
print(f"实际值: {result:.20f}")
print()
print("这不是 Python 的 bug, 而是所有使用 IEEE 754 的语言都有的问题!")
print("原因: 0.1 在二进制中是无限循环小数, 就像 1/3 在十进制中是 0.333...")
````

### 2.3 CHAR (字符类型)

**定义**: 单个字符 --- 一个字母、数字、符号或空格.

**底层原理**: 每个字符对应一个数字编码:
- **ASCII** (American Standard Code for Information Interchange):
  - 7-bit 编码, 128 个字符 (0-127)
  - 'A' = 65, 'B' = 66, ... 'Z' = 90
  - 'a' = 97, 'b' = 98, ... 'z' = 122
  - '0' = 48, '1' = 49, ... '9' = 57
- **Unicode**: ASCII 的超集, 支持全球所有文字 (包括中文!)
  - UTF-8: 变长编码, 1-4 字节
  - 可以表示超过 14 万个字符

> **注意**: 在 Python 中没有单独的 CHAR 类型, 字符就是长度为 1 的 STRING. 但在 pseudocode 和许多语言 (如 Java, C) 中, CHAR 是独立类型.

````python
# CHAR 字符演示
letter = 'A'
digit = '7'
symbol = '@'

# 字符和 ASCII 码的转换
print("=== ASCII 码表 (部分) ===")
print(f"'{letter}' -> ASCII: {ord(letter)}")
print(f"'{digit}' -> ASCII: {ord(digit)}")
print(f"'{symbol}' -> ASCII: {ord(symbol)}")
print()

# ASCII 码转字符
print(f"ASCII 65 -> 字符: '{chr(65)}'")
print(f"ASCII 97 -> 字符: '{chr(97)}'")
print()

# 有趣的发现: 大小写字母差 32!
print("=== 大小写转换的秘密 ===")
for c in "HELLO":
    print(f"'{c}' (ASCII {ord(c)}) + 32 = '{chr(ord(c)+32)}' (ASCII {ord(c)+32})")
print()

# Unicode: 中文字符
chinese_char = '中'
print(f"'{chinese_char}' 的 Unicode 编码: U+{ord(chinese_char):04X} (十进制: {ord(chinese_char)})")
````

### 2.4 STRING (字符串类型)

**定义**: 由零个或多个字符组成的序列.

**内存中的存储**: 字符串在内存中是一个字符数组, 通常以两种方式表示结尾:
1. **Null-terminated** (C 语言): 末尾加 `\0` 字符
2. **Length-prefixed** (Python, Java): 额外存储长度信息

| 操作 | 说明 | Pseudocode |
|------|------|-----------|
| 长度 | 获取字符数 | `LENGTH(s)` |
| 连接 | 合并字符串 | `s1 & s2` |
| 截取 | 取子串 | `SUBSTRING(s, start, length)` |
| 查找 | 搜索位置 | 各语言不同 |

````python
# STRING 字符串演示
name = "Cambridge"
empty = ""  # 空字符串, 长度为 0

print(f"字符串: '{name}'")
print(f"长度: {len(name)}")
print(f"第一个字符: '{name[0]}'")
print(f"最后一个字符: '{name[-1]}'")
print()

# 字符串操作
first = "Hello"
second = "World"
combined = first + " " + second  # 连接 (concatenation)
print(f"连接: '{combined}'")
print(f"子串: '{combined[0:5]}'")
print(f"大写: '{combined.upper()}'")
print()

# 字符串是不可变的 (immutable)!
s = "Hello"
# s[0] = 'h'  # 这会报错! 字符串不能修改单个字符
s_new = 'h' + s[1:]  # 只能创建新字符串
print(f"原始: '{s}' -> 修改后: '{s_new}'")
print("字符串是 immutable 的, 每次 '修改' 都创建新对象")
````

### 2.5 BOOLEAN (布尔类型)

**定义**: 只有两个可能的值 --- `TRUE` 或 `FALSE`.

**历史**: 以英国数学家 **George Boole (1815-1864)** 命名. 他发明了布尔代数 (Boolean Algebra), 用数学方法处理逻辑推理. 这成为了现代计算机的理论基础!

**存储**: 理论上只需要 **1 bit** (0 或 1), 但实际上大多数语言使用 1 byte (8 bits) 来存储, 因为内存的最小寻址单位通常是 1 byte.

**布尔运算**:
| 运算 | 符号 | 含义 |
|------|------|------|
| AND | `AND` / `&&` | 两者都为真才为真 |
| OR | `OR` / `\|\|` | 至少一个为真就为真 |
| NOT | `NOT` / `!` | 取反 |

````python
# BOOLEAN 布尔类型演示
is_student = True
has_passed = False

print(f"是学生? {is_student} (类型: {type(is_student).__name__})")
print(f"已通过? {has_passed}")
print()

# 布尔运算
print("=== 布尔运算真值表 ===")
print(f"True AND True   = {True and True}")
print(f"True AND False  = {True and False}")
print(f"True OR False   = {True or False}")
print(f"False OR False  = {False or False}")
print(f"NOT True        = {not True}")
print(f"NOT False       = {not False}")
print()

# 布尔值本质上是 0 和 1
print("=== BOOLEAN 的数值本质 ===")
print(f"True + True = {True + True}")
print(f"True * 10 = {True * 10}")
print(f"False * 10 = {False * 10}")
print(f"int(True) = {int(True)}, int(False) = {int(False)}")
````

### 2.6 DATE (日期类型)

**定义**: 表示日历中的一个特定日期.

**格式问题**: 不同国家使用不同的日期格式, 这常常导致混淆!

| 格式 | 示例 | 使用地区 |
|------|------|---------|
| DD/MM/YYYY | 14/04/2026 | 英国, 中国, 大部分欧洲 |
| MM/DD/YYYY | 04/14/2026 | 美国 |
| YYYY-MM-DD | 2026-04-14 | ISO 8601 国际标准 |

**计算机如何存储日期?**
- 方法1: 存储为字符串 (简单但不利于计算)
- 方法2: 存储为从某个 "纪元 (epoch)" 开始的天数/秒数
  - Unix 纪元: 1970年1月1日 00:00:00 UTC
  - 当前时刻 = 从1970年至今的秒数 (一个大整数!)

> **Y2K 千年虫问题**: 1999年前, 很多程序用 2 位数存储年份 (99 = 1999). 到 2000 年时, "00" 被解释为 1900 而非 2000, 全球花费数十亿美元修复!

````python
from datetime import date, datetime
import time

# DATE 日期演示
today = date.today()
print(f"今天的日期: {today}")
print(f"年: {today.year}, 月: {today.month}, 日: {today.day}")
print()

# 不同格式输出
print("=== 不同日期格式 ===")
print(f"ISO 格式:  {today.isoformat()}")
print(f"英式格式:  {today.strftime('%d/%m/%Y')}")
print(f"美式格式:  {today.strftime('%m/%d/%Y')}")
print(f"中文格式:  {today.strftime('%Y年%m月%d日')}")
print()

# Unix 时间戳
timestamp = time.time()
print(f"=== Unix 时间戳 ===")
print(f"当前时间戳: {timestamp:.0f} 秒")
print(f"即从 1970-01-01 00:00:00 UTC 至今的秒数")
print(f"转换回日期: {datetime.fromtimestamp(timestamp)}")
````

## 3. 类型转换 / 类型铸造 (Type Conversion / Casting)

有时我们需要将一种数据类型转换为另一种. 这叫做 **类型转换 (type casting)**.

**两种转换方式**:
1. **隐式转换 (Implicit)**: 编译器自动转换, 不会丢失信息 (如 int -> float)
2. **显式转换 (Explicit)**: 程序员手动指定, 可能丢失信息 (如 float -> int)

| 转换 | 函数 | 示例 | 注意 |
|------|------|------|------|
| String -> Integer | `int()` | `int("42")` -> 42 | 字符串必须是有效数字 |
| String -> Float | `float()` | `float("3.14")` -> 3.14 | 同上 |
| Integer -> String | `str()` | `str(42)` -> "42" | 总是安全的 |
| Float -> Integer | `int()` | `int(3.7)` -> 3 | 截断, 不是四舍五入! |
| Integer -> Float | `float()` | `float(42)` -> 42.0 | 总是安全的 |
| Char -> ASCII | `ord()` | `ord('A')` -> 65 | |
| ASCII -> Char | `chr()` | `chr(65)` -> 'A' | |

````python
# 类型转换演示
print("=== 显式类型转换 (Explicit Casting) ===")

# String -> Integer
age_str = "17"
age_int = int(age_str)
print(f"int('17') = {age_int}, 类型: {type(age_int).__name__}")

# Float -> Integer (注意: 截断而非四舍五入!)
pi = 3.99
pi_int = int(pi)
print(f"int(3.99) = {pi_int} (截断, 不是四舍五入!)")
print(f"round(3.99) = {round(pi)} (这才是四舍五入)")

# Integer -> String
num = 42
num_str = str(num)
print(f"str(42) = '{num_str}', 类型: {type(num_str).__name__}")
print()

# 类型转换可能失败!
print("=== 转换失败示例 ===")
try:
    result = int("hello")
except ValueError as e:
    print(f"int('hello') 失败! 错误: {e}")

try:
    result = int("3.14")
except ValueError as e:
    print(f"int('3.14') 失败! 需要先转为 float: int(float('3.14')) = {int(float('3.14'))}")
print()

# 隐式转换
print("=== 隐式转换 (Implicit) ===")
result = 5 + 3.0  # int + float -> float
print(f"5 + 3.0 = {result}, 类型: {type(result).__name__}")
result2 = "Score: " + str(95)  # Python 不会自动将 int 转为 str
print(f"'Score: ' + str(95) = '{result2}'")
````

## 4. 记录结构 (Record / Composite Data Type)

### 什么是 Record?

**Record (记录)** 是一种 **复合数据类型 (composite data type)**, 它将多个 **不同类型** 的相关数据组合在一起, 形成一个逻辑单元.

### 现实世界的类比

想想一张 **学生信息卡**:

```
+-----------------------------------+
|  学生信息卡                        |
|  -------------------------------- |
|  姓名 (STRING):   张三             |
|  年龄 (INTEGER):  17              |
|  成绩 (REAL):     85.5            |
|  在校 (BOOLEAN):  TRUE            |
|  入学日期 (DATE): 01/09/2024      |
+-----------------------------------+
```

这张卡片上有不同类型的信息, 但它们都属于同一个学生. Record 就是这样的结构!

### 为什么需要 Record?
- **组织性**: 相关数据放在一起, 代码更清晰
- **可读性**: `student.name` 比 `name_array[i]` 更易懂
- **可维护性**: 添加新字段只需修改 Record 定义

### Pseudocode 中的 Record 定义

```
// 定义 Record 类型
TYPE StudentRecord
    DECLARE Name : STRING
    DECLARE Age : INTEGER
    DECLARE Grade : REAL
    DECLARE IsEnrolled : BOOLEAN
ENDTYPE

// 声明 Record 变量
DECLARE Student1 : StudentRecord

// 赋值
Student1.Name <- "Zhang San"
Student1.Age <- 17
Student1.Grade <- 85.5
Student1.IsEnrolled <- TRUE

// 读取
OUTPUT Student1.Name
OUTPUT Student1.Grade
```

**要点**:
- `TYPE...ENDTYPE` 定义记录的结构 (模板)
- 用 `.` (点号) 访问记录中的字段
- 每个字段有自己的名称和类型
- 可以创建多个同类型的记录变量

````python
# Python 中实现 Record --- 方法1: dataclass (推荐)
from dataclasses import dataclass

@dataclass
class StudentRecord:
    name: str
    age: int
    grade: float
    is_enrolled: bool

# 创建 Record 实例
student1 = StudentRecord(name="Zhang San", age=17, grade=85.5, is_enrolled=True)
student2 = StudentRecord(name="Li Si", age=16, grade=92.0, is_enrolled=True)

print("=== Student Records ===")
print(f"学生1: {student1}")
print(f"学生2: {student2}")
print()

# 访问字段 (用点号, 就像 pseudocode 一样!)
print(f"学生1的姓名: {student1.name}")
print(f"学生1的成绩: {student1.grade}")
print()

# 修改字段
student1.grade = 90.0
print(f"学生1更新后的成绩: {student1.grade}")
````

````python
# Python 中实现 Record --- 方法2: namedtuple (不可变)
from collections import namedtuple

# 定义 Record 类型
BookRecord = namedtuple('BookRecord', ['title', 'author', 'price', 'pages'])

# 创建实例
book1 = BookRecord(title="Python Programming", author="John", price=29.99, pages=350)
book2 = BookRecord(title="Data Structures", author="Jane", price=35.50, pages=420)

print("=== Book Records ===")
print(f"Book 1: {book1}")
print(f"书名: {book1.title}")
print(f"价格: ${book1.price}")
print()

# namedtuple 是不可变的! (immutable)
try:
    book1.price = 19.99  # 这会报错!
except AttributeError as e:
    print(f"修改失败: {e}")
    print("namedtuple 创建后不能修改, 适合用于不变的记录")
````

````python
# Record 数组: 存储多条记录
from dataclasses import dataclass

@dataclass
class Employee:
    emp_id: int
    name: str
    department: str
    salary: float

# 创建 Record 数组
employees = [
    Employee(1001, "Alice Wang", "Engineering", 85000),
    Employee(1002, "Bob Chen", "Marketing", 72000),
    Employee(1003, "Carol Liu", "Engineering", 91000),
    Employee(1004, "David Zhang", "HR", 68000),
    Employee(1005, "Eva Li", "Engineering", 88000),
]

# 遍历和操作
print("=== 员工列表 ===")
print(f"{'ID':<6} {'姓名':<15} {'部门':<15} {'薪资':>10}")
print("-" * 50)
for emp in employees:
    print(f"{emp.emp_id:<6} {emp.name:<15} {emp.department:<15} ${emp.salary:>9,.0f}")

print()

# 按部门筛选
eng_team = [e for e in employees if e.department == "Engineering"]
print(f"Engineering 部门共 {len(eng_team)} 人")
avg_salary = sum(e.salary for e in eng_team) / len(eng_team)
print(f"Engineering 平均薪资: ${avg_salary:,.0f}")
````

## 5. 数据类型总结对照表

| 数据类型 | 描述 | 存储大小 | 示例值 | 使用场景 |
|---------|------|---------|--------|---------|
| **INTEGER** | 整数 | 通常 4 bytes | 42, -17, 0 | 计数, 索引, 年龄 |
| **REAL** | 浮点数 | 4 或 8 bytes | 3.14, -0.5 | 价格, 温度, 科学计算 |
| **CHAR** | 单个字符 | 1-4 bytes | 'A', '7', '@' | 等级(A/B/C), 性别(M/F) |
| **STRING** | 字符序列 | 可变 | "Hello" | 姓名, 地址, 文本 |
| **BOOLEAN** | 真/假 | 1 byte | TRUE, FALSE | 开关状态, 条件标志 |
| **DATE** | 日期 | 视实现 | 14/04/2026 | 生日, 截止日期 |
| **RECORD** | 复合类型 | 各字段之和 | Student{...} | 组织相关数据 |

````python
# 练习参考答案 (先自己尝试!)

# 练习 2 答案:
print("=== 练习 2: 类型转换结果 ===")
print(f"1. int(7.9) = {int(7.9)}")
print(f"2. str(100) = '{str(100)}'")
print(f"3. float('3') = {float('3')}")
print(f"4. bool(0) = {bool(0)}")
print(f"5. ord('Z') = {ord('Z')}")
print(f"6. chr(48) = '{chr(48)}'")
print()

# 练习 3 参考答案:
# Pseudocode:
# TYPE BookRecord
#     DECLARE Title : STRING
#     DECLARE Author : STRING
#     DECLARE ISBN : STRING
#     DECLARE YearPublished : INTEGER
#     DECLARE IsLent : BOOLEAN
#     DECLARE Price : REAL
# ENDTYPE

from dataclasses import dataclass

@dataclass
class BookRecord:
    title: str
    author: str
    isbn: str
    year_published: int
    is_lent: bool
    price: float

book = BookRecord(
    title="Introduction to Algorithms",
    author="Cormen et al.",
    isbn="978-0262033848",
    year_published=2009,
    is_lent=False,
    price=75.00
)
print("=== 练习 3: BookRecord ===")
print(book)
````
