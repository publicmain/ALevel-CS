## 1. 什么是计算思维 (Computational Thinking)？

### 1.1 定义

> **计算思维**是一种用**计算机科学的方法**来思考和解决问题的思维方式。
> 它不仅仅是“会写代码”，而是一种**系统化的问题解决框架**。

想象你是一个建筑师，面对客户说“我想要一栋房子”：
- 你不会立刻开始搬砖（写代码）
- 你会先**理解需求**、**画设计图**、**分解工序**、**选择合适的材料**

计算思维就是计算机科学家的“建筑设计能力”。

### 1.2 计算思维的四大核心技能

```
        计算思维 (Computational Thinking)
        ┌──────────┬──────────┬──────────┐
        │          │          │          │
   抽象 Abstraction  分解 Decomposition  模式识别 Pattern Recognition
        │          │          │          │
        └──────────┴──────────┴──────────┘
                       │
              算法设计 Algorithm Design
```

| 技能 | 英文 | 核心思想 |
|------|------|----------|
| 抽象 | Abstraction | 忽略不必要的细节，保留关键信息 |
| 分解 | Decomposition | 把大问题拆成小问题 |
| 模式识别 | Pattern Recognition | 发现问题之间的相似性 |
| 算法设计 | Algorithm Design | 设计一步步的解决方案 |

### 1.3 为什么计算思维是 CS 的核心？

**计算思维 ≠ 编程。** 编程只是把计算思维的结果“翻译”成计算机能理解的语言。

真正困难的从来不是语法，而是**如何思考问题**。一个优秀的软件工程师，
70% 的时间在思考和设计，30% 的时间在编码。

> **CS 冷知识**: Google、Amazon 等科技巨头的面试中，计算思维能力（而非特定语言的知识）
> 是最重要的考察内容。这就是为什么面试题往往是“设计一个系统”或“解决一个抽象问题”。

## 2. 抽象 (Abstraction)

### 2.1 什么是抽象？

> **抽象**就是**隐藏不必要的细节，只保留解决问题所需要的关键信息**。

这是人类思维中最自然的能力之一。你每天都在使用抽象：

| 现实场景 | 你关注的（保留） | 你忽略的（隐藏） |
|----------|------------------|------------------|
| 开车 | 方向盘、油门、刹车 | 发动机内部如何燃烧汽油 |
| 用手机 | 触摸屏上的图标 | CPU 如何处理指令 |
| 看地图 | 道路和目的地 | 每棵树、每栋楼的位置 |

### 2.2 经典案例：伦敦地铁图 (London Underground Map)

1933年，Harry Beck 设计了著名的伦敦地铁图。这是 **抽象** 的完美范例：

**真实地理 vs 地铁图：**

| 方面 | 真实地理 | 地铁图（抽象后） |
|------|----------|------------------|
| 站点距离 | 精确距离（有的近有的远） | 等间距排列 |
| 线路走向 | 弯弯曲曲 | 只用水平、垂直、45° 线 |
| 地面建筑 | 密密麻麻的街道和建筑 | 完全不显示 |
| 保留信息 | — | 站名、换乘关系、线路颜色 |

**为什么这样做？** 因为乘客只需要知道：
1. 我在哪一站？
2. 我要去哪一站？
3. 在哪里换乘？

地面上的真实距离和方向？完全不重要！

### 2.3 为什么抽象是必要的？

1. **管理复杂性 (Manage Complexity)**: 现实世界太复杂，无法一次性处理所有信息
2. **聚焦核心问题**: 忽略干扰信息，集中精力解决关键问题
3. **使沟通更高效**: 团队成员可以在同一抽象层次上讨论
4. **支持分层设计**: 不同层次的抽象处理不同层次的问题

> **深度理解**: 计算机科学本身就是一层层的抽象！
> - 最底层：电信号（0和1）
> - 机器码：用二进制表示指令
> - 汇编语言：用助记符代替二进制
> - 高级语言：用接近人类语言的方式编程
> - 应用软件：用户只看到按钮和界面
>
> 每一层都在前一层的基础上做抽象！

### 2.4 实践：为图书馆系统建立抽象模型

假设你要设计一个 **图书馆管理系统 (Library Management System)**。

真实的图书馆有无数细节：
- 书架的颜色、材质
- 图书管理员的衣服
- 走廊的宽度
- 灯光的亮度
- 每本书的重量

但对于一个 **管理系统** 来说，我们需要抽象——只保留相关信息：

下面的代码会帮你分析哪些信息应该保留，哪些应该忽略：

````python
# 图书馆系统抽象模型 - 交互式分析

print('=' * 60)
print('  图书馆管理系统 - 抽象分析')
print('=' * 60)

# All possible attributes
attributes = [
    ('书名 (Book Title)', True, '需要知道这是什么书'),
    ('作者 (Author)', True, '需要按作者搜索'),
    ('ISBN编号', True, '唯一标识每本书'),
    ('出版日期 (Publish Date)', True, '管理版本信息'),
    ('是否已借出 (Is Borrowed)', True, '核心业务逻辑'),
    ('借阅者姓名 (Borrower Name)', True, '追踪谁借了书'),
    ('借阅日期 (Borrow Date)', True, '计算是否逾期'),
    ('书的重量 (Book Weight)', False, '与借阅管理无关'),
    ('书架颜色 (Shelf Color)', False, '纯装饰信息'),
    ('书的纸张厚度 (Paper Thickness)', False, '与管理无关'),
    ('图书馆大楼高度 (Building Height)', False, '与软件系统无关'),
    ('管理员衣服颜色', False, '与系统功能无关'),
]

print()
print('分析每个属性是否应该保留在抽象模型中：')
print('-' * 60)

keep = []
discard = []

for attr, relevant, reason in attributes:
    status = '✓ 保留' if relevant else '✗ 丢弃'
    symbol = '🟢' if relevant else '🔴'
    print(f'  {symbol} {attr:30s} → {status}  ({reason})')
    if relevant:
        keep.append(attr)
    else:
        discard.append(attr)

print()
print(f'保留的属性数量: {len(keep)}')
print(f'丢弃的属性数量: {len(discard)}')
print()
print('抽象模型（只包含关键属性）:')
print('┌─────────────────────────────────┐')
print('│     Library Book (图书)          │')
print('├─────────────────────────────────┤')
for attr in keep:
    print(f'│  - {attr:29s}│')
print('└─────────────────────────────────┘')
````

## 3. 分解 (Decomposition)

### 3.1 什么是分解？

> **分解**是将一个**大的、复杂的问题**拆分成**更小的、更容易管理的子问题**。

### 3.2 现实世界类比：建造一栋房子

没有人会说“来，咱们一口气把房子盖好”。建房子的过程是：

```
                    建造房子
         ┌──────────┼──────────┐
         │          │          │
      基础工程    主体工程    装修工程
      ┌──┴──┐   ┌──┴──┐   ┌──┴──┐
      │     │   │     │   │     │
    挖地基 浇水泥 砌墙 盖屋顶 水电 粉刷
```

每个子问题：
- **更容易理解**（“砌墙”比“建房子”简单得多）
- **可以分配给不同的人**（水电工做水电，油漆工做粉刷）
- **可以独立测试**（检查地基是否牢固，不用等整栋楼盖完）

### 3.3 分解在编程中的意义

分解直接导致了编程中的核心概念：

| 分解的结果 | 编程概念 | 说明 |
|-----------|---------|------|
| 子问题 | **模块 (Module)** | 独立的代码单元 |
| 子任务 | **过程 (Procedure)** | 执行特定任务的代码块 |
| 子计算 | **函数 (Function)** | 接收输入、返回输出的代码块 |

> **深度理解**: 分解不仅是一种设计方法，它还体现了软件工程的两大原则：
> 1. **单一职责原则 (Single Responsibility Principle)**: 每个模块只做一件事
> 2. **分而治之 (Divide and Conquer)**: 这是算法设计中最重要的策略之一！

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(1, 1, figsize=(16, 10))
ax.set_xlim(0, 16)
ax.set_ylim(0, 10)
ax.axis('off')
ax.set_title('分解树 (Decomposition Tree): 学生成绩管理系统', fontsize=16, fontweight='bold', pad=20)

def draw_box(ax, x, y, text, color='#4FC3F7', w=2.2, h=0.7):
    rect = patches.FancyBboxPatch((x - w/2, y - h/2), w, h,
        boxstyle='round,pad=0.1', facecolor=color, edgecolor='#333', linewidth=1.5)
    ax.add_patch(rect)
    ax.text(x, y, text, ha='center', va='center', fontsize=9, fontweight='bold')

def draw_line(ax, x1, y1, x2, y2):
    ax.plot([x1, x2], [y1, y2], 'k-', linewidth=1.2)

# Level 0: Root
draw_box(ax, 8, 9, '学生成绩管理系统', '#FF8A65', w=3, h=0.8)

# Level 1: Main modules
level1 = [
    (2.5, 7, '学生管理'),
    (6, 7, '成绩管理'),
    (9.5, 7, '报告生成'),
    (13, 7, '用户界面'),
]
for x, y, text in level1:
    draw_box(ax, x, y, text, '#FFF176')
    draw_line(ax, 8, 9 - 0.4, x, y + 0.35)

# Level 2: Sub-modules
level2_groups = [
    # Under 学生管理
    [(1, 5, '添加学生'), (2.5, 5, '删除学生'), (4, 5, '修改信息')],
    # Under 成绩管理
    [(5, 5, '录入成绩'), (6, 5, '修改成绩'), (7, 5, '查询成绩')],
    # Under 报告生成
    [(8.5, 5, '成绩排名'), (9.5, 5, '平均分'), (10.5, 5, '成绩单')],
    # Under 用户界面
    [(12, 5, '登录界面'), (13, 5, '主菜单'), (14, 5, '数据显示')],
]

parents_x = [2.5, 6, 9.5, 13]
for gi, group in enumerate(level2_groups):
    for x, y, text in group:
        draw_box(ax, x, y, text, '#C5E1A5', w=1.8, h=0.6)
        draw_line(ax, parents_x[gi], 7 - 0.35, x, y + 0.3)

# Level 3: Even finer (under 录入成绩)
level3 = [
    (4, 3, '验证输入'),
    (5, 3, '保存数据'),
    (6, 3, '确认反馈'),
]
for x, y, text in level3:
    draw_box(ax, x, y, text, '#E1BEE7', w=1.6, h=0.6)
    draw_line(ax, 5, 5 - 0.3, x, y + 0.3)

# Legend
ax.text(1, 1.5, '每一层都在上一层的基础上进一步细化', fontsize=11, color='#555')
ax.text(1, 0.8, '分解到每个子模块都足够简单、可以直接编码为止', fontsize=11, color='#555')

plt.tight_layout()
plt.show()
````

### 3.4 实践：分解“学生管理系统”

让我们一步步分解一个学生管理系统，体会分解的过程：

````python
# 交互式分解演示

def show_decomposition():
    system = {
        '学生管理系统 (Student Management System)': {
            '1. 学生信息模块 (Student Info Module)': [
                '1.1 注册新学生 (Register Student)',
                '1.2 更新学生信息 (Update Info)',
                '1.3 删除学生记录 (Delete Record)',
                '1.4 搜索学生 (Search Student)',
            ],
            '2. 课程管理模块 (Course Module)': [
                '2.1 添加课程 (Add Course)',
                '2.2 学生选课 (Enroll Student)',
                '2.3 退课 (Drop Course)',
            ],
            '3. 成绩模块 (Grade Module)': [
                '3.1 录入成绩 (Enter Grades)',
                '3.2 计算平均分 (Calculate Average)',
                '3.3 生成排名 (Generate Ranking)',
            ],
            '4. 报告模块 (Report Module)': [
                '4.1 个人成绩单 (Individual Report)',
                '4.2 班级汇总 (Class Summary)',
                '4.3 统计图表 (Statistics Charts)',
            ],
        }
    }

    print('分解过程演示:')
    print('=' * 60)

    for system_name, modules in system.items():
        print(f'\n  {system_name}')
        print(f'  {"─" * 50}')

        for module_name, sub_tasks in modules.items():
            print(f'\n    ├── {module_name}')
            for i, task in enumerate(sub_tasks):
                if i == len(sub_tasks) - 1:
                    print(f'    │   └── {task}')
                else:
                    print(f'    │   ├── {task}')

    print('\n' + '=' * 60)
    print('\n关键原则:')
    print('  1. 每个模块有清晰的职责')
    print('  2. 模块之间相对独立')
    print('  3. 每个子任务足够小，可以直接实现')
    print('  4. 子任务的名字就暗示了需要的函数/过程')

show_decomposition()
````

## 4. 模式识别 (Pattern Recognition) —— 额外拓展

### 4.1 什么是模式识别？

> **模式识别**是在不同问题中发现**相似性和共同模式**的能力。

如果你发现两个问题有相似的结构，那么一个问题的解决方案可能可以**复用**到另一个问题上。

### 4.2 例子

| 问题 A | 问题 B | 共同模式 |
|--------|--------|----------|
| 在一组数中找最大值 | 在一组字符串中找最长的 | **遍历 + 比较 + 记录最优** |
| 图书馆管理系统 | 视频店管理系统 | **物品登记 + 借出/归还 + 逾期检查** |
| 计算购物车总价 | 计算学生总分 | **遍历 + 累加** |

### 4.3 为什么模式识别很重要？

1. **避免重复劳动**: 不用每次都从零开始
2. **利用已有经验**: 已经解决过的问题模式可以直接套用
3. **设计模式 (Design Patterns)**: 软件工程中有一整套“设计模式”，就是对常见问题模式的标准解决方案

> **CS 冷知识**: 著名的《设计模式：可复用面向对象软件的基础》(Design Patterns, GoF)
> 一书总结了 23 种经典设计模式，至今仍是软件工程师的必读书目。

````python
# 模式识别示例：相同的模式，不同的问题

print('模式识别示例："遍历 + 比较 + 记录最优" 模式')
print('=' * 50)

# Problem 1: Find maximum number
print('\n--- 问题1: 找最大数 ---')
numbers = [34, 67, 12, 89, 45, 23, 78]
print(f'数据: {numbers}')

best = numbers[0]  # pattern: initialize with first
for item in numbers:  # pattern: traverse
    if item > best:  # pattern: compare
        best = item  # pattern: update best
print(f'最大值: {best}')

# Problem 2: Find longest string
print('\n--- 问题2: 找最长字符串 ---')
words = ['cat', 'elephant', 'dog', 'hippopotamus', 'ant']
print(f'数据: {words}')

best = words[0]  # SAME pattern!
for item in words:
    if len(item) > len(best):
        best = item
print(f'最长单词: {best}')

# Problem 3: Find student with highest score
print('\n--- 问题3: 找最高分学生 ---')
students = [('Alice', 85), ('Bob', 92), ('Charlie', 78), ('Diana', 96)]
print(f'数据: {students}')

best = students[0]  # SAME pattern!
for item in students:
    if item[1] > best[1]:
        best = item
print(f'最高分: {best[0]} ({best[1]}分)')

print('\n' + '=' * 50)
print('发现了吗？三个问题的代码结构几乎一模一样！')
print('这就是模式识别的价值——一旦你掌握了这个模式，')
print('所有"找最优"的问题都可以用同样的方法解决。')
````

## 5. 算法设计 (Algorithm Design)

### 5.1 什么是算法 (Algorithm)？

> **算法是一组有限的、明确的、有效的指令，用来解决特定问题或完成特定任务。**

生活中的算法无处不在：
- 食谱就是一个算法（步骤明确，最终做出一道菜）
- 导航路线就是一个算法（一步步告诉你怎么走）
- 洗衣机的工作流程也是一个算法

### 5.2 好算法的三个特征

| 特征 | 英文 | 含义 | 反例 |
|------|------|------|------|
| **有限性** | Finite | 必须在有限步骤内结束 | 无限循环 |
| **明确性** | Unambiguous | 每一步都有且只有一种理解方式 | “加一些盐” |
| **有效性** | Effective | 每一步都是可执行的 | “计算一个不可计算的数” |

### 5.3 表达算法的三种方式

在 Cambridge 9618 考试中，你需要掌握三种表达算法的方式：

| 方式 | 英文 | 特点 | 适用场景 |
|------|------|------|----------|
| 结构化英语 | Structured English | 接近自然语言，易读 | 初步描述，与非技术人员沟通 |
| 流程图 | Flowchart | 图形化，直观 | 展示程序流程和判断逻辑 |
| 伪代码 | Pseudocode | 接近代码，精确 | 详细设计，准备编码 |

下面我们用一个简单的例子来展示三种表达方式：

````python
# 用三种方式表达同一个算法：判断一个数是正数、负数还是零

print('算法：判断一个数是正数、负数还是零')
print('=' * 60)

print('\n1. 结构化英语 (Structured English):')
print('-' * 40)
structured_english = [
    'GET the number from the user',
    'IF the number is greater than zero THEN',
    '    DISPLAY "The number is positive"',
    'ELSE IF the number is less than zero THEN',
    '    DISPLAY "The number is negative"',
    'ELSE',
    '    DISPLAY "The number is zero"',
    'ENDIF',
]
for line in structured_english:
    print(f'  {line}')

print('\n2. 伪代码 (Pseudocode - Cambridge 9618 style):')
print('-' * 40)
pseudocode = [
    'INPUT Number',
    'IF Number > 0',
    '  THEN',
    '    OUTPUT "Positive"',
    '  ELSE',
    '    IF Number < 0',
    '      THEN',
    '        OUTPUT "Negative"',
    '      ELSE',
    '        OUTPUT "Zero"',
    '    ENDIF',
    'ENDIF',
]
for line in pseudocode:
    print(f'  {line}')

print('\n3. 流程图 (Flowchart) - 见下方图形')
print('-' * 40)
print('（使用 matplotlib 绘制，见下一个单元格）')
````

````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots(1, 1, figsize=(8, 12))
ax.set_xlim(0, 10)
ax.set_ylim(0, 14)
ax.axis('off')
ax.set_title('流程图: 判断正数/负数/零', fontsize=14, fontweight='bold')

# Helper functions
def draw_oval(ax, x, y, text, w=2, h=0.6):
    oval = mpatches.FancyBboxPatch((x-w/2, y-h/2), w, h,
        boxstyle='round,pad=0.2', facecolor='#FFCDD2', edgecolor='black', linewidth=1.5)
    ax.add_patch(oval)
    ax.text(x, y, text, ha='center', va='center', fontsize=10, fontweight='bold')

def draw_rect(ax, x, y, text, w=2.5, h=0.6):
    rect = mpatches.FancyBboxPatch((x-w/2, y-h/2), w, h,
        boxstyle='square,pad=0.1', facecolor='#C8E6C9', edgecolor='black', linewidth=1.5)
    ax.add_patch(rect)
    ax.text(x, y, text, ha='center', va='center', fontsize=9)

def draw_diamond(ax, x, y, text, size=0.8):
    diamond = plt.Polygon([(x, y+size), (x+size*1.5, y), (x, y-size), (x-size*1.5, y)],
        facecolor='#FFF9C4', edgecolor='black', linewidth=1.5)
    ax.add_patch(diamond)
    ax.text(x, y, text, ha='center', va='center', fontsize=9)

def draw_parallelogram(ax, x, y, text, w=2.5, h=0.6):
    offset = 0.3
    para = plt.Polygon([
        (x-w/2+offset, y-h/2), (x+w/2+offset, y-h/2),
        (x+w/2-offset, y+h/2), (x-w/2-offset, y+h/2)
    ], facecolor='#BBDEFB', edgecolor='black', linewidth=1.5)
    ax.add_patch(para)
    ax.text(x, y, text, ha='center', va='center', fontsize=9)

def arrow(ax, x1, y1, x2, y2):
    ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
        arrowprops=dict(arrowstyle='->', lw=1.5))

# Draw flowchart
draw_oval(ax, 5, 13, 'START')
arrow(ax, 5, 12.7, 5, 12.1)

draw_parallelogram(ax, 5, 11.8, 'INPUT Number')
arrow(ax, 5, 11.5, 5, 10.5)

draw_diamond(ax, 5, 9.7, 'Number > 0?')
arrow(ax, 6.2, 9.7, 8, 9.7)
ax.text(6.5, 9.9, 'Yes', fontsize=9, color='green')
draw_parallelogram(ax, 8, 9.7, 'OUTPUT\n"Positive"', w=2, h=0.6)

arrow(ax, 5, 8.9, 5, 8.0)
ax.text(5.15, 8.5, 'No', fontsize=9, color='red')

draw_diamond(ax, 5, 7.2, 'Number < 0?')
arrow(ax, 6.2, 7.2, 8, 7.2)
ax.text(6.5, 7.4, 'Yes', fontsize=9, color='green')
draw_parallelogram(ax, 8, 7.2, 'OUTPUT\n"Negative"', w=2, h=0.6)

arrow(ax, 5, 6.4, 5, 5.5)
ax.text(5.15, 6.0, 'No', fontsize=9, color='red')

draw_parallelogram(ax, 5, 5.2, 'OUTPUT "Zero"')

# All paths converge
arrow(ax, 8, 9.1, 8, 4.0)
arrow(ax, 8, 6.6, 8, 4.0)
arrow(ax, 5, 4.9, 5, 4.0)
ax.plot([5, 8], [4.0, 4.0], 'k-', linewidth=1.5)
ax.plot([6.5, 6.5], [4.0, 3.5], 'k-', linewidth=1.5)
arrow(ax, 6.5, 3.5, 6.5, 3.0)

draw_oval(ax, 6.5, 2.7, 'END')

# Legend
ax.text(0.5, 1.5, '图例:', fontsize=10, fontweight='bold')
draw_oval(ax, 1.5, 0.8, '终端', w=1.2, h=0.4)
draw_rect(ax, 3.5, 0.8, '处理', w=1.2, h=0.4)
draw_diamond(ax, 5.5, 0.8, '判断', size=0.35)
draw_parallelogram(ax, 7.8, 0.8, '输入/输出', w=1.5, h=0.4)

plt.tight_layout()
plt.show()
````

### 5.4 将算法转化为 Python 代码

下面我们把上面的算法直接用 Python 实现，你会发现伪代码和 Python 非常接近：

````python
# 将伪代码转化为 Python 代码

# Pseudocode:               # Python:
# INPUT Number               number = int(input(...))
# IF Number > 0 THEN         if number > 0:
#   OUTPUT "Positive"           print("Positive")
# ELSE                       else:
#   IF Number < 0 THEN           if number < 0:
#     OUTPUT "Negative"              print("Negative")
#   ELSE                         else:
#     OUTPUT "Zero"                  print("Zero")

# 用几个测试值来演示
test_values = [42, -7, 0, 100, -33]

print('测试算法:')
print('-' * 30)
for number in test_values:
    if number > 0:
        result = 'Positive (正数)'
    elif number < 0:
        result = 'Negative (负数)'
    else:
        result = 'Zero (零)'
    print(f'  Number = {number:5d}  =>  {result}')
````
