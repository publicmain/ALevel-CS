## 1. 标识符表 (Identifier Table)

### 1.1 什么是标识符表？

> **标识符表**是一个记录算法中所有变量、常量名称及其信息的表格。

在开始写算法之前，专业的做法是先列出你会用到的所有“名字”——
就像做饭之前先把所有食材摆在台面上。

### 1.2 标识符表的格式

| Identifier (标识符) | Data Type (数据类型) | Description (描述) |
|---------------------|---------------------|--------------------|
| `StudentName` | STRING | 学生的姓名 |
| `Mark` | INTEGER | 学生的分数 (0-100) |
| `Average` | REAL | 平均分 |
| `Grade` | CHAR | 等级 (A/B/C/D/F) |
| `IsPassed` | BOOLEAN | 是否及格 |

### 1.3 为什么标识符表很重要？

1. **文档化 (Documentation)**: 其他人（包括未来的你）能快速理解每个变量的用途
2. **避免混乱**: 提前规划好变量名，避免随意命名导致的混乱
3. **类型安全**: 明确每个变量的数据类型，减少类型错误
4. **考试要求**: Cambridge 9618 考试经常要求你写出标识符表！

### 1.4 命名规范

好的标识符命名应该是 **自说明的 (self-documenting)**：

| 差的命名 | 好的命名 | 原因 |
|----------|----------|------|
| `x` | `StudentAge` | 含义清晰 |
| `temp` | `MaxTemperature` | 具体明确 |
| `n` | `NumberOfStudents` | 读得懂 |
| `flag` | `IsValidInput` | 表达了布尔含义 |

## 2. 伪代码基础 (Pseudocode Fundamentals)

### 2.1 什么是伪代码？

> **伪代码 (Pseudocode)** 是一种介于自然语言和编程语言之间的“假代码”。
> 它使用编程语言的结构，但不受特定语言语法的限制。

**“伪”的含义**: 伪代码不是真正的代码，不能直接在计算机上运行，
但它足够精确，可以很容易地翻译成任何编程语言。

### 2.2 Cambridge 9618 伪代码的基本语法

#### 输入和输出

| 操作 | 伪代码 | Python 等价 |
|------|--------|------------|
| 输入 | `INPUT Variable` | `variable = input()` |
| 输入 | `READ Variable` | `variable = input()` |
| 输出 | `OUTPUT Expression` | `print(expression)` |
| 输出 | `PRINT Expression` | `print(expression)` |

#### 赋值

| 操作 | 伪代码 | Python 等价 |
|------|--------|------------|
| 赋值 | `Variable ← Expression` | `variable = expression` |

注意：Cambridge 使用左箭头 `←` 表示赋值，而不是 `=`。
这是为了区分“赋值”和“相等比较”。

> **深度理解**: 为什么不用 `=`？因为在数学中 `x = x + 1` 是不可能的（方程无解），
> 但在编程中 `x ← x + 1` 很清楚地表示“把 x+1 的值赋给 x”。
> 使用 `←` 避免了这种歧义。

## 3. 三种基本控制结构

任何算法都可以用三种基本结构组合而成——这是计算机科学的一个基本定理
（Bohm-Jacopini 定理，1966年）。

### 3.1 顺序 (Sequence)

> **顺序**是最简单的结构：指令按从上到下的顺序，一条接一条执行。

```
伪代码：                   流程图：
                          ┌─────────┐
Step1                     │  Step1  │
Step2                     └────┬────┘
Step3                          │
                          ┌────▼────┐
                          │  Step2  │
                          └────┬────┘
                               │
                          ┌────▼────┐
                          │  Step3  │
                          └─────────┘
```

**示例**: 计算矩形面积

```
INPUT Length
INPUT Width
Area ← Length * Width
OUTPUT "The area is: ", Area
```

````python
# 顺序结构示例：计算矩形面积

print('顺序结构 (Sequence) 示例')
print('=' * 40)
print()

# 模拟伪代码执行过程
Length = 8
Width = 5
print(f'INPUT Length    =>  Length = {Length}')
print(f'INPUT Width     =>  Width = {Width}')

Area = Length * Width
print(f'Area <- Length * Width  =>  Area = {Length} * {Width} = {Area}')
print(f'OUTPUT Area     =>  The area is: {Area}')
print()
print('每条语句按顺序执行，没有跳转，没有判断。')
````

### 3.2 选择 (Selection)

> **选择**结构根据条件的真假，选择执行不同的代码分支。

#### IF...THEN...ELSE...ENDIF

```
IF condition
  THEN
    statements (when TRUE)
  ELSE
    statements (when FALSE)
ENDIF
```

也可以嵌套（nested）：

```
IF condition1
  THEN
    statements
  ELSE
    IF condition2
      THEN
        statements
      ELSE
        statements
    ENDIF
ENDIF
```

#### CASE OF...ENDCASE

当有多个可能的值需要判断时，使用 CASE 比多层 IF 更清晰：

```
CASE OF Variable
  value1 : statements
  value2 : statements
  value3 : statements
  OTHERWISE : statements
ENDCASE
```

> **类比**: IF 就像“岔路口”——你看到路标（条件），然后选择左走还是右走。
> CASE 就像“转盘”——根据你要去的方向（变量的值），选择对应的出口。

````python
# 选择结构示例

print('选择结构 (Selection) 示例')
print('=' * 50)

# Example 1: IF...THEN...ELSE
print('\n--- 示例1: IF...THEN...ELSE ---')
print('伪代码:')
print('  INPUT Mark')
print('  IF Mark >= 60')
print('    THEN')
print('      OUTPUT "Pass"')
print('    ELSE')
print('      OUTPUT "Fail"')
print('  ENDIF')
print()

# Python execution
test_marks = [85, 60, 45, 100, 33]
print('Python 执行结果:')
for mark in test_marks:
    if mark >= 60:
        result = 'Pass'
    else:
        result = 'Fail'
    print(f'  Mark = {mark:3d}  =>  {result}')

# Example 2: CASE OF
print('\n--- 示例2: CASE OF ---')
print('伪代码:')
print('  INPUT Grade')
print('  CASE OF Grade')
print('    "A" : OUTPUT "Excellent"')
print('    "B" : OUTPUT "Good"')
print('    "C" : OUTPUT "Satisfactory"')
print('    "D" : OUTPUT "Needs Improvement"')
print('    OTHERWISE : OUTPUT "Invalid Grade"')
print('  ENDCASE')
print()

# Python execution using match (or if-elif)
test_grades = ['A', 'B', 'C', 'D', 'X']
print('Python 执行结果:')
for grade in test_grades:
    if grade == 'A':
        msg = 'Excellent'
    elif grade == 'B':
        msg = 'Good'
    elif grade == 'C':
        msg = 'Satisfactory'
    elif grade == 'D':
        msg = 'Needs Improvement'
    else:
        msg = 'Invalid Grade'
    print(f'  Grade = "{grade}"  =>  {msg}')
````

### 3.3 循环 (Iteration)

> **循环**结构允许一组语句重复执行多次。

Cambridge 9618 伪代码有三种循环：

#### 1. FOR...TO...NEXT（计数循环）

当你**知道**要重复多少次时使用：

```
FOR Counter ← StartValue TO EndValue
    statements
NEXT Counter
```

#### 2. WHILE...DO...ENDWHILE（前测循环）

**先检查条件**，条件为 TRUE 才执行循环体：

```
WHILE condition DO
    statements
ENDWHILE
```

特点：循环体**可能一次都不执行**（如果一开始条件就是 FALSE）

#### 3. REPEAT...UNTIL（后测循环）

**先执行循环体**，然后检查条件，条件为 TRUE 时**停止**：

```
REPEAT
    statements
UNTIL condition
```

特点：循环体**至少执行一次**

> **关键区别**:
> - WHILE: 条件为 TRUE 时**继续**循环
> - REPEAT...UNTIL: 条件为 TRUE 时**停止**循环
> - 这是初学者最容易混淆的点！

````python
# 三种循环结构对比

print('三种循环结构 (Iteration) 对比')
print('=' * 60)

# FOR loop
print('\n--- 1. FOR...TO...NEXT ---')
print('伪代码: FOR i <- 1 TO 5 ... NEXT i')
print('执行过程:')
total = 0
for i in range(1, 6):
    total += i
    print(f'  i = {i}, total = {total}')
print(f'结果: 1+2+3+4+5 = {total}')

# WHILE loop
print('\n--- 2. WHILE...DO...ENDWHILE ---')
print('伪代码: WHILE Number > 0 DO ... ENDWHILE')
print('场景: 反转一个数字的各位')
number = 12345
print(f'原始数字: {number}')
print('执行过程:')
temp = number
reversed_digits = []
while temp > 0:
    digit = temp % 10
    reversed_digits.append(str(digit))
    print(f'  temp = {temp}, digit = {digit}')
    temp = temp // 10
print(f'反转结果: {"".join(reversed_digits)}')

# REPEAT...UNTIL loop
print('\n--- 3. REPEAT...UNTIL ---')
print('伪代码: REPEAT ... UNTIL Password = "secret"')
print('场景: 密码验证（模拟）')
passwords_to_try = ['hello', 'wrong', '123', 'secret']
attempt = 0
while True:  # Python doesn't have repeat...until, simulate with while True
    password = passwords_to_try[attempt]
    print(f'  尝试 {attempt+1}: 输入 "{password}"', end='')
    attempt += 1
    if password == 'secret':  # UNTIL condition
        print(' => 正确！循环结束')
        break
    else:
        print(' => 错误，继续...')

print(f'\n总共尝试了 {attempt} 次')
````

````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, axes = plt.subplots(1, 3, figsize=(18, 8))

def draw_box(ax, x, y, text, color, w=2.2, h=0.6):
    rect = mpatches.FancyBboxPatch((x-w/2, y-h/2), w, h,
        boxstyle='round,pad=0.1', facecolor=color, edgecolor='black', linewidth=1.2)
    ax.add_patch(rect)
    ax.text(x, y, text, ha='center', va='center', fontsize=8, fontweight='bold')

def draw_diamond(ax, x, y, text, size=0.6):
    d = plt.Polygon([(x, y+size), (x+size*1.3, y), (x, y-size), (x-size*1.3, y)],
        facecolor='#FFF9C4', edgecolor='black', linewidth=1.2)
    ax.add_patch(d)
    ax.text(x, y, text, ha='center', va='center', fontsize=7)

def arrow(ax, x1, y1, x2, y2):
    ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
        arrowprops=dict(arrowstyle='->', lw=1.2))

for ax_item in axes:
    ax_item.set_xlim(0, 6)
    ax_item.set_ylim(0, 8)
    ax_item.axis('off')

# FOR loop
ax = axes[0]
ax.set_title('FOR...TO...NEXT', fontsize=12, fontweight='bold')
draw_box(ax, 3, 7.2, 'Counter <- Start', '#C8E6C9')
arrow(ax, 3, 6.9, 3, 6.3)
draw_diamond(ax, 3, 5.7, 'Counter\n<= End?')
ax.text(3.9, 5.9, 'Yes', fontsize=8, color='green')
arrow(ax, 3, 5.1, 3, 4.5)
draw_box(ax, 3, 4.2, 'Execute Body', '#BBDEFB')
arrow(ax, 3, 3.9, 3, 3.3)
draw_box(ax, 3, 3.0, 'Counter <- Counter+1', '#C8E6C9')
# Loop back
ax.annotate('', xy=(1, 5.7), xytext=(1, 3.0),
    arrowprops=dict(arrowstyle='->', lw=1.2))
ax.plot([1, 1], [3.0, 3.0], 'k-')
ax.plot([1, 3], [3.0, 3.0], 'k-', alpha=0)
ax.plot([1.0, 1.78], [5.7, 5.7], 'k-')
ax.plot([1.0, 1.0], [3.0, 5.7], 'k-')
# Exit
ax.text(4.1, 5.4, 'No', fontsize=8, color='red')
ax.annotate('', xy=(5, 5.7), xytext=(3.78, 5.7),
    arrowprops=dict(arrowstyle='->', lw=1.2))
draw_box(ax, 5, 5.7, 'Exit', '#FFCDD2', w=1)
ax.text(3, 1.8, 'Known iterations', ha='center', fontsize=9, style='italic', color='#666')

# WHILE loop
ax = axes[1]
ax.set_title('WHILE...DO...ENDWHILE', fontsize=12, fontweight='bold')
draw_diamond(ax, 3, 6.5, 'Condition\nTRUE?')
ax.text(3.9, 6.7, 'Yes', fontsize=8, color='green')
arrow(ax, 3, 5.9, 3, 5.3)
draw_box(ax, 3, 5.0, 'Execute Body', '#BBDEFB')
# Loop back
ax.plot([1.0, 1.78], [6.5, 6.5], 'k-')
ax.plot([1.0, 1.0], [5.0, 6.5], 'k-')
ax.annotate('', xy=(1, 6.5), xytext=(1, 5.0),
    arrowprops=dict(arrowstyle='->', lw=1.2))
# Exit
ax.text(4.1, 6.2, 'No', fontsize=8, color='red')
ax.annotate('', xy=(5, 6.5), xytext=(3.78, 6.5),
    arrowprops=dict(arrowstyle='->', lw=1.2))
draw_box(ax, 5, 6.5, 'Exit', '#FFCDD2', w=1)
ax.text(3, 3.5, 'Check FIRST', ha='center', fontsize=10, fontweight='bold', color='#D32F2F')
ax.text(3, 2.8, 'May execute 0 times', ha='center', fontsize=9, style='italic', color='#666')

# REPEAT...UNTIL loop
ax = axes[2]
ax.set_title('REPEAT...UNTIL', fontsize=12, fontweight='bold')
draw_box(ax, 3, 6.5, 'Execute Body', '#BBDEFB')
arrow(ax, 3, 6.2, 3, 5.5)
draw_diamond(ax, 3, 4.9, 'Condition\nTRUE?')
# Exit
ax.text(3.9, 4.6, 'Yes', fontsize=8, color='green')
ax.annotate('', xy=(5, 4.9), xytext=(3.78, 4.9),
    arrowprops=dict(arrowstyle='->', lw=1.2))
draw_box(ax, 5, 4.9, 'Exit', '#FFCDD2', w=1)
# Loop back
ax.text(1.6, 4.6, 'No', fontsize=8, color='red')
ax.plot([1.78, 1.0], [4.9, 4.9], 'k-')
ax.plot([1.0, 1.0], [4.9, 6.5], 'k-')
ax.annotate('', xy=(1, 6.5), xytext=(1, 5.5),
    arrowprops=dict(arrowstyle='->', lw=1.2))
ax.plot([1, 1.88], [6.5, 6.5], 'k-')
ax.text(3, 3.0, 'Execute FIRST', ha='center', fontsize=10, fontweight='bold', color='#D32F2F')
ax.text(3, 2.3, 'Always executes at least once', ha='center', fontsize=9, style='italic', color='#666')

plt.suptitle('三种循环结构的流程图对比', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

## 4. 在三种表达方式之间转换

### 4.1 转换方法

结构化英语 ↔ 伪代码 ↔ 流程图 ↔ Python 代码

**转换的关键**: 识别出算法中的三种基本结构（顺序、选择、循环），然后用目标格式的语法重新表达。

| 概念 | 结构化英语 | 伪代码 | Python |
|------|-----------|--------|--------|
| 输入 | GET / ASK | INPUT | `input()` |
| 输出 | DISPLAY / SHOW | OUTPUT | `print()` |
| 赋值 | SET...TO | ← | `=` |
| 判断 | IF...THEN | IF...THEN...ENDIF | `if...elif...else` |
| 多选 | DEPENDING ON | CASE OF...ENDCASE | `match` / `if-elif` |
| 计数循环 | REPEAT N TIMES | FOR...TO...NEXT | `for i in range()` |
| 前测循环 | WHILE...DO | WHILE...ENDWHILE | `while:` |
| 后测循环 | DO...UNTIL | REPEAT...UNTIL | `while True: ...break` |

## 5. 流程图符号 (Flowchart Symbols)

### 5.1 标准符号

| 符号 | 形状 | 名称 | 用途 |
|------|------|------|------|
| ⬭ | 椭圆/圆角矩形 | Terminal (终端) | 表示开始 (START) 或结束 (END) |
| ▭ | 矩形 | Process (处理) | 执行操作，如计算、赋值 |
| ◇ | 菱形 | Decision (判断) | IF 条件判断，有 Yes/No 两个出口 |
| ▱ | 平行四边形 | Input/Output (输入/输出) | 接收输入或显示输出 |
| → | 箭头 | Flow Arrow (流向箭头) | 指示执行顺序 |

### 5.2 流程图绘制规则

1. **只能有一个 START 和一个 END**
2. **流向一般从上到下、从左到右**
3. **每个判断框必须有两个出口**（Yes/No 或 True/False）
4. **所有路径最终都必须到达 END**
5. **箭头不能交叉**（如果必须交叉，用“桥接”表示）

````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots(1, 1, figsize=(12, 6))
ax.set_xlim(0, 12)
ax.set_ylim(0, 6)
ax.axis('off')
ax.set_title('流程图标准符号 (Standard Flowchart Symbols)', fontsize=14, fontweight='bold')

# 1. Terminal (oval)
oval = mpatches.FancyBboxPatch((0.5, 4), 2, 1,
    boxstyle='round,pad=0.3', facecolor='#FFCDD2', edgecolor='black', linewidth=1.5)
ax.add_patch(oval)
ax.text(1.5, 4.5, 'START/END', ha='center', va='center', fontsize=10, fontweight='bold')
ax.text(1.5, 3.5, 'Terminal\n(终端)', ha='center', va='center', fontsize=9, color='#555')

# 2. Process (rectangle)
rect = mpatches.FancyBboxPatch((3.5, 4), 2, 1,
    boxstyle='square,pad=0.05', facecolor='#C8E6C9', edgecolor='black', linewidth=1.5)
ax.add_patch(rect)
ax.text(4.5, 4.5, 'x <- x + 1', ha='center', va='center', fontsize=10, fontweight='bold')
ax.text(4.5, 3.5, 'Process\n(处理)', ha='center', va='center', fontsize=9, color='#555')

# 3. Decision (diamond)
diamond = plt.Polygon([(7.5, 5.2), (8.8, 4.5), (7.5, 3.8), (6.2, 4.5)],
    facecolor='#FFF9C4', edgecolor='black', linewidth=1.5)
ax.add_patch(diamond)
ax.text(7.5, 4.5, 'x > 0?', ha='center', va='center', fontsize=10, fontweight='bold')
ax.text(7.5, 3.2, 'Decision\n(判断)', ha='center', va='center', fontsize=9, color='#555')

# 4. I/O (parallelogram)
para = plt.Polygon([(9.8, 4), (11.8, 4), (11.4, 5), (9.4, 5)],
    facecolor='#BBDEFB', edgecolor='black', linewidth=1.5)
ax.add_patch(para)
ax.text(10.6, 4.5, 'INPUT x', ha='center', va='center', fontsize=10, fontweight='bold')
ax.text(10.6, 3.5, 'Input/Output\n(输入/输出)', ha='center', va='center', fontsize=9, color='#555')

# 5. Arrow
ax.annotate('', xy=(3, 1.5), xytext=(1, 1.5),
    arrowprops=dict(arrowstyle='->', lw=2.5, color='black'))
ax.text(2, 1.0, 'Flow Arrow (流向箭头)', ha='center', va='center', fontsize=9, color='#555')

plt.tight_layout()
plt.show()
````

## 6. 逐步细化 (Stepwise Refinement)

### 6.1 什么是逐步细化？

> **逐步细化**是一种 **自顶向下 (Top-Down)** 的设计方法：
> 从问题的高层描述开始，逐步将每一步细化为更具体、更详细的步骤，
> 直到每一步都可以直接转化为代码。

### 6.2 实例：计算平均成绩并评定等级

**第一层（最粗略）**:
```
1. 获取所有学生的成绩
2. 计算平均分
3. 根据平均分评定等级
4. 输出结果
```

**第二层（细化每一步）**:
```
1. 获取所有学生的成绩
   1.1 询问学生数量
   1.2 对每个学生，输入其成绩
   1.3 将成绩累加到总分

2. 计算平均分
   2.1 平均分 = 总分 / 学生数量

3. 根据平均分评定等级
   3.1 如果平均分 >= 90，等级为 A
   3.2 如果平均分 >= 80，等级为 B
   3.3 如果平均分 >= 70，等级为 C
   3.4 如果平均分 >= 60，等级为 D
   3.5 否则，等级为 F

4. 输出结果
   4.1 显示平均分
   4.2 显示等级
```

**第三层（直接可编码的伪代码）**:
```
INPUT NumberOfStudents
Total ← 0
FOR Counter ← 1 TO NumberOfStudents
    INPUT Mark
    Total ← Total + Mark
NEXT Counter
Average ← Total / NumberOfStudents
IF Average >= 90
  THEN Grade ← 'A'
  ELSE IF Average >= 80
    THEN Grade ← 'B'
    ELSE IF Average >= 70
      THEN Grade ← 'C'
      ELSE IF Average >= 60
        THEN Grade ← 'D'
        ELSE Grade ← 'F'
OUTPUT "Average: ", Average
OUTPUT "Grade: ", Grade
```

````python
# 逐步细化的最终实现

print('逐步细化 (Stepwise Refinement) 完整实现')
print('=' * 50)

# Simulated input
marks = [78, 85, 92, 67, 88, 73, 95, 81]
number_of_students = len(marks)

print(f'\n学生人数: {number_of_students}')
print(f'各学生成绩: {marks}')
print()

# Step 1: Get marks and accumulate total
print('--- 步骤 1: 累加总分 ---')
total = 0
for counter in range(number_of_students):
    total = total + marks[counter]
    print(f'  读入 Mark[{counter+1}] = {marks[counter]}, Total = {total}')

# Step 2: Calculate average
print('\n--- 步骤 2: 计算平均分 ---')
average = total / number_of_students
print(f'  Average = {total} / {number_of_students} = {average:.2f}')

# Step 3: Determine grade
print('\n--- 步骤 3: 评定等级 ---')
if average >= 90:
    grade = 'A'
elif average >= 80:
    grade = 'B'
elif average >= 70:
    grade = 'C'
elif average >= 60:
    grade = 'D'
else:
    grade = 'F'
print(f'  {average:.2f} >= 80? => Grade = {grade}')

# Step 4: Output
print('\n--- 步骤 4: 输出结果 ---')
print(f'  Average: {average:.2f}')
print(f'  Grade: {grade}')
````

| `NOT` | 非（取反） | `IF NOT (x = 0)` | 条件为 FALSE 时返回 TRUE |

| F | F | F | F | T |

### 8.2 实例 2：统计字符串中的元音字母

**问题**: 输入一个字符串，统计其中元音字母 (A, E, I, O, U) 的个数。

#### 伪代码:
```
INPUT Text
VowelCount ← 0
FOR Index ← 1 TO LENGTH(Text)
    CurrentChar ← UCASE(MID(Text, Index, 1))
    IF CurrentChar = 'A' OR CurrentChar = 'E' OR CurrentChar = 'I'
       OR CurrentChar = 'O' OR CurrentChar = 'U'
      THEN
        VowelCount ← VowelCount + 1
    ENDIF
NEXT Index
OUTPUT "Number of vowels: ", VowelCount
```

#### 标识符表:

| Identifier | Data Type | Description |
|-----------|-----------|-------------|
| Text | STRING | 输入的字符串 |
| VowelCount | INTEGER | 元音字母计数 |
| Index | INTEGER | 循环计数器 |
| CurrentChar | CHAR | 当前正在检查的字符（大写） |

````python
# 实例 2: 统计元音字母 - 带详细 trace

print('统计字符串中的元音字母')
print('=' * 50)

def count_vowels(text):
    vowels = 'AEIOU'
    vowel_count = 0
    print(f'  输入: "{text}"')
    print(f'  逐字符检查:')
    for index, char in enumerate(text):
        upper_char = char.upper()
        is_vowel = upper_char in vowels
        if is_vowel:
            vowel_count += 1
            print(f'    [{index}] "{char}" -> "{upper_char}" => VOWEL! (count = {vowel_count})')
    print(f'  结果: {vowel_count} 个元音字母')
    return vowel_count

test_strings = ['Hello World', 'AEIOU', 'Python', 'Cambridge']
for s in test_strings:
    count_vowels(s)
    print()
````

### 8.3 实例 3：简单计算器

**问题**: 输入两个数和一个运算符，计算结果。

#### 伪代码:
```
INPUT Num1
INPUT Operator
INPUT Num2
CASE OF Operator
    '+' : Result ← Num1 + Num2
    '-' : Result ← Num1 - Num2
    '*' : Result ← Num1 * Num2
    '/' : IF Num2 <> 0
            THEN
              Result ← Num1 / Num2
            ELSE
              OUTPUT "Error: Division by zero"
          ENDIF
    OTHERWISE : OUTPUT "Invalid operator"
ENDCASE
OUTPUT Num1, Operator, Num2, "=", Result
```

````python
# 实例 3: 简单计算器

print('简单计算器 (Simple Calculator)')
print('=' * 50)

def calculate(num1, operator, num2):
    if operator == '+':
        result = num1 + num2
    elif operator == '-':
        result = num1 - num2
    elif operator == '*':
        result = num1 * num2
    elif operator == '/':
        if num2 != 0:
            result = num1 / num2
        else:
            print(f'  {num1} {operator} {num2} => Error: Division by zero!')
            return None
    else:
        print(f'  Invalid operator: "{operator}"')
        return None
    print(f'  {num1} {operator} {num2} = {result}')
    return result

# Test cases
test_calcs = [
    (10, '+', 5),
    (20, '-', 8),
    (7, '*', 6),
    (15, '/', 4),
    (10, '/', 0),
    (5, '%', 3),
]

for num1, op, num2 in test_calcs:
    calculate(num1, op, num2)
````

## 9. 伪代码执行器 (Pseudocode Executor)

下面是一个简化的伪代码执行器，可以帮助你理解伪代码是如何一步步执行的。

它支持：
- 赋值 (`Variable <- Expression`)
- 输出 (`OUTPUT ...`)
- FOR 循环 (`FOR...TO...NEXT`)
- IF 语句 (`IF...THEN...ENDIF`)

````python
# 简化的伪代码执行器
# 注意：这是一个教学工具，只支持非常基本的伪代码

class SimplePseudocodeExecutor:
    def __init__(self):
        self.variables = {}
        self.output_lines = []
        self.trace = []

    def execute(self, pseudocode_lines):
        self.variables = {}
        self.output_lines = []
        self.trace = []
        self._execute_block(pseudocode_lines, 0, len(pseudocode_lines))
        return self.output_lines

    def _execute_block(self, lines, start, end):
        i = start
        while i < end:
            line = lines[i].strip()
            if not line:
                i += 1
                continue

            # Assignment: Variable <- Expression
            if '<-' in line:
                parts = line.split('<-')
                var_name = parts[0].strip()
                expr = parts[1].strip()
                value = self._eval_expr(expr)
                self.variables[var_name] = value
                self.trace.append(f'  SET {var_name} = {value}')
                i += 1

            # OUTPUT
            elif line.startswith('OUTPUT'):
                expr = line[6:].strip()
                value = self._eval_expr(expr)
                self.output_lines.append(str(value))
                self.trace.append(f'  OUTPUT: {value}')
                i += 1

            # FOR loop
            elif line.startswith('FOR'):
                # Parse: FOR var <- start TO end
                rest = line[3:].strip()
                var_part, end_part = rest.split('TO')
                var_name, start_val_str = var_part.split('<-')
                var_name = var_name.strip()
                start_val = int(self._eval_expr(start_val_str.strip()))
                end_val = int(self._eval_expr(end_part.strip()))

                # Find matching NEXT
                next_i = i + 1
                depth = 1
                while next_i < end:
                    if lines[next_i].strip().startswith('FOR'):
                        depth += 1
                    elif lines[next_i].strip().startswith('NEXT'):
                        depth -= 1
                        if depth == 0:
                            break
                    next_i += 1

                # Execute loop body
                for val in range(start_val, end_val + 1):
                    self.variables[var_name] = val
                    self.trace.append(f'  LOOP: {var_name} = {val}')
                    self._execute_block(lines, i + 1, next_i)

                i = next_i + 1

            else:
                i += 1

    def _eval_expr(self, expr):
        # Replace variable names with values
        expr = expr.strip()
        if expr.startswith('"') and expr.endswith('"'):
            return expr[1:-1]
        # Try to evaluate as math expression
        try:
            eval_expr = expr
            for var, val in self.variables.items():
                eval_expr = eval_expr.replace(var, str(val))
            return eval(eval_expr)
        except Exception:
            return expr


# Demo: Run pseudocode
executor = SimplePseudocodeExecutor()

print('伪代码执行器演示')
print('=' * 50)

# Example: Sum of 1 to 5
pseudocode = [
    'Sum <- 0',
    'FOR i <- 1 TO 5',
    '    Sum <- Sum + i',
    'NEXT i',
    'OUTPUT Sum',
]

print('\n伪代码:')
for line in pseudocode:
    print(f'  {line}')

print('\n执行过程:')
result = executor.execute(pseudocode)
for t in executor.trace:
    print(t)

print(f'\n输出: {result}')
print(f'变量: {executor.variables}')
````

````python
# 练习参考答案

print('练习 1 参考答案 - 标识符表')
print('=' * 60)
table = [
    ('Celsius', 'REAL', '输入的摄氏温度'),
    ('Fahrenheit', 'REAL', '转换后的华氏温度'),
]
print(f'{"Identifier":<15} {"Data Type":<12} {"Description"}')
print('-' * 50)
for name, dt, desc in table:
    print(f'{name:<15} {dt:<12} {desc}')

print()
print('练习 2 参考答案 - 找最大值和最小值')
print('=' * 60)
pseudocode_answer = [
    'INPUT FirstNumber',
    'MaxValue <- FirstNumber',
    'MinValue <- FirstNumber',
    'FOR Counter <- 2 TO 10',
    '    INPUT Number',
    '    IF Number > MaxValue',
    '      THEN',
    '        MaxValue <- Number',
    '    ENDIF',
    '    IF Number < MinValue',
    '      THEN',
    '        MinValue <- Number',
    '    ENDIF',
    'NEXT Counter',
    'OUTPUT "Maximum: ", MaxValue',
    'OUTPUT "Minimum: ", MinValue',
]
for line in pseudocode_answer:
    print(f'  {line}')

print()
print('Python 验证:')
import random
random.seed(42)
numbers = [random.randint(1, 100) for _ in range(10)]
print(f'  数据: {numbers}')
max_val = numbers[0]
min_val = numbers[0]
for num in numbers[1:]:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num
print(f'  Maximum: {max_val}')
print(f'  Minimum: {min_val}')
````

````python
# 练习 4 参考答案 - 伪代码转 Python

print('练习 4 参考答案 - 统计单词数')
print('=' * 50)
print()
print('伪代码 => Python 转换:')
print()

# The pseudocode translated to Python:
sentence = 'The quick brown fox jumps over the lazy dog'
print(f'Input: "{sentence}"')
print()

space_count = 0
for index in range(len(sentence)):
    if sentence[index] == ' ':
        space_count += 1

word_count = space_count + 1
print(f'Space count: {space_count}')
print(f'Word count: {word_count}')
print()

print('练习 5 参考答案 - 猜数字游戏')
print('=' * 50)
print()
print('伪代码:')
guess_pseudocode = [
    'Secret <- RANDOM(1, 100)',
    'Attempts <- 0',
    'REPEAT',
    '    INPUT Guess',
    '    Attempts <- Attempts + 1',
    '    IF Guess > Secret',
    '      THEN',
    '        OUTPUT "Too high!"',
    '    ENDIF',
    '    IF Guess < Secret',
    '      THEN',
    '        OUTPUT "Too low!"',
    '    ENDIF',
    'UNTIL Guess = Secret',
    'OUTPUT "Correct! Attempts: ", Attempts',
]
for line in guess_pseudocode:
    print(f'  {line}')

# Simulate
print()
print('模拟执行（使用二分法猜测）:')
import random
random.seed(123)
secret = random.randint(1, 100)
print(f'Secret number: {secret}')
low, high = 1, 100
attempts = 0
while True:
    guess = (low + high) // 2
    attempts += 1
    if guess > secret:
        print(f'  Attempt {attempts}: Guess={guess} => Too high!')
        high = guess - 1
    elif guess < secret:
        print(f'  Attempt {attempts}: Guess={guess} => Too low!')
        low = guess + 1
    else:
        print(f'  Attempt {attempts}: Guess={guess} => Correct!')
        break
print(f'Total attempts: {attempts}')
````

### 核心知识点回顾

```
算法表示与伪代码
│
├── 标识符表 (Identifier Table)
│   └── 变量名 + 数据类型 + 描述
│
├── 伪代码语法 (Pseudocode Syntax)
│   ├── INPUT / OUTPUT
│   ├── 赋值: ←
│   ├── 顺序 (Sequence)
│   ├── 选择 (Selection): IF...ENDIF / CASE OF...ENDCASE
│   └── 循环 (Iteration): FOR / WHILE / REPEAT...UNTIL
│
├── 流程图 (Flowcharts)
│   ├── 终端 (Terminal): 椭圆
│   ├── 处理 (Process): 矩形
│   ├── 判断 (Decision): 菱形
│   └── 输入/输出 (I/O): 平行四边形
│
├── 逐步细化 (Stepwise Refinement)
│   └── 自顶向下，逐层细化
│
├── 逻辑语句 (Logic Statements)
│   └── AND / OR / NOT
│
└── 三种表达方式的相互转换
    └── 结构化英语 ↔ 流程图 ↔ 伪代码 ↔ Python
```

### 考试建议

1. **伪代码格式**: 严格按照 Cambridge 9618 的格式书写，注意缩进和关键字大写
2. **流程图**: 符号必须正确，箭头方向清晰，每个判断必须标注 Yes/No
3. **标识符表**: 变量命名要有意义，数据类型要准确

5. **转换**: 能够在三种表达方式之间自由转换是必考技能
