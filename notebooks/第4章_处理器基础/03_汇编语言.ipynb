# 第4章 处理器基础 - 4.2 汇编语言 (Assembly Language)

## Cambridge International AS & A Level Computer Science (9618)

---

> **核心问题**：CPU只懂0和1，人类怎么跟它"对话"？汇编语言就是答案。

汇编语言是最接近机器码的编程语言，理解它能帮助你真正理解计算机是如何工作的。

## 1. 机器码 vs 汇编语言

### 机器码 (Machine Code)
- CPU唯一能直接执行的语言
- 全部由 **0和1** 组成
- 例如：`0001 0000 11001000` 表示 "将地址200的值加载到ACC"
- 对人类来说几乎 **不可能阅读和编写**

### 汇编语言 (Assembly Language)
- 使用 **助记符 (Mnemonics)** 代替二进制操作码
- 例如：`LDD 200` 就是上面那条机器码的汇编形式
- 与机器码 **一一对应**（一条汇编指令 = 一条机器码指令）
- 需要 **汇编器 (Assembler)** 翻译成机器码

### 转换过程

```
汇编语言 (Assembly)     汇编器 (Assembler)     机器码 (Machine Code)
   LDD 200         ───────────────→      0001 0000 11001000
   ADD 201         ───────────────→      0010 0000 11001001
   STO 202         ───────────────→      0011 0000 11001010
```

> **类比**：
> - 机器码 = 摩斯电码（滴滴答答，很难读）
> - 汇编语言 = 电报的文字版本（容易理解，但需要翻译成摩斯电码才能发送）
> - 高级语言 = 普通对话（最容易，但离机器最远）

## 2. 指令分类

AS & A Level考试中需要掌握的汇编指令：

### 2.1 数据传输指令 (Data Movement)

| 指令 | 格式 | 含义 | 寻址方式 |
|------|------|------|----------|
| **LDM** | `LDM #n` | ACC ← n | 立即寻址 |
| **LDD** | `LDD addr` | ACC ← memory[addr] | 直接寻址 |
| **LDI** | `LDI addr` | ACC ← memory[memory[addr]] | 间接寻址 |
| **LDX** | `LDX addr` | ACC ← memory[addr + IX] | 变址寻址 |
| **LDR** | `LDR #n` | IX ← n | 立即寻址(IX) |
| **STO** | `STO addr` | memory[addr] ← ACC | 直接寻址 |
| **MOV** | `MOV reg` | reg ← ACC | 寄存器 |

### 2.2 算术指令 (Arithmetic)

| 指令 | 格式 | 含义 |
|------|------|------|
| **ADD** | `ADD addr` | ACC ← ACC + memory[addr] |
| **SUB** | `SUB addr` | ACC ← ACC - memory[addr] |
| **INC** | `INC reg` | reg ← reg + 1 |
| **DEC** | `DEC reg` | reg ← reg - 1 |

### 2.3 比较与跳转 (Compare & Branch)

| 指令 | 格式 | 含义 |
|------|------|------|
| **CMP** | `CMP addr` | 比较 ACC 与 memory[addr] |
| **CMP** | `CMP #n` | 比较 ACC 与 n |
| **JMP** | `JMP label` | 无条件跳转到label |
| **JPE** | `JPE label` | 如果比较结果相等则跳转 |
| **JNE** | `JNE label` | 如果比较结果不相等则跳转 |
| **JGT** | `JGT label` | 如果ACC > 比较值则跳转 |
| **JLT** | `JLT label` | 如果ACC < 比较值则跳转 |

### 2.4 输入/输出与结束

| 指令 | 格式 | 含义 |
|------|------|------|
| **IN** | `IN` | ACC ← 用户输入 |
| **OUT** | `OUT` | 输出ACC的值 |
| **END** | `END` | 程序结束 |

## 3. 寻址方式 (Addressing Modes) — 关键考点！

寻址方式决定了如何找到操作数（数据）。这是考试中的重点和难点！

### 3.1 立即寻址 (Immediate Addressing)

**操作数本身就是要使用的值**

```
LDM #5     → ACC = 5（#号表示立即寻址）
```

类比：有人告诉你"拿5个苹果"——5就是你要的数字，不需要去别处查找。

### 3.2 直接寻址 (Direct Addressing)

**操作数是数据存储的内存地址**

```
LDD 200    → ACC = memory[200]（去地址200拿数据）
```

类比：有人告诉你"去200号储物柜拿东西"——200是地址，不是你要的值。

### 3.3 间接寻址 (Indirect Addressing)

**操作数是一个地址，该地址中存储的是另一个地址（指向数据的指针）**

```
LDI 300    → temp = memory[300] → ACC = memory[temp]
```

类比：有人告诉你"去300号柜子，里面有一张纸条告诉你真正去哪里取东西"。
这就是 **指针 (Pointer)** 的概念！

### 3.4 变址寻址 (Indexed Addressing)

**有效地址 = 操作数 + IX（变址寄存器）的值**

```
LDR #2     → IX = 2
LDX 500    → ACC = memory[500 + 2] = memory[502]
```

类比：基地址是500（数组起始位置），IX是偏移量2（第3个元素），所以访问502。
**这就是数组访问的底层实现！**

### 3.5 相对寻址 (Relative Addressing)

**有效地址 = PC + 偏移量**

```
JMP +3     → PC = PC + 3（向前跳3个地址）
```

优点：代码可以在内存中任意位置运行（位置无关代码 position-independent code）。

## 5. 两遍汇编器 (Two-Pass Assembler)

汇编器将汇编语言翻译成机器码。为什么需要"两遍"？

### 问题：前向引用 (Forward Reference)

```
      JMP LOOP_END    ; 这里引用了LOOP_END，但还没定义！
      ADD 200
LOOP_END:
      OUT
      END
```

第一遍看到 `JMP LOOP_END` 时，还不知道 `LOOP_END` 在哪里！

### 第一遍 (Pass 1)：构建符号表

1. 逐行读取源代码
2. 为每条指令分配内存地址
3. 遇到标签 (Label) 时，记录标签名和对应地址到 **符号表 (Symbol Table)**

### 第二遍 (Pass 2)：生成机器码

1. 再次逐行读取源代码
2. 将每条指令的助记符转换为操作码
3. 使用符号表替换标签为实际地址
4. 生成最终的机器码

## 6. 两遍汇编器模拟器

````python
class TwoPassAssembler:
    OPCODES = {
        'LDM': '0001', 'LDD': '0010', 'LDI': '0011', 'LDX': '0100',
        'LDR': '0101', 'STO': '0110', 'ADD': '0111', 'SUB': '1000',
        'INC': '1001', 'DEC': '1010', 'CMP': '1011', 'JMP': '1100',
        'JPE': '1101', 'JNE': '1110', 'OUT': '1111', 'END': '0000',
    }
    
    def __init__(self):
        self.symbol_table = {}
        self.machine_code = []
    
    def pass1(self, source_lines):
        print('=== PASS 1: \u6784\u5efa\u7b26\u53f7\u8868 ===')
        addr = 0
        cleaned = []
        for line in source_lines:
            line = line.strip()
            if not line or line.startswith(';'):
                continue
            # Check for label
            if ':' in line:
                parts = line.split(':')
                label = parts[0].strip()
                self.symbol_table[label] = addr
                print(f'  \u53d1\u73b0\u6807\u7b7e: {label} -> \u5730\u5740 {addr}')
                rest = parts[1].strip()
                if rest:
                    cleaned.append((addr, rest))
                    addr += 1
            else:
                cleaned.append((addr, line))
                addr += 1
        print(f'\n  \u7b26\u53f7\u8868: {self.symbol_table}')
        return cleaned
    
    def pass2(self, cleaned_lines):
        print('\n=== PASS 2: \u751f\u6210\u673a\u5668\u7801 ===')
        for addr, line in cleaned_lines:
            parts = line.split()
            mnemonic = parts[0]
            operand_str = parts[1] if len(parts) > 1 else ''
            
            opcode_bin = self.OPCODES.get(mnemonic, '????')
            
            # Resolve operand
            if operand_str in self.symbol_table:
                operand_val = self.symbol_table[operand_str]
                note = f'(\u6807\u7b7e{operand_str}->\u5730\u5740{operand_val})'
            elif operand_str.startswith('#'):
                operand_val = int(operand_str[1:])
                note = '(\u7acb\u5373\u503c)'
            elif operand_str.isdigit():
                operand_val = int(operand_str)
                note = '(\u5730\u5740)'
            else:
                operand_val = 0
                note = ''
            
            operand_bin = format(operand_val, '08b')
            machine = f'{opcode_bin} {operand_bin}'
            self.machine_code.append((addr, machine))
            print(f'  [{addr:03d}] {line:15s} -> {machine}  {note}')
    
    def assemble(self, source):
        lines = source.strip().split('\n')
        cleaned = self.pass1(lines)
        self.pass2(cleaned)
        return self.machine_code

# Example program: sum numbers until result >= 100
program = """
; \u7a0b\u5e8f\uff1a\u8ba1\u7b97 20 + 30 + 50\uff0c\u8f93\u51fa\u7ed3\u679c
START:  LDM #0
        ADD NUM1
        ADD NUM2
        ADD NUM3
        STO RESULT
        OUT
        END
NUM1:   LDM #20
NUM2:   LDM #30
NUM3:   LDM #50
RESULT: LDM #0
"""

asm = TwoPassAssembler()
asm.assemble(program)
````

## 7. 汇编程序追踪 (Trace Table)

追踪表是考试中常见的题型。让我们编写一个自动追踪工具：

````python
class AssemblyTracer:
    def __init__(self):
        self.ACC = 0
        self.IX = 0
        self.PC = 0
        self.memory = {}
        self.instructions = []
        self.labels = {}
        self.cmp_flag = None  # 'EQ', 'GT', 'LT'
        self.trace = []
        self.output = []
    
    def load(self, program_text):
        lines = program_text.strip().split('\n')
        addr = 0
        for line in lines:
            line = line.strip()
            if not line or line.startswith(';'):
                continue
            if ':' in line:
                parts = line.split(':')
                self.labels[parts[0].strip()] = addr
                rest = parts[1].strip()
                if rest:
                    self.instructions.append(rest)
                    addr += 1
            else:
                self.instructions.append(line)
                addr += 1
    
    def get_val(self, operand_str):
        if operand_str.startswith('#'):
            return int(operand_str[1:])
        addr = int(operand_str) if operand_str.isdigit() else self.labels.get(operand_str, 0)
        return self.memory.get(addr, 0)
    
    def get_addr(self, operand_str):
        if operand_str.isdigit():
            return int(operand_str)
        return self.labels.get(operand_str, 0)
    
    def run(self, data=None):
        if data:
            for addr, val in data.items():
                self.memory[addr] = val
        
        self.PC = 0
        steps = 0
        print(f'{"Step":>4} | {"Instruction":>15} | {"ACC":>6} | {"IX":>4} | {"Output":>8} | Note')
        print('-' * 70)
        
        while self.PC < len(self.instructions) and steps < 50:
            inst = self.instructions[self.PC]
            parts = inst.split()
            op = parts[0]
            operand = parts[1] if len(parts) > 1 else ''
            old_pc = self.PC
            self.PC += 1
            note = ''
            out_str = ''
            
            if op == 'LDM':
                self.ACC = int(operand.replace('#', ''))
                note = f'ACC <- {self.ACC}'
            elif op == 'LDD':
                addr = self.get_addr(operand)
                self.ACC = self.memory.get(addr, 0)
                note = f'ACC <- mem[{addr}]={self.ACC}'
            elif op == 'LDI':
                addr = self.get_addr(operand)
                ptr = self.memory.get(addr, 0)
                self.ACC = self.memory.get(ptr, 0)
                note = f'ACC <- mem[mem[{addr}]]'
            elif op == 'LDX':
                addr = self.get_addr(operand)
                eff = addr + self.IX
                self.ACC = self.memory.get(eff, 0)
                note = f'ACC <- mem[{addr}+{self.IX}]'
            elif op == 'LDR':
                self.IX = int(operand.replace('#', ''))
                note = f'IX <- {self.IX}'
            elif op == 'STO':
                addr = self.get_addr(operand)
                self.memory[addr] = self.ACC
                note = f'mem[{addr}] <- {self.ACC}'
            elif op == 'ADD':
                if operand.startswith('#'):
                    val = int(operand[1:])
                else:
                    addr = self.get_addr(operand)
                    val = self.memory.get(addr, 0)
                self.ACC += val
                note = f'ACC <- ACC+{val}'
            elif op == 'SUB':
                if operand.startswith('#'):
                    val = int(operand[1:])
                else:
                    addr = self.get_addr(operand)
                    val = self.memory.get(addr, 0)
                self.ACC -= val
                note = f'ACC <- ACC-{val}'
            elif op == 'INC':
                if operand == 'ACC':
                    self.ACC += 1
                elif operand == 'IX':
                    self.IX += 1
                note = f'{operand} += 1'
            elif op == 'DEC':
                if operand == 'ACC':
                    self.ACC -= 1
                elif operand == 'IX':
                    self.IX -= 1
                note = f'{operand} -= 1'
            elif op == 'CMP':
                if operand.startswith('#'):
                    cmp_val = int(operand[1:])
                else:
                    addr = self.get_addr(operand)
                    cmp_val = self.memory.get(addr, 0)
                if self.ACC == cmp_val:
                    self.cmp_flag = 'EQ'
                elif self.ACC > cmp_val:
                    self.cmp_flag = 'GT'
                else:
                    self.cmp_flag = 'LT'
                note = f'CMP {self.ACC} vs {cmp_val} -> {self.cmp_flag}'
            elif op == 'JMP':
                self.PC = self.labels.get(operand, 0)
                note = f'Jump to {operand}(={self.PC})'
            elif op == 'JPE':
                if self.cmp_flag == 'EQ':
                    self.PC = self.labels.get(operand, 0)
                    note = f'EQ: Jump to {operand}'
                else:
                    note = 'Not EQ: no jump'
            elif op == 'JNE':
                if self.cmp_flag != 'EQ':
                    self.PC = self.labels.get(operand, 0)
                    note = f'NE: Jump to {operand}'
                else:
                    note = 'EQ: no jump'
            elif op == 'JGT':
                if self.cmp_flag == 'GT':
                    self.PC = self.labels.get(operand, 0)
                    note = f'GT: Jump to {operand}'
                else:
                    note = 'Not GT: no jump'
            elif op == 'JLT':
                if self.cmp_flag == 'LT':
                    self.PC = self.labels.get(operand, 0)
                    note = f'LT: Jump to {operand}'
                else:
                    note = 'Not LT: no jump'
            elif op == 'OUT':
                out_str = str(self.ACC)
                self.output.append(self.ACC)
                note = f'Output {self.ACC}'
            elif op == 'END':
                print(f'{steps+1:>4} | {inst:>15} | {self.ACC:>6} | {self.IX:>4} | {"":>8} | END')
                break
            
            steps += 1
            print(f'{steps:>4} | {inst:>15} | {self.ACC:>6} | {self.IX:>4} | {out_str:>8} | {note}')
        
        print(f'\nOutput: {self.output}')

# Demo: count from 1 to 5
tracer = AssemblyTracer()
tracer.load("""
        LDM #0
LOOP:   ADD #1
        OUT
        CMP #5
        JNE LOOP
        END
""")
print('\u7a0b\u5e8f: \u4ece1\u6570\u52305')
print('=' * 70)
tracer.run()
````

## 8. 实例：用变址寻址访问数组

变址寻址最典型的应用就是数组访问。下面的程序演示如何遍历一个数组并求和：

````python
tracer2 = AssemblyTracer()
tracer2.load("""
        LDM #0
        STO 60
        LDR #0
LOOP:   LDX 50
        ADD 60
        STO 60
        INC IX
        LDD 60
        CMP #0
        LDM #0
        ADD 60
        CMP #0
        LDD 60
        OUT
        END
""")
# Array at addresses 50-54, sum stored at 60
print('\u7a0b\u5e8f: \u8bbf\u95ee\u6570\u7ec4\u5e76\u7d2f\u52a0')
print('\u6570\u7ec4\u5730\u574050-54: [10, 20, 30, 40, 50]')
print('=' * 70)
tracer2.run(data={50: 10, 51: 20, 52: 30, 53: 40, 54: 50, 60: 0})
````

## 11. 汇编语言与高级语言的对比

### 一个简单的例子：两数相加

**Python:**
```python
a = 10
b = 20
c = a + b
print(c)
```

**Assembly:**
```
LDD 200    ; 加载a的值
ADD 201    ; 加上b的值
STO 202    ; 存储到c
OUT        ; 输出
END
```

| 特征 | 汇编语言 | 高级语言 |
|------|---------|--------|
| 可读性 | 低 | 高 |
| 执行效率 | 极高（接近机器码） | 较低（需要编译/解释） |
| 可移植性 | 低（特定CPU架构） | 高（跨平台） |
| 开发速度 | 慢 | 快 |
| 硬件控制 | 直接控制 | 间接控制 |
| 适用场景 | 驱动程序、嵌入式系统、性能关键代码 | 应用软件、Web开发等 |

## 12. 拓展：实际应用中的汇编语言

你可能觉得汇编语言很"古老"，但它在现代计算中仍然至关重要：

1. **操作系统内核**：Linux内核的启动代码用汇编编写
2. **嵌入式系统**：微控制器（Arduino底层）经常使用汇编
3. **游戏引擎**：关键性能代码可能用汇编优化
4. **安全研究**：逆向工程和漏洞分析需要阅读汇编
5. **编译器开发**：编译器的最终输出就是汇编/机器码

> **知识拓展**：x86汇编（Intel/AMD）和ARM汇编是当今最常见的两种汇编语言。
