## 1. 链表 (Linked List)

### 1.1 为什么需要链表？

数组（Python的list）有一个限制：**插入和删除元素需要移动大量元素**。

| 操作 | 数组 | 链表 |
|------|------|------|
| 按索引访问 | O(1) | O(n) |
| 在头部插入 | O(n) - 需要移动所有元素 | O(1) - 只需改指针 |
| 在中间插入 | O(n) | O(1)（找到位置后） |
| 删除元素 | O(n) | O(1)（找到位置后） |

链表的核心思想：**每个元素不仅存储数据，还存储指向下一个元素的指针(pointer)**。

### 1.2 节点 (Node) 的概念

```
节点 = 数据 (Data) + 指针 (Pointer/Next)

[Data|Next] -> [Data|Next] -> [Data|Next] -> None
  头节点        中间节点        尾节点
  (Head)                       (Tail)
```

````python
# ============================================
# 链表的完整实现
# ============================================

class Node:
    '''链表节点'''
    def __init__(self, data):
        self.data = data    # 存储数据
        self.next = None    # 指向下一个节点的指针
    
    def __repr__(self):
        return f'Node({self.data})'

class LinkedList:
    '''单向链表 (Singly Linked List)'''
    
    def __init__(self):
        self.head = None    # 头指针
        self.size = 0       # 链表长度
    
    def is_empty(self):
        '''检查链表是否为空'''
        return self.head is None
    
    def prepend(self, data):
        '''在头部插入 - O(1)'''
        new_node = Node(data)
        new_node.next = self.head  # 新节点指向原来的头
        self.head = new_node       # 更新头指针
        self.size += 1
    
    def append(self, data):
        '''在尾部插入 - O(n)'''
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:  # 遍历到尾部
                current = current.next
            current.next = new_node
        self.size += 1
    
    def insert_at(self, index, data):
        '''在指定位置插入 - O(n)'''
        if index == 0:
            self.prepend(data)
            return
        new_node = Node(data)
        current = self.head
        for i in range(index - 1):
            if current is None:
                raise IndexError('Index out of range')
            current = current.next
        new_node.next = current.next
        current.next = new_node
        self.size += 1
    
    def delete(self, data):
        '''删除第一个匹配的元素 - O(n)'''
        if self.head is None:
            return False
        if self.head.data == data:
            self.head = self.head.next
            self.size -= 1
            return True
        current = self.head
        while current.next is not None:
            if current.next.data == data:
                current.next = current.next.next  # 跳过被删节点
                self.size -= 1
                return True
            current = current.next
        return False
    
    def search(self, data):
        '''查找元素 - O(n)'''
        current = self.head
        index = 0
        while current is not None:
            if current.data == data:
                return index
            current = current.next
            index += 1
        return -1
    
    def traverse(self):
        '''遍历并返回所有元素'''
        elements = []
        current = self.head
        while current is not None:
            elements.append(current.data)
            current = current.next
        return elements
    
    def __str__(self):
        elements = self.traverse()
        return ' -> '.join(str(e) for e in elements) + ' -> None'

# 演示
ll = LinkedList()
print('=== \u94fe\u8868\u64cd\u4f5c\u6f14\u793a ===')
print(f'\u7a7a\u94fe\u8868: {ll}')

# 插入操作
for val in [10, 20, 30, 40, 50]:
    ll.append(val)
print(f'\u8ffd\u52a0 10,20,30,40,50: {ll}')

ll.prepend(5)
print(f'\u5934\u90e8\u63d2\u5165 5: {ll}')

ll.insert_at(3, 25)
print(f'\u4f4d\u7f6e3\u63d2\u5165 25: {ll}')

# 查找
idx = ll.search(30)
print(f'\n\u67e5\u627e 30: \u627e\u5230\u5728\u7d22\u5f15 {idx}')

# 删除
ll.delete(25)
print(f'\u5220\u9664 25: {ll}')

ll.delete(5)
print(f'\u5220\u9664 5: {ll}')

````

## 2. 二叉树 (Binary Tree) 与二叉搜索树 (BST)

### 2.1 什么是二叉树？

**二叉树 (Binary Tree)** 是一种树形数据结构，每个节点最多有**两个子节点**：
- **左子节点 (Left Child)**
- **右子节点 (Right Child)**

### 2.2 什么是二叉搜索树 (BST)？

**二叉搜索树** 是一种特殊的二叉树，满足以下规则：
- 左子树中所有节点的值 **小于** 根节点
- 右子树中所有节点的值 **大于** 根节点
- 左右子树也分别是BST

### 为什么BST这么有用？

因为它可以实现 **O(log n)** 的查找、插入和删除操作（平衡情况下）！

### 2.3 三种遍历方式

| 遍历方式 | 顺序 | 英文名 | 助记 |
|----------|------|--------|------|
| 前序遍历 | 根→左→右 | Pre-order | **根**先行 |
| 中序遍历 | 左→根→右 | In-order | 根在**中**间 |
| 后序遍历 | 左→右→根 | Post-order | 根最**后** |

> **重要**: BST的中序遍历结果是**有序的**！这是BST最重要的性质之一。

````python
# ============================================
# 二叉搜索树的完整实现
# ============================================

class TreeNode:
    '''\u4e8c\u53c9\u6811\u8282\u70b9'''
    def __init__(self, value):
        self.value = value
        self.left = None    # \u5de6\u5b50\u8282\u70b9
        self.right = None   # \u53f3\u5b50\u8282\u70b9

class BinarySearchTree:
    '''\u4e8c\u53c9\u641c\u7d22\u6811 (BST)'''
    
    def __init__(self):
        self.root = None
    
    def insert(self, value):
        '''\u63d2\u5165\u8282\u70b9 - O(log n) \u5e73\u5747'''
        if self.root is None:
            self.root = TreeNode(value)
        else:
            self._insert_recursive(self.root, value)
    
    def _insert_recursive(self, node, value):
        if value < node.value:
            if node.left is None:
                node.left = TreeNode(value)
            else:
                self._insert_recursive(node.left, value)
        elif value > node.value:
            if node.right is None:
                node.right = TreeNode(value)
            else:
                self._insert_recursive(node.right, value)
        # \u5982\u679c\u76f8\u7b49\uff0c\u4e0d\u63d2\u5165\uff08BST\u4e0d\u5141\u8bb8\u91cd\u590d\u503c\uff09
    
    def search(self, value):
        '''\u67e5\u627e\u8282\u70b9 - O(log n) \u5e73\u5747'''
        return self._search_recursive(self.root, value)
    
    def _search_recursive(self, node, value):
        if node is None:
            return False
        if value == node.value:
            return True
        elif value < node.value:
            return self._search_recursive(node.left, value)  # \u5f80\u5de6\u627e
        else:
            return self._search_recursive(node.right, value)  # \u5f80\u53f3\u627e
    
    def inorder(self, node=None, result=None):
        '''\u4e2d\u5e8f\u904d\u5386: \u5de6 -> \u6839 -> \u53f3 (\u7ed3\u679c\u6709\u5e8f!)'''
        if result is None:
            result = []
            node = self.root
        if node is not None:
            self.inorder(node.left, result)
            result.append(node.value)
            self.inorder(node.right, result)
        return result
    
    def preorder(self, node=None, result=None):
        '''\u524d\u5e8f\u904d\u5386: \u6839 -> \u5de6 -> \u53f3'''
        if result is None:
            result = []
            node = self.root
        if node is not None:
            result.append(node.value)
            self.preorder(node.left, result)
            self.preorder(node.right, result)
        return result
    
    def postorder(self, node=None, result=None):
        '''\u540e\u5e8f\u904d\u5386: \u5de6 -> \u53f3 -> \u6839'''
        if result is None:
            result = []
            node = self.root
        if node is not None:
            self.postorder(node.left, result)
            self.postorder(node.right, result)
            result.append(node.value)
        return result
    
    def delete(self, value):
        '''\u5220\u9664\u8282\u70b9'''
        self.root = self._delete_recursive(self.root, value)
    
    def _delete_recursive(self, node, value):
        if node is None:
            return None
        if value < node.value:
            node.left = self._delete_recursive(node.left, value)
        elif value > node.value:
            node.right = self._delete_recursive(node.right, value)
        else:
            # \u627e\u5230\u8981\u5220\u9664\u7684\u8282\u70b9
            if node.left is None:
                return node.right
            elif node.right is None:
                return node.left
            else:
                # \u6709\u4e24\u4e2a\u5b50\u8282\u70b9\uff1a\u627e\u53f3\u5b50\u6811\u7684\u6700\u5c0f\u503c\u66ff\u4ee3
                min_node = node.right
                while min_node.left:
                    min_node = min_node.left
                node.value = min_node.value
                node.right = self._delete_recursive(node.right, min_node.value)
        return node

# \u6f14\u793a
bst = BinarySearchTree()
values = [50, 30, 70, 20, 40, 60, 80]
print('\u63d2\u5165\u987a\u5e8f:', values)
for v in values:
    bst.insert(v)

print(f'\n\u524d\u5e8f\u904d\u5386 (Pre-order):  {bst.preorder()}')
print(f'\u4e2d\u5e8f\u904d\u5386 (In-order):   {bst.inorder()}  <- \u6ce8\u610f\uff1a\u7ed3\u679c\u662f\u6709\u5e8f\u7684\uff01')
print(f'\u540e\u5e8f\u904d\u5386 (Post-order): {bst.postorder()}')

print(f'\n\u67e5\u627e 40: {bst.search(40)}')
print(f'\u67e5\u627e 55: {bst.search(55)}')

bst.delete(30)
````

## 3. 栈和队列的高级实现

### 3.1 循环队列 (Circular Queue)

**为什么需要循环队列？**

普通队列在数组实现中有一个问题：**出队后前面的空间就浪费了**。

循环队列用 **取模运算 (%)** 让数组"首尾相连"，解决空间浪费问题。

````python
# ============================================
# 循环队列实现
# ============================================

class CircularQueue:
    '''\u5faa\u73af\u961f\u5217 (Circular Queue)'''
    
    def __init__(self, capacity):
        self.capacity = capacity
        self.queue = [None] * capacity
        self.front = 0    # \u961f\u5934\u6307\u9488
        self.rear = -1    # \u961f\u5c3e\u6307\u9488
        self.size = 0     # \u5f53\u524d\u5143\u7d20\u6570
    
    def is_empty(self):
        return self.size == 0
    
    def is_full(self):
        return self.size == self.capacity
    
    def enqueue(self, item):
        '''\u5165\u961f - O(1)'''
        if self.is_full():
            raise Exception('Queue is full!')
        self.rear = (self.rear + 1) % self.capacity  # \u5173\u952e\uff1a\u53d6\u6a21\u5b9e\u73b0\u5faa\u73af
        self.queue[self.rear] = item
        self.size += 1
    
    def dequeue(self):
        '''\u51fa\u961f - O(1)'''
        if self.is_empty():
            raise Exception('Queue is empty!')
        item = self.queue[self.front]
        self.queue[self.front] = None  # \u6e05\u9664
        self.front = (self.front + 1) % self.capacity  # \u53d6\u6a21\u5faa\u73af
        self.size -= 1
        return item
    
    def peek(self):
        '''\u67e5\u770b\u961f\u5934 - O(1)'''
        if self.is_empty():
            raise Exception('Queue is empty!')
        return self.queue[self.front]
    
    def display(self):
        '''\u663e\u793a\u961f\u5217\u72b6\u6001'''
        status = []
        for i in range(self.capacity):
            if self.queue[i] is not None:
                marker = ''
                if i == self.front and i == self.rear:
                    marker = ' <-F,R'
                elif i == self.front:
                    marker = ' <-F'
                elif i == self.rear:
                    marker = ' <-R'
                status.append(f'[{self.queue[i]}{marker}]')
            else:
                status.append('[  ]')
        return ' '.join(status)

# \u6f14\u793a
cq = CircularQueue(5)
print('=== \u5faa\u73af\u961f\u5217\u6f14\u793a ===')
print(f'\u521d\u59cb: {cq.display()}')

for item in ['A', 'B', 'C', 'D']:
    cq.enqueue(item)
    print(f'\u5165\u961f {item}: {cq.display()}')

print(f'\n\u51fa\u961f: {cq.dequeue()}')
print(f'\u51fa\u961f: {cq.dequeue()}')
print(f'\u72b6\u6001: {cq.display()}')

cq.enqueue('E')
cq.enqueue('F')
print(f'\n\u5165\u961f E,F: {cq.display()}')
````

### 3.2 优先队列 (Priority Queue)

**普通队列**: 先进先出 (FIFO)

**优先队列**: 优先级最高的先出

应用场景：操作系统进程调度、急诊室病人排队、Dijkstra最短路径算法

````python
# ============================================
# 优先队列实现（简单版）
# ============================================

class PriorityQueue:
    '''\u4f18\u5148\u961f\u5217 - \u4f18\u5148\u7ea7\u6570\u5b57\u8d8a\u5c0f\u4f18\u5148\u7ea7\u8d8a\u9ad8'''
    
    def __init__(self):
        self.items = []  # \u5b58\u50a8 (priority, data) \u5143\u7ec4
    
    def is_empty(self):
        return len(self.items) == 0
    
    def enqueue(self, item, priority):
        '''\u5165\u961f - \u6309\u4f18\u5148\u7ea7\u63d2\u5165\u5230\u6b63\u786e\u4f4d\u7f6e'''
        entry = (priority, item)
        # \u627e\u5230\u5408\u9002\u7684\u4f4d\u7f6e\u63d2\u5165
        inserted = False
        for i in range(len(self.items)):
            if priority < self.items[i][0]:
                self.items.insert(i, entry)
                inserted = True
                break
        if not inserted:
            self.items.append(entry)
    
    def dequeue(self):
        '''\u51fa\u961f - \u53d6\u51fa\u4f18\u5148\u7ea7\u6700\u9ad8\u7684'''
        if self.is_empty():
            raise Exception('Queue is empty!')
        return self.items.pop(0)
    
    def display(self):
        return [(item, f'\u4f18\u5148\u7ea7:{pri}') for pri, item in self.items]

# \u6f14\u793a\uff1a\u533b\u9662\u6025\u8bca\u5ba4
print('=== \u6025\u8bca\u5ba4\u4f18\u5148\u961f\u5217\u6f14\u793a ===')
pq = PriorityQueue()
pq.enqueue('\u611f\u5192\u60a3\u8005', 5)       # \u4f4e\u4f18\u5148\u7ea7
pq.enqueue('\u9aa8\u6298\u60a3\u8005', 3)       # \u4e2d\u4f18\u5148\u7ea7
pq.enqueue('\u5fc3\u810f\u75c5\u53d1\u4f5c', 1)   # \u6700\u9ad8\u4f18\u5148\u7ea7
pq.enqueue('\u8f7b\u5fae\u64e6\u4f24', 7)       # \u6700\u4f4e\u4f18\u5148\u7ea7
pq.enqueue('\u4e25\u91cd\u8fc7\u654f', 2)       # \u9ad8\u4f18\u5148\u7ea7

print(f'\u961f\u5217\u72b6\u6001: {pq.display()}')
print()
while not pq.is_empty():
    pri, patient = pq.dequeue()
````

## 4. 字典/哈希表 (Hash Table)

### 4.1 为什么哈希表这么快？

哈希表通过 **哈希函数 (Hash Function)** 将键 (key) 映射到数组索引，实现 **O(1)** 平均查找时间。

**核心思想**：`index = hash(key) % table_size`

### 4.2 冲突处理 (Collision Handling)

当两个不同的键映射到同一个索引时，称为 **冲突 (Collision)**。

常见解决方法：
1. **链地址法 (Chaining)**：每个桶存储一个链表
2. **开放地址法 (Open Addressing)**：寻找下一个空位

````python
# ============================================
# 哈希表实现（链地址法）
# ============================================

class HashTable:
    '''\u54c8\u5e0c\u8868 - \u94fe\u5730\u5740\u6cd5\u89e3\u51b3\u51b2\u7a81'''
    
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(size)]  # \u6bcf\u4e2a\u6876\u662f\u4e00\u4e2a\u5217\u8868
        self.count = 0
    
    def _hash(self, key):
        '''\u54c8\u5e0c\u51fd\u6570\uff1a\u5c06\u952e\u6620\u5c04\u5230\u7d22\u5f15'''
        if isinstance(key, int):
            return key % self.size
        # \u5b57\u7b26\u4e32\u54c8\u5e0c\uff1a\u5c06\u6bcf\u4e2a\u5b57\u7b26\u7684ASCII\u503c\u52a0\u8d77\u6765
        hash_val = 0
        for char in str(key):
            hash_val += ord(char)
        return hash_val % self.size
    
    def put(self, key, value):
        '''\u63d2\u5165/\u66f4\u65b0\u952e\u503c\u5bf9 - O(1) \u5e73\u5747'''
        index = self._hash(key)
        # \u68c0\u67e5\u952e\u662f\u5426\u5df2\u5b58\u5728
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index][i] = (key, value)  # \u66f4\u65b0
                return
        self.table[index].append((key, value))  # \u63d2\u5165
        self.count += 1
    
    def get(self, key):
        '''\u83b7\u53d6\u503c - O(1) \u5e73\u5747'''
        index = self._hash(key)
        for k, v in self.table[index]:
            if k == key:
                return v
        raise KeyError(f'Key {key} not found')
    
    def delete(self, key):
        '''\u5220\u9664\u952e\u503c\u5bf9 - O(1) \u5e73\u5747'''
        index = self._hash(key)
        for i, (k, v) in enumerate(self.table[index]):
            if k == key:
                self.table[index].pop(i)
                self.count -= 1
                return True
        return False
    
    def display(self):
        '''\u663e\u793a\u54c8\u5e0c\u8868\u7ed3\u6784'''
        for i, bucket in enumerate(self.table):
            if bucket:
                items = ', '.join(f'{k}:{v}' for k, v in bucket)
                print(f'  \u6876[{i}]: {items}')
            else:
                print(f'  \u6876[{i}]: (\u7a7a)')

# \u6f14\u793a
ht = HashTable(7)
print('=== \u54c8\u5e0c\u8868\u6f14\u793a ===')

students = [('Alice', 95), ('Bob', 87), ('Charlie', 92),
            ('Diana', 78), ('Eve', 88), ('Frank', 91)]

for name, score in students:
    index = ht._hash(name)
    print(f'  \u63d2\u5165 {name}({score}): hash("{name}") = {index}')
    ht.put(name, score)

print(f'\n\u54c8\u5e0c\u8868\u72b6\u6001:')
ht.display()

print(f'\n\u67e5\u627e Alice: {ht.get("Alice")}')
````

## 5. 图数据结构 (Graph)

### 5.1 什么是图？

图由 **顶点 (Vertices/Nodes)** 和 **边 (Edges)** 组成：
- **无向图 (Undirected Graph)**：边没有方向
- **有向图 (Directed Graph)**：边有方向
- **带权图 (Weighted Graph)**：边有权重/距离

### 5.2 图的表示方法

| 方法 | 优点 | 缺点 |
|------|------|------|
| 邻接矩阵 (Adjacency Matrix) | 查边O(1) | 空间O(V²)，稀疏图浪费 |
| 邻接表 (Adjacency List) | 空间O(V+E)，遍历快 | 查边O(V) |

````python
# ============================================
# 图的两种表示方法
# ============================================

class GraphAdjList:
    '''\u56fe - \u90bb\u63a5\u8868\u5b9e\u73b0'''
    
    def __init__(self, directed=False):
        self.adj = {}  # \u90bb\u63a5\u8868: {\u9876\u70b9: [(\u76f8\u90bb\u9876\u70b9, \u6743\u91cd), ...]}
        self.directed = directed
    
    def add_vertex(self, vertex):
        if vertex not in self.adj:
            self.adj[vertex] = []
    
    def add_edge(self, u, v, weight=1):
        self.add_vertex(u)
        self.add_vertex(v)
        self.adj[u].append((v, weight))
        if not self.directed:
            self.adj[v].append((u, weight))
    
    def display(self):
        for vertex in sorted(self.adj.keys()):
            neighbors = [(v, w) for v, w in self.adj[vertex]]
            print(f'  {vertex}: {neighbors}')
    
    def bfs(self, start):
        '''\u5e7f\u5ea6\u4f18\u5148\u641c\u7d22 (BFS)'''
        visited = set()
        queue = [start]
        visited.add(start)
        order = []
        while queue:
            vertex = queue.pop(0)
            order.append(vertex)
            for neighbor, _ in self.adj.get(vertex, []):
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)
        return order
    
    def dfs(self, start, visited=None):
        '''\u6df1\u5ea6\u4f18\u5148\u641c\u7d22 (DFS)'''
        if visited is None:
            visited = set()
        visited.add(start)
        order = [start]
        for neighbor, _ in self.adj.get(start, []):
            if neighbor not in visited:
                order.extend(self.dfs(neighbor, visited))
        return order

# \u6f14\u793a\uff1a\u57ce\u5e02\u4ea4\u901a\u7f51\u7edc
g = GraphAdjList()
edges = [
    ('A', 'B', 4), ('A', 'C', 2),
    ('B', 'D', 3), ('B', 'E', 1),
    ('C', 'D', 5), ('C', 'F', 6),
    ('D', 'E', 2), ('E', 'F', 4)
]

for u, v, w in edges:
    g.add_edge(u, v, w)

print('=== \u56fe\u7684\u90bb\u63a5\u8868\u8868\u793a ===')
g.display()

print(f'\nBFS\u4ece A \u5f00\u59cb: {g.bfs("A")}')
````

````python
# ============================================
# 练习4: 实现链表反转
# ============================================

def reverse_linked_list(ll):
    '''反转链表 - O(n)时间, O(1)空间'''
    prev = None
    current = ll.head
    
    while current is not None:
        next_node = current.next  # 保存下一个节点
        current.next = prev       # 反转指针方向
        prev = current            # 前移prev
        current = next_node       # 前移current
    
    ll.head = prev  # 更新头指针
    return ll

# 测试
ll = LinkedList()
for v in [1, 2, 3, 4, 5]:
    ll.append(v)
print(f'原始链表: {ll}')

reverse_linked_list(ll)
print(f'反转后:   {ll}')

print('\n\u2705 \u94fe\u8868\u53cd\u8f6c\u6210\u529f\uff01')
````
