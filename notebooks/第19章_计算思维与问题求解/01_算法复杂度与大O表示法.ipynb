## 1. 为什么需要衡量算法效率？

### 1.1 一个真实的问题

假设你要在一个有 **100万条记录** 的数据库中查找一个学生的信息：

- **方法A**：从头到尾逐一检查 → 最坏情况需要检查 **100万次**
- **方法B**：使用二分查找 → 最坏情况只需要检查约 **20次**（因为 log₂(1,000,000) ≈ 20）

两种方法都能得到正确答案，但效率差了 **5万倍**！

### 1.2 为什么不能只看运行时间？

你可能会想："直接用秒表计时不就行了？" 但这有几个问题：

| 问题 | 说明 |
|------|------|
| **硬件差异** | 同一算法在不同电脑上运行时间不同 |
| **数据差异** | 同一算法对不同输入数据表现不同 |
| **规模差异** | 小数据量时差异不明显，大数据量才能看出区别 |

所以我们需要一种 **与硬件无关、与具体数据无关** 的方式来描述算法效率 → 这就是 **大O表示法 (Big O Notation)**。

### 1.3 核心思想

> **大O表示法描述的是：当输入规模 n 趋向无穷大时，算法的运行时间（或空间）如何增长。**

它关注的不是精确的操作次数，而是 **增长的趋势(growth rate)**。

## 2. 大O表示法 (Big O Notation) 详解

### 2.1 什么是大O？

大O表示法用一个函数来描述算法的 **最坏情况 (worst case)** 时间复杂度。

**规则：**
1. **忽略常数系数**：O(2n) → O(n)，O(100) → O(1)
2. **只保留最高阶项**：O(n² + n) → O(n²)
3. **关注n趋向无穷时的行为**

### 为什么要忽略常数？

因为当 n 足够大时，**增长率** 才是决定性因素：
- 当 n = 1000 时：n² = 1,000,000 而 100n = 100,000
- 当 n = 10000 时：n² = 100,000,000 而 100n = 1,000,000

无论常数多大，高阶项终究会支配一切。

### 2.2 常见的复杂度类型

| 复杂度 | 名称 | 例子 | 特点 |
|--------|------|------|------|
| O(1) | 常数 Constant | 数组按索引访问 | 无论n多大，时间不变 |
| O(log n) | 对数 Logarithmic | 二分查找 | 每次排除一半数据 |
| O(n) | 线性 Linear | 遍历数组 | 时间与n成正比 |
| O(n log n) | 线性对数 Linearithmic | 归并排序、快速排序 | 最优比较排序 |
| O(n²) | 平方 Quadratic | 冒泡排序、嵌套循环 | 两层嵌套循环 |
| O(2ⁿ) | 指数 Exponential | 暴力求解旅行商问题 | 每增加1个输入，时间翻倍 |

### 2.3 每种复杂度的直观理解

让我们通过具体的代码来理解每种复杂度：

````python
# ============================================
# 每种复杂度的代码示例
# ============================================

# O(1) - 常数时间：无论数组多大，访问时间都一样
def constant_example(arr):
    '''访问数组的第一个元素'''
    return arr[0]  # 直接通过索引访问，1步完成

# O(log n) - 对数时间：每次操作排除一半数据
def binary_search(arr, target):
    '''二分查找 - 前提是数组已排序'''
    low, high = 0, len(arr) - 1
    steps = 0
    while low <= high:
        steps += 1
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid, steps
        elif arr[mid] < target:
            low = mid + 1  # 排除左半部分
        else:
            high = mid - 1  # 排除右半部分
    return -1, steps

# O(n) - 线性时间：需要检查每个元素
def linear_search(arr, target):
    '''线性查找'''
    steps = 0
    for i in range(len(arr)):
        steps += 1
        if arr[i] == target:
            return i, steps
    return -1, steps

# O(n^2) - 平方时间：嵌套循环
def bubble_sort(arr):
    '''冒泡排序'''
    arr = arr.copy()
    steps = 0
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            steps += 1
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr, steps

# 演示
import random
sizes = [100, 1000, 10000]
print('=' * 65)
print(f'{"n":>8} | {"\u7ebf\u6027\u67e5\u627e\u6b65\u6570":>12} | {"\u4e8c\u5206\u67e5\u627e\u6b65\u6570":>12} | {"\u6bd4\u503c":>8}')
print('=' * 65)
for size in sizes:
    arr = list(range(size))
    target = size - 1  # 最坏情况
    _, lin_steps = linear_search(arr, target)
    _, bin_steps = binary_search(arr, target)
    ratio = lin_steps / bin_steps if bin_steps > 0 else 0
    print(f'{size:>8} | {lin_steps:>12} | {bin_steps:>12} | {ratio:>8.0f}x')
print('=' * 65)
````

## 3. 时间复杂度 vs 空间复杂度

### 3.1 两个维度

| 维度 | 含义 | 关注点 |
|------|------|--------|
| **时间复杂度 (Time Complexity)** | 算法执行需要多少时间 | 操作次数随n的增长 |
| **空间复杂度 (Space Complexity)** | 算法需要多少额外内存 | 额外内存随n的增长 |

### 3.2 为什么空间复杂度也很重要？

- 在嵌入式设备（如手机、IoT设备）中，内存非常有限
- 有时候我们会 **用空间换时间**（Space-Time Tradeoff），比如缓存、哈希表
- 有时候我们会 **用时间换空间**，比如重新计算而不是存储结果

### 3.3 例子

```python
# 方法1：时间O(n^2)，空间O(1)
# 检查数组是否有重复元素 - 不使用额外空间
for i in range(n):
    for j in range(i+1, n):
        if arr[i] == arr[j]: return True

# 方法2：时间O(n)，空间O(n)
# 使用集合（哈希表）- 用空间换时间
seen = set()
for x in arr:
    if x in seen: return True
    seen.add(x)
```

````python
# ============================================
# 时间 vs 空间复杂度的实际对比
# ============================================
import time

def has_duplicates_brute(arr):
    '''方法1: O(n^2)时间, O(1)空间'''
    comparisons = 0
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            comparisons += 1
            if arr[i] == arr[j]:
                return True, comparisons
    return False, comparisons

def has_duplicates_set(arr):
    '''方法2: O(n)时间, O(n)空间'''
    seen = set()
    comparisons = 0
    for x in arr:
        comparisons += 1
        if x in seen:
            return True, comparisons
        seen.add(x)
    return False, comparisons

# 测试
import random
sizes = [100, 500, 1000, 2000, 5000]
brute_times = []
set_times = []

print(f'{"n":>6} | {"\u66b4\u529b\u6cd5\u65f6\u95f4(ms)":>14} | {"\u54c8\u5e0c\u96c6\u65f6\u95f4(ms)":>14} | {"\u52a0\u901f\u6bd4":>8}')
print('-' * 55)

for size in sizes:
    arr = list(range(size))  # 无重复，最坏情况
    
    start = time.perf_counter()
    has_duplicates_brute(arr)
    t1 = (time.perf_counter() - start) * 1000
    brute_times.append(t1)
    
    start = time.perf_counter()
    has_duplicates_set(arr)
    t2 = (time.perf_counter() - start) * 1000
    set_times.append(t2)
    
    speedup = t1 / t2 if t2 > 0 else float('inf')
    print(f'{size:>6} | {t1:>14.3f} | {t2:>14.3f} | {speedup:>7.1f}x')

````

````python
# ============================================
# 可视化：时间-空间权衡对比
# ============================================
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['WenQuanYi Zen Hei', 'Noto Sans CJK SC', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

fig, ax = plt.subplots(figsize=(10, 6))

x = np.array(sizes)
width = np.array([s * 0.08 for s in sizes])

bars1 = ax.bar(np.arange(len(sizes)) - 0.2, brute_times, 0.35,
               label='\u66b4\u529b\u6cd5 O(n\u00b2) - \u65f6\u95f4\u957f\u4f46\u7701\u7a7a\u95f4', color='#e74c3c', alpha=0.8)
bars2 = ax.bar(np.arange(len(sizes)) + 0.2, set_times, 0.35,
               label='\u54c8\u5e0c\u96c6 O(n) - \u65f6\u95f4\u77ed\u4f46\u8017\u7a7a\u95f4', color='#2ecc71', alpha=0.8)

ax.set_xlabel('\u6570\u636e\u89c4\u6a21 n', fontsize=12)
ax.set_ylabel('\u8fd0\u884c\u65f6\u95f4 (ms)', fontsize=12)
ax.set_title('\u65f6\u95f4-\u7a7a\u95f4\u6743\u8861 (Time-Space Tradeoff)', fontsize=14, fontweight='bold')
ax.set_xticks(range(len(sizes)))
ax.set_xticklabels([str(s) for s in sizes])
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
````

## 4. 常见算法的复杂度分析

### 4.1 查找算法

| 算法 | 时间复杂度（最坏） | 时间复杂度（平均） | 空间复杂度 | 前提条件 |
|------|---------------------|---------------------|------------|----------|
| 线性查找 (Linear Search) | O(n) | O(n) | O(1) | 无 |
| 二分查找 (Binary Search) | O(log n) | O(log n) | O(1) | 数据已排序 |
| 哈希查找 (Hash Lookup) | O(n) | O(1) | O(n) | 哈希表 |

### 4.2 排序算法

| 算法 | 最好 | 平均 | 最坏 | 空间 | 稳定性 |
|------|------|------|------|------|--------|
| 冒泡排序 (Bubble Sort) | O(n) | O(n²) | O(n²) | O(1) | 稳定 |
| 插入排序 (Insertion Sort) | O(n) | O(n²) | O(n²) | O(1) | 稳定 |
| 归并排序 (Merge Sort) | O(n log n) | O(n log n) | O(n log n) | O(n) | 稳定 |
| 快速排序 (Quick Sort) | O(n log n) | O(n log n) | O(n²) | O(log n) | 不稳定 |

### 4.3 数据结构操作

| 操作 | 数组 | 链表 | 哈希表 | BST（平衡） |
|------|------|------|--------|-------------|
| 查找 | O(n) | O(n) | O(1)平均 | O(log n) |
| 插入 | O(n) | O(1) | O(1)平均 | O(log n) |
| 删除 | O(n) | O(1) | O(1)平均 | O(log n) |

````python
# ============================================
# 实际测量：验证理论复杂度
# ============================================
import time
import random
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['WenQuanYi Zen Hei', 'Noto Sans CJK SC', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

def measure_time(func, *args, repeats=3):
    '''测量函数运行时间，取多次平均'''
    times = []
    for _ in range(repeats):
        start = time.perf_counter()
        func(*args)
        times.append(time.perf_counter() - start)
    return min(times) * 1000  # 返回最快时间(ms)

def linear_search_simple(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def binary_search_simple(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# 测量不同规模下的查找时间
sizes = [1000, 5000, 10000, 50000, 100000, 500000]
linear_times = []
binary_times = []

for size in sizes:
    arr = list(range(size))
    target = size - 1  # 最坏情况
    
    lt = measure_time(linear_search_simple, arr, target)
    bt = measure_time(binary_search_simple, arr, target)
    linear_times.append(lt)
    binary_times.append(bt)

# 可视化
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 左图：实际测量时间
ax1 = axes[0]
ax1.plot(sizes, linear_times, 'ro-', linewidth=2, markersize=8, label='\u7ebf\u6027\u67e5\u627e O(n)')
ax1.plot(sizes, binary_times, 'bs-', linewidth=2, markersize=8, label='\u4e8c\u5206\u67e5\u627e O(log n)')
ax1.set_xlabel('\u6570\u636e\u89c4\u6a21 n', fontsize=12)
ax1.set_ylabel('\u8fd0\u884c\u65f6\u95f4 (ms)', fontsize=12)
ax1.set_title('\u5b9e\u9645\u6d4b\u91cf\u65f6\u95f4', fontsize=13, fontweight='bold')
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)

# 右图：理论预测 vs 实际
ax2 = axes[1]
# 归一化到第一个数据点进行比较
n_arr = np.array(sizes, dtype=float)
theory_linear = n_arr / n_arr[0] * linear_times[0]
theory_log = np.log2(n_arr) / np.log2(n_arr[0]) * binary_times[0]

ax2.plot(sizes, linear_times, 'ro', markersize=8, label='\u7ebf\u6027\u67e5\u627e(\u5b9e\u6d4b)')
ax2.plot(sizes, theory_linear, 'r--', linewidth=1.5, label='O(n) \u7406\u8bba\u9884\u6d4b')
ax2.plot(sizes, binary_times, 'bs', markersize=8, label='\u4e8c\u5206\u67e5\u627e(\u5b9e\u6d4b)')
ax2.plot(sizes, theory_log, 'b--', linewidth=1.5, label='O(log n) \u7406\u8bba\u9884\u6d4b')
ax2.set_xlabel('\u6570\u636e\u89c4\u6a21 n', fontsize=12)
ax2.set_ylabel('\u8fd0\u884c\u65f6\u95f4 (ms)', fontsize=12)
ax2.set_title('\u7406\u8bba\u9884\u6d4b vs \u5b9e\u9645\u6d4b\u91cf', fontsize=13, fontweight='bold')
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## 5. 深入：为什么O(n log n)是基于比较的排序的下界？

### 5.1 决策树论证

这是一个非常优雅的数学论证：

1. **任何基于比较的排序**都可以用一棵 **决策树 (Decision Tree)** 来表示
2. 对于n个元素，有 **n!** 种可能的排列方式
3. 决策树的叶子节点数 ≥ n!（每种排列至少对应一个叶子）
4. 高度为h的二叉树最多有 **2^h** 个叶子
5. 所以：2^h ≥ n!
6. 取对数：h ≥ log₂(n!)
7. 由Stirling近似：log₂(n!) ≈ n·log₂(n) - n·log₂(e) = **O(n log n)**

### 5.2 这意味着什么？

> **没有任何基于比较的排序算法能比 O(n log n) 更快！**

- 归并排序和快速排序（平均情况）已经达到了这个理论最优值
- 这就是为什么它们被称为"高效排序算法"
- 注意：非比较的排序（如计数排序、基数排序）可以做到O(n)，但有其他限制条件

## 6. 如何分析代码的复杂度？

### 6.1 基本规则

1. **顺序语句**：取最大的 → O(f(n)) + O(g(n)) = O(max(f(n), g(n)))
2. **单层循环**：循环次数 → 通常是O(n)
3. **嵌套循环**：相乘 → 两层O(n)循环 = O(n²)
4. **每次减半的循环**：O(log n)
5. **递归**：需要分析递归关系式

````python
# ============================================
# 练习：分析以下代码的复杂度
# ============================================

# 例1：这是什么复杂度？
def example1(n):
    count = 0
    for i in range(n):       # n次
        for j in range(n):   # 每次又n次
            count += 1       # 总共 n*n = n^2 次
    return count

# 例2：这是什么复杂度？
def example2(n):
    count = 0
    i = n
    while i > 1:
        count += 1
        i = i // 2           # 每次减半
    return count

# 例3：这是什么复杂度？
def example3(n):
    count = 0
    for i in range(n):       # n次
        j = n
        while j > 1:         # 每次log(n)次
            count += 1
            j = j // 2
    return count

# 验证
for n in [10, 100, 1000]:
    print(f'n={n:>5}: example1={example1(n):>10} (n\u00b2={n**2:>10}) | '
          f'example2={example2(n):>4} (log\u2082n\u2248{int(np.log2(n)):>3}) | '
          f'example3={example3(n):>8} (n\u00b7log\u2082n\u2248{int(n*np.log2(n)):>8})')

print('\n\u7b54\u6848\uff1a')
print('example1: O(n\u00b2) - \u4e24\u5c42\u5d4c\u5957\u5faa\u73af')
print('example2: O(log n) - \u6bcf\u6b21\u51cf\u534a')
````

### 练习3：编程挑战
编写两个函数来查找数组中的两个数之和等于目标值：
1. 一个O(n²)的暴力解法
2. 一个O(n)的哈希表解法
3. 用不同大小的数组测试并比较两者的性能

````python
# ============================================
# 练习3的参考答案
# ============================================
import time

# 方法1: O(n^2) 暴力解法
def two_sum_brute(arr, target):
    '''暴力查找：检查所有配对'''
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] + arr[j] == target:
                return (i, j)
    return None

# 方法2: O(n) 哈希表解法
def two_sum_hash(arr, target):
    '''哈希表查找：用字典记录已见过的数'''
    seen = {}  # 值 -> 索引
    for i, num in enumerate(arr):
        complement = target - num
        if complement in seen:  # O(1)查找
            return (seen[complement], i)
        seen[num] = i
    return None

# 测试正确性
test_arr = [2, 7, 11, 15, 1, 8]
target = 9
print(f'测试数组: {test_arr}, 目标和: {target}')
print(f'暴力法结果: {two_sum_brute(test_arr, target)}')
print(f'哈希法结果: {two_sum_hash(test_arr, target)}')

# 性能对比
print(f'\n{"n":>8} | {"暴力法(ms)":>12} | {"哈希法(ms)":>12} | {"加速比":>8}')
print('-' * 50)
import random
for size in [1000, 5000, 10000, 20000]:
    arr = list(range(size))
    random.shuffle(arr)
    target = arr[-1] + arr[-2]  # 确保有解
    
    start = time.perf_counter()
    two_sum_brute(arr, target)
    t1 = (time.perf_counter() - start) * 1000
    
    start = time.perf_counter()
    two_sum_hash(arr, target)
    t2 = (time.perf_counter() - start) * 1000
    
    ratio = t1 / t2 if t2 > 0 else float('inf')
````

## 本节小结 (Summary)

| 核心概念 | 要点 |
|----------|------|
| **大O表示法** | 描述算法效率随输入规模增长的趋势，忽略常数和低阶项 |
| **时间复杂度** | 算法执行时间随n的增长关系 |

| **分析方法** | 数循环层数、看每次操作规模变化、分析递归关系 |
