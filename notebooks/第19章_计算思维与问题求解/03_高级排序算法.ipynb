## 1. 分治思想 (Divide and Conquer)

### 1.1 什么是分治？

分治是一种强大的算法设计策略，包含三个步骤：

1. **分 (Divide)**：将大问题分解为若干个较小的子问题
2. **治 (Conquer)**：递归地解决每个子问题
3. **合 (Combine)**：将子问题的解合并为原问题的解

### 1.2 为什么分治能提高效率？

直觉上理解：
- 冒泡排序对n个元素需要 n² 次比较
- 如果我们把n个元素分成两半，各自排序，再合并呢？
  - 每半边排序：2 × (n/2)² = n²/2 次比较
  - 合并：n 次比较
  - 总计：n²/2 + n → 比 n² 少了！
- 如果我们**继续递归地分下去**，就能达到 O(n log n)！

这就是归并排序和快速排序的核心思想。

## 2. 归并排序 (Merge Sort)

### 2.1 算法思路

1. **分**：将数组从中间分成左右两半
2. **治**：递归地对左半和右半分别排序
3. **合**：将两个已排序的子数组合并为一个有序数组

### 2.2 关键操作：合并 (Merge)

合并两个有序数组的过程：
- 用两个指针分别指向左右数组的开头
- 比较两个指针所指的元素，较小的放入结果数组
- 移动该指针，重复直到所有元素都放入结果

### 2.3 复杂度分析

| 情况 | 时间复杂度 | 说明 |
|------|-----------|------|
| 最好 | O(n log n) | 总是需要分和合 |
| 平均 | O(n log n) | 稳定的性能 |
| 最坏 | O(n log n) | 不受输入数据影响 |
| 空间 | O(n) | 需要额外数组存储合并结果 |

**为什么是O(n log n)？**
- 分的层数 = log₂(n)（每次减半）
- 每层合并的总工作量 = O(n)
- 总计 = O(n) × O(log n) = O(n log n)

```python
# ============================================
# 归并排序的完整实现（带详细注释）
# ============================================

def merge_sort(arr, depth=0):
    '''
    归并排序 (Merge Sort)
    - 分治策略：先分后合
    - 时间复杂度: O(n log n)
    - 空间复杂度: O(n)
    - 稳定排序
    '''
    # 基线条件：长度为0或1的数组已经是有序的
    if len(arr) <= 1:
        return arr
    
    # 第1步：分 (Divide) - 找中间点，分成两半
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]
    
    indent = '  ' * depth
    if len(arr) <= 16:  # 小数组时打印过程
        print(f'{indent}分: {arr} -> {left_half} | {right_half}')
    
    # 第2步：治 (Conquer) - 递归排序左右两半
    left_sorted = merge_sort(left_half, depth + 1)
    right_sorted = merge_sort(right_half, depth + 1)
    
    # 第3步：合 (Combine) - 合并两个有序数组
    merged = merge(left_sorted, right_sorted)
    
    if len(arr) <= 16:
        print(f'{indent}合: {left_sorted} + {right_sorted} -> {merged}')
    
    return merged

def merge(left, right):
    '''合并两个有序数组'''
    result = []
    i = j = 0
    
    # 比较两个数组的元素，较小的先放入结果
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:  # <= 保证稳定性
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    # 将剩余元素加入结果
    result.extend(left[i:])
    result.extend(right[j:])
    return result

# 演示
print('=' * 50)
print('归并排序过程演示')
print('=' * 50)
test = [38, 27, 43, 3, 9, 82, 10]
print(f'\n原始数组: {test}\n')
result = merge_sort(test)
```

## 3. 快速排序 (Quick Sort)

### 3.1 算法思路

1. **选基准 (Pivot)**：从数组中选择一个元素作为基准
2. **分区 (Partition)**：将数组分为三部分
   - 小于基准的元素
   - 等于基准的元素
   - 大于基准的元素
3. **递归**：对小于和大于基准的两部分分别递归排序

### 3.2 与归并排序的区别

| 特性 | 归并排序 | 快速排序 |
|------|----------|----------|
| **分的方式** | 按位置（中间切开） | 按值（基准分区） |
| **主要工作** | 在"合"的阶段 | 在"分"的阶段 |
| **额外空间** | O(n) | O(log n) 平均 |
| **最坏情况** | O(n log n) | O(n²) |
| **稳定性** | 稳定 | 不稳定 |
| **实际速度** | 较快 | 通常更快（常数因子小） |

### 3.3 为什么快排的最坏情况是O(n²)？

当每次选的基准恰好是最大或最小元素时：
- 分区只能排除1个元素
- 需要分n次，每次扫描剩余元素
- 总计：n + (n-1) + (n-2) + ... + 1 = n(n+1)/2 = O(n²)

**如何避免？** 随机选择基准 (Random Pivot) 或三数取中法 (Median of Three)

```python
# ============================================
# 快速排序的完整实现（带详细注释）
# ============================================

def quick_sort(arr, depth=0):
    '''
    快速排序 (Quick Sort)
    - 选基准、分区、递归
    - 时间复杂度: 平均O(n log n), 最坏O(n²)
    - 空间复杂度: O(log n) 平均
    - 不稳定排序
    '''
    # 基线条件
    if len(arr) <= 1:
        return arr
    
    indent = '  ' * depth
    
    # 第1步：选择基准（这里选最后一个元素）
    pivot = arr[-1]
    
    # 第2步：分区
    left = [x for x in arr[:-1] if x <= pivot]   # 小于等于基准的
    right = [x for x in arr[:-1] if x > pivot]    # 大于基准的
    
    if len(arr) <= 12:
        print(f'{indent}基准={pivot}: {arr} -> [{left}] [{pivot}] [{right}]')
    
    # 第3步：递归排序左右两部分，然后拼接
    return quick_sort(left, depth+1) + [pivot] + quick_sort(right, depth+1)

# 演示
print('=' * 60)
print('快速排序过程演示')
print('=' * 60)
test = [38, 27, 43, 3, 9, 82, 10]
print(f'\n原始数组: {test}\n')
result = quick_sort(test)
```

## 4. 归并排序 vs 快速排序 vs 冒泡排序 性能对比

让我们用实际数据来验证理论分析：

```python
# ============================================
# 三种排序算法的性能实测对比
# ============================================
import time
import random
import sys

# 增加递归深度限制
sys.setrecursionlimit(20000)

def merge_sort_clean(arr):
    '''归并排序（不打印）'''
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort_clean(arr[:mid])
    right = merge_sort_clean(arr[mid:])
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

def quick_sort_clean(arr):
    '''快速排序（不打印，随机基准）'''
    if len(arr) <= 1:
        return arr
    pivot_idx = random.randint(0, len(arr) - 1)
    pivot = arr[pivot_idx]
    left = [x for i, x in enumerate(arr) if x < pivot or (x == pivot and i < pivot_idx)]
    middle = [x for x in arr if x == pivot]
    right = [x for i, x in enumerate(arr) if x > pivot or (x == pivot and i > pivot_idx)]
    return quick_sort_clean(left) + middle + quick_sort_clean(right)

def bubble_sort_clean(arr):
    '''冒泡排序'''
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

# 性能测试
sizes = [100, 500, 1000, 2000, 5000]
bubble_times = []
merge_times = []
quick_times = []

print(f'{"n":>6} | {"\u5192\u6ce1(ms)":>10} | {"\u5f52\u5e76(ms)":>10} | {"\u5feb\u6392(ms)":>10} | {"\u5192\u6ce1/\u5f52\u5e76":>10}')
print('-' * 65)

for size in sizes:
    arr = list(range(size))
    random.shuffle(arr)
    
    # 冒泡排序
    start = time.perf_counter()
    bubble_sort_clean(arr)
    bt = (time.perf_counter() - start) * 1000
    bubble_times.append(bt)
    
    # 归并排序
    start = time.perf_counter()
    merge_sort_clean(arr)
    mt = (time.perf_counter() - start) * 1000
    merge_times.append(mt)
    
    # 快速排序
    start = time.perf_counter()
    quick_sort_clean(arr)
    qt = (time.perf_counter() - start) * 1000
    quick_times.append(qt)
    
    ratio = bt / mt if mt > 0 else float('inf')
    print(f'{size:>6} | {bt:>10.2f} | {mt:>10.2f} | {qt:>10.2f} | {ratio:>9.1f}x')

```

## 5. 深入：Python的内置sort用的是什么算法？

### 5.1 TimSort

Python 的 `list.sort()` 和 `sorted()` 使用的是 **TimSort** 算法，由 Tim Peters 在2002年发明。

**TimSort 的核心思想：**

1. 它是 **归并排序** 和 **插入排序** 的混合体
2. 先将数组分成小块（称为 "run"），每个 run 用插入排序
3. 然后用归并排序将这些 run 合并

**为什么这样设计？**
- 插入排序在 **小数组** 上非常快（缓存友好，常数因子小）
- 归并排序在 **大数组** 上保证O(n log n)
- TimSort 结合了两者的优点！

| 特性 | TimSort |
|------|--------|
| 最好情况 | O(n) — 已排序数据 |
| 平均情况 | O(n log n) |
| 最坏情况 | O(n log n) |
| 空间复杂度 | O(n) |
| 稳定性 | 稳定 |

让我们验证 Python 内置排序比我们手写的快多少：

```python
# ============================================
# Python 内置sort vs 手写排序
# ============================================
import time
import random

sizes = [1000, 5000, 10000, 50000, 100000]
builtin_times = []
merge_manual_times = []

print(f'{"n":>8} | {"\u624b\u5199\u5f52\u5e76(ms)":>14} | {"\u5185\u7f6esorted(ms)":>14} | {"\u52a0\u901f\u6bd4":>8}')
print('-' * 55)

for size in sizes:
    arr = list(range(size))
    random.shuffle(arr)
    
    # 手写归并排序
    start = time.perf_counter()
    merge_sort_clean(arr)
    mt = (time.perf_counter() - start) * 1000
    merge_manual_times.append(mt)
    
    # 内置sorted
    start = time.perf_counter()
    sorted(arr)
    bt = (time.perf_counter() - start) * 1000
    builtin_times.append(bt)
    
    ratio = mt / bt if bt > 0 else float('inf')
    print(f'{size:>8} | {mt:>14.3f} | {bt:>14.3f} | {ratio:>7.1f}x')

print('\n\u7ed3\u8bba\uff1aPython\u5185\u7f6e\u7684sorted()\u6bd4\u624b\u5199\u7684\u5f52\u5e76\u6392\u5e8f\u5feb\u5f88\u591a\uff01')
```

```python
# ============================================
# 练习3: 实现原地快速排序 (In-place Quick Sort)
# ============================================

def quick_sort_inplace(arr, low=0, high=None):
    '''
    原地快速排序 - 不需要额外数组空间
    提示：需要一个partition函数来原地分区
    '''
    if high is None:
        high = len(arr) - 1
    
    if low < high:
        # TODO: 实现分区并获取基准位置
        pivot_idx = partition(arr, low, high)
        # TODO: 递归排序左右两部分
        quick_sort_inplace(arr, low, pivot_idx - 1)
        quick_sort_inplace(arr, pivot_idx + 1, high)

def partition(arr, low, high):
    '''
    分区函数：
    - 选arr[high]为基准
    - 将小于基准的放左边，大于基准的放右边
    - 返回基准的最终位置
    '''
    pivot = arr[high]
    i = low - 1  # i指向"小于区"的最后一个元素
    
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]  # 交换
    
    arr[i + 1], arr[high] = arr[high], arr[i + 1]  # 把基准放到正确位置
    return i + 1

# 测试
import random
test = [38, 27, 43, 3, 9, 82, 10]
print(f'原始: {test}')
quick_sort_inplace(test)
print(f'排序: {test}')

# 验证
test2 = list(range(100))
random.shuffle(test2)
quick_sort_inplace(test2)
assert test2 == list(range(100)), '排序结果不正确！'
```

## 本节小结 (Summary)

| 核心概念 | 要点 |
|----------|------|
| **分治思想** | 分解 → 递归解决 → 合并，是高效算法的关键 |
| **归并排序** | 稳定的O(n log n)，需要O(n)额外空间 |
| **快速排序** | 平均O(n log n)，原地排序，但最坏O(n²) |
| **基准选择** | 随机基准可以避免快排的最坏情况 |
| **TimSort** | Python内置排序，结合归并和插入排序的优点 |
