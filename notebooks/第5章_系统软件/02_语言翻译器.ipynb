## 1. 翻译问题 (The Translation Problem)

### 1.1 语言的鸿沟 (The Language Gap)

这是计算机科学中最基本的问题之一 (*one of the most fundamental problems in CS*)：

- **人类 (Humans)** 用接近自然语言的 **高级语言 (High-Level Language, HLL)** 编程
  - 例子 (Example)：`if score > 90: print("Excellent!")`
- **CPU** 只理解 **机器码 (Machine Code)** —— 二进制的 0 和 1 (*binary 0s and 1s*)
  - 例子：`10110000 01100001`

这就像一个中国人想和一个只懂阿拉伯语的人交流 —— **需要翻译官 (we need a translator)！**

### 1.2 语言的层级 (The Hierarchy of Languages)

| 层级 (Level) | 英文 | 例子 | 特点 |
|---|---|---|---|
| 第 5 代 | **Natural Language** | "Please print 'Hello'" | 自然语言，仍在研究 |
| 第 4 代 | **4GL (Fourth-Generation Language)** | SQL, MATLAB | 特定领域，声明式 |
| 第 3 代 | **High-Level Language (HLL)** | Python, Java, C | 接近英语，通用 |
| 第 2 代 | **Assembly Language** | `MOV AL, 61h` | 机器码的助记符 |
| 第 1 代 | **Machine Code** | `10110000 01100001` | CPU 直接执行的二进制 |

### 1.3 为什么不直接用机器码编程？(Why Not Program in Machine Code Directly?)

早期的程序员确实是直接用机器码编程的 (*early programmers did this*)。问题在于：

| 问题 (Problem) | 说明 (Explanation) |
|------|------|
| 极难编写 (Hard to write) | 必须记住每条指令的二进制编码 |
| 极难阅读 (Unreadable) | `10110000 01100001` 究竟是什么意思？ |
| 极难调试 (Hard to debug) | 一个 0 或 1 打错了，几乎不可能定位 |
| 不可移植 (Not portable) | 不同 CPU 架构有不同机器码 (*x86 vs ARM vs RISC-V*) |

> **定义 (Definition)：Language Translator (语言翻译器)** — 一种系统软件，把人类可读的代码 **翻译 (translate)** 成 CPU 可执行的机器码。
>
> *A **Language Translator** is system software that translates human-readable code into machine code the CPU can execute.*

## 2. 三种翻译器 (Three Types of Translators)

### 2.1 汇编器 (Assembler)

> **定义 (Definition)：Assembler (汇编器)** — 把 **汇编语言 (Assembly Language)** 翻译成 **机器码 (Machine Code)** 的程序，每条汇编指令 **一一对应 (1-to-1 mapping)** 一条机器码指令。
>
> *An **Assembler** translates Assembly Language into Machine Code, with a one-to-one correspondence between each assembly instruction and machine code instruction.*

```
Assembly:     MOV AL, 61h      ; 把数值 61h 放入 AL 寄存器
                  |            ; (move value 61h into AL register)
                  | Assembler 翻译 (translate)
                  v
Machine Code: 10110000 01100001
```

**关键特性 (Key Features)：**
- 翻译很 **直接 (direct)** —— 每条 **助记符 (mnemonic)** 对应一条二进制指令
- 汇编语言是 **低级语言 (Low-Level Language)**，与硬件紧密相关 (*tightly coupled to hardware*)
- 我们在第 4 章已经学过汇编语言的基本概念

### 2.2 编译器 (Compiler)

> **定义 (Definition)：Compiler (编译器)** — 将 **整个高级语言程序 (the entire HLL program)** **一次性 (in one go)** 翻译成机器码或目标代码。
>
> *A **Compiler** translates the entire high-level source program into machine code or object code in one go.*

```
源代码 (Source Code)          可执行文件 (Executable File)
  program.c          --->      program.exe
  (human-readable)  Compiler   (machine-executable)
```

**关键特性 (Key Features)：**
- 翻译 **整个程序 (whole program)** 后再运行 —— 编译和执行是分开的两个阶段
- 产出 **独立的可执行文件 (standalone executable)** (.exe)，运行时不再需要编译器
- 编译过程 **较慢 (slow)**，但编译后的程序 **运行很快 (fast execution)**
- **一次编译，多次运行 (Compile Once, Run Many Times)** —— 不改源代码就不需要重新编译
- 编译完成后给出 **所有错误的完整报告 (full error report)**

> **类比 (Analogy)：** 编译器就像翻译一本书 (*translating a book*) —— 翻译整本书需要很长时间，但翻译完后任何人都可以直接读翻译版，不需要翻译官在场。

**使用编译器的语言 (Compiled Languages)：** C, C++, Go, Rust, Swift

### 2.3 解释器 (Interpreter)

> **定义 (Definition)：Interpreter (解释器)** — **逐行 (line by line)** 翻译并执行高级语言代码的程序。
>
> *An **Interpreter** translates and executes high-level language code line by line.*

```
Line 1: x = 10        ---> 翻译 (translate) ---> 执行 (execute) ---> 结果 (result)
Line 2: y = 20        ---> 翻译              ---> 执行              ---> 结果
Line 3: print(x + y)  ---> 翻译              ---> 执行              ---> 30
```

**关键特性 (Key Features)：**
- **逐行 (line-by-line)** 翻译并立即执行 —— 翻译和执行交替进行 (*interleaved*)
- **不产生 (no)** 独立可执行文件 —— 每次运行都需要解释器
- 整体运行 **较慢 (slower)**（每次运行都要重新翻译）
- 遇到第一个错误就 **立即停止 (halts immediately)** —— 适合调试
- 提供 **即时反馈 (immediate feedback)** —— 改一行代码就能马上看到效果

> **类比 (Analogy)：** 解释器就像 **同声传译 (simultaneous interpreter)** —— 说一句翻译一句，速度较慢，但可以随时停下来纠正错误。翻译官必须一直在场 (*always present*)。

**使用解释器的语言 (Interpreted Languages)：** Python*, JavaScript*, Ruby, PHP

*注：Python 和 JavaScript 实际使用 **混合方式 (hybrid)**，下面详细讲解。

## 3. 编译器 vs 解释器：详细对比 (Compiler vs Interpreter: Detailed Comparison)

| 特征 (Feature) | Compiler (编译器) | Interpreter (解释器) |
|------|-------------------|---------------------|
| **翻译方式 (Translation)** | 一次翻译整个程序 (*translates the whole program at once*) | 逐行翻译并执行 (*translates line by line*) |
| **输出 (Output)** | 独立可执行文件 (*standalone executable*, .exe) | 无独立可执行文件 (*no standalone output*) |
| **翻译速度 (Translation Speed)** | 慢 (*slow* — analyses whole program) | 快 (*fast* — one line at a time) |
| **执行速度 (Execution Speed)** | 快 (*fast* — already machine code) | 慢 (*slow* — re-translates each run) |
| **错误报告 (Error Reporting)** | 编译后给出完整错误列表 (*full list after compilation*) | 遇第一个错误即停止 (*stops at first error*) |
| **运行时需要 (Runtime Need)** | 不需要编译器 (*compiler not needed*) | 必须有解释器 (*interpreter required*) |
| **代码安全 (Code Security)** | 只分发机器码，源码保密 (*source hidden*) | 必须分发源码 (*source must be distributed*) |
| **调试 (Debugging)** | 改一行要重新编译整个程序 | 改一行立即测试 (*immediate testing*) |
| **内存使用 (Memory Usage)** | 运行时内存占用小 | 解释器本身占用内存 |
| **适用阶段 (Best For)** | **生产/发布 (Production / Release)** | **开发/调试 (Development / Debugging)** |
| **代表语言 (Examples)** | C, C++, Go, Rust, Swift | Python\*, JavaScript\*, Ruby, PHP |

\*Python 和 JavaScript 实际上使用 **混合方式 (hybrid approach)**，下面会详细讲解。

### 3.1 中间代码 (Intermediate Code)

许多现代编译器不会直接生成机器码，而是先生成 **中间代码 (Intermediate Code)** 或 **目标代码 (Object Code)**，再由另一个工具处理：

```
Source Code --[Compiler Front-End]--> Intermediate Code --[Compiler Back-End / Linker]--> Machine Code
源代码      ---[编译器前端]         --> 中间代码          ---[编译器后端 / 链接器]       --> 机器码
```

- **编译器前端 (Compiler Front-End)** — 词法、语法、语义分析
- **编译器后端 (Compiler Back-End)** — 代码优化、目标代码生成
- **链接器 (Linker)** — 把多个目标文件和库链接成最终可执行文件
- **加载器 (Loader)** — 把可执行文件加载进内存并启动

## 4. 什么时候用哪个？(When to Use Which?)

### 4.1 开发阶段 (Development Phase) → 偏好 Interpreter

- 程序员需要 **快速反馈 (rapid feedback)** —— 改一行看效果
- 需要 **逐行调试 (line-by-line debugging)** —— 定位 bug
- **不需要 (not critical)** 最佳运行速度

### 4.2 生产/发布阶段 (Production / Release Phase) → 偏好 Compiler

- 程序需要 **最快运行速度 (maximum execution speed)**
- 需要 **保护源代码 (protect source code)** 不被用户看到
- 用户电脑上 **不需要安装 (no need to install)** 编译器/解释器

### 4.3 现实中的混合方式 (Hybrid in Practice)

> 很多现代语言都采用了 **混合方式 (hybrid approach)** 来兼顾两者的优点：
> - **Java** — `javac` 编译成 **字节码 (bytecode)**，`JVM` 解释 + **JIT 编译 (Just-In-Time Compilation)**
> - **Python** — 自动编译成 `.pyc` 字节码，`PVM` 解释执行
> - **C#** — 编译成 **CIL (Common Intermediate Language)**，由 `.NET` 运行时执行
> - **JavaScript** — V8 引擎将 JS 解析为 AST，然后 JIT 编译成机器码

### 互动实验：编译 vs 解释的速度差异概念演示

````python
import time

print('=== Compiler vs Interpreter: Speed Concept Demo ===')
print()

# Simulating "Compiler" approach
# Step 1: "Compile" - prepare all work upfront
print('--- Simulating COMPILER approach ---')
compile_start = time.perf_counter()

# "Compilation": pre-compute all results
numbers = list(range(1, 10001))
squared = [n * n for n in numbers]  # compile all at once
compile_time = time.perf_counter() - compile_start

# "Execution": just read pre-computed results
exec_start = time.perf_counter()
total = sum(squared)  # fast: already computed
exec_time = time.perf_counter() - exec_start

print(f'  Compile phase: {compile_time*1000:.3f} ms')
print(f'  Execute phase: {exec_time*1000:.3f} ms')
print(f'  Result: {total}')
print()

# Simulating "Interpreter" approach
print('--- Simulating INTERPRETER approach ---')
interp_start = time.perf_counter()

# "Interpret": translate and execute each line
total2 = 0
for n in range(1, 10001):
    # Each iteration: parse + execute (simulated)
    result = n * n
    total2 += result
interp_time = time.perf_counter() - interp_start

print(f'  Total time: {interp_time*1000:.3f} ms')
print(f'  Result: {total2}')
print()

print('Key takeaway (关键要点):')
print('  Compiler: slow upfront (compile), fast execution (run many times)')
print('  Interpreter: no upfront cost, but slower overall for repeated runs')
print('  编译器：前期慢（编译），后期快（反复运行）')
print('  解释器：无前期成本，但反复运行时整体较慢')
````

## 5. 混合翻译方式 (Hybrid Translation Approaches)

### 5.1 Java 的方式：编译 + 解释 (Java's Approach: Compile + Interpret)

Java 采用了一种非常聪明的混合方式 (*a clever hybrid approach*)：

```
Step 1: Java Compiler (javac)
  Hello.java  --->  Hello.class (Bytecode 字节码)

Step 2: JVM (Java Virtual Machine, Java 虚拟机)
  Hello.class  --->  JVM interprets / JIT compiles  --->  执行 (execute)
```

#### 什么是字节码？(What Is Bytecode?)

> **定义 (Definition)：Bytecode (字节码)** — 一种 **中间代码 (intermediate code)**，为特定 **虚拟机 (Virtual Machine)** 设计的指令集，不依赖任何具体的 CPU 架构。
>
> *Bytecode is intermediate code designed for a specific **Virtual Machine (VM)**, independent of any particular CPU architecture.*

- **不是** 特定 CPU 的机器码 (*not machine code for any specific CPU*)
- 比源代码更 **接近机器码 (closer to machine code)** —— 执行更快
- 但 **不依赖特定硬件 (hardware-independent)** —— 可跨平台运行

#### JVM (Java Virtual Machine) 做什么？(What Does the JVM Do?)

- 读取字节码并在当前平台上执行 (*executes bytecode on the current platform*)
- 可以 **解释执行 (interpret)** 字节码
- 也可以使用 **JIT 编译器 (Just-In-Time Compiler)** 将 **热点代码 (hot code)** 编译为本机机器码 —— 兼顾速度
- **垃圾回收 (Garbage Collection, GC)** — 自动管理内存

#### 好处："Write Once, Run Anywhere" (一次编写，到处运行)

- 同一个 `.class` 文件可以在 Windows、macOS、Linux 上运行
- 只需要在每个平台安装对应的 JVM 即可
- 这种特性叫做 **平台无关性 (Platform Independence)** 或 **跨平台 (Cross-Platform)**

### 5.2 Python 的方式 (Python's Approach)

Python 的工作方式与 Java 类似 (*Python works similarly to Java*)：

```
Step 1: Python Compiler (自动进行 / automatic, usually invisible)
  hello.py  --->  hello.pyc (Bytecode 字节码)

Step 2: PVM (Python Virtual Machine, Python 虚拟机)
  hello.pyc  --->  PVM interprets  --->  执行 (execute)
```

- Python 会自动将 `.py` 文件编译成 `.pyc` **字节码文件 (bytecode file)**，存放在 `__pycache__` 目录
- 然后 **PVM (Python Virtual Machine)** 解释执行这些字节码
- 这就是为什么 Python 有时被称为"解释型语言" (*interpreted language*)，但严格来说它使用了 **编译+解释的混合方式**

### 5.3 为什么混合方式好？(Why Are Hybrid Approaches Beneficial?)

| 优点 (Advantage) | 说明 (Explanation) |
|---|---|
| **速度 (Speed)** | 字节码比源代码更紧凑、执行更快 |
| **跨平台 (Portability)** | 字节码在任何有虚拟机的平台都能运行 |
| **调试友好 (Debug-friendly)** | 保留了解释器的灵活性 |
| **安全 (Security)** | 虚拟机可以做 **沙箱 (sandboxing)** |
| **优化 (Optimisation)** | JIT 可根据运行时信息进行动态优化 |

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, axes = plt.subplots(2, 1, figsize=(15, 8))

def draw_pipeline(ax, title, steps):
    ax.set_title(title, fontsize=13, fontweight='bold')
    total_width = sum(s[2] for s in steps) + 0.5 * (len(steps) - 1)
    x = (15 - total_width) / 2
    for label, color, width in steps:
        rect = patches.FancyBboxPatch((x, 0.3), width, 1.4,
                                       boxstyle='round,pad=0.15',
                                       facecolor=color, edgecolor='black', lw=2)
        ax.add_patch(rect)
        ax.text(x + width/2, 1.0, label, ha='center', va='center',
                fontsize=9, fontweight='bold')
        if x + width < 14:
            ax.annotate('', xy=(x + width + 0.5, 1.0), xytext=(x + width, 1.0),
                        arrowprops=dict(arrowstyle='->', lw=2))
        x += width + 0.5
    ax.set_xlim(0, 15)
    ax.set_ylim(-0.2, 2.3)
    ax.axis('off')

# Java pipeline
draw_pipeline(axes[0], 'Java: Compile to Bytecode + JVM (JIT)', [
    ('.java\nSource Code', '#BBDEFB', 2),
    ('javac\nCompiler', '#FFF9C4', 2),
    ('.class\nBytecode', '#C8E6C9', 2),
    ('JVM\n(Interpret/JIT)', '#FFCCBC', 2.5),
    ('Machine Code\nExecution', '#E1BEE7', 2.5),
])

# Python pipeline
draw_pipeline(axes[1], 'Python: Compile to Bytecode + PVM', [
    ('.py\nSource Code', '#BBDEFB', 2),
    ('Python\nCompiler', '#FFF9C4', 2),
    ('.pyc\nBytecode', '#C8E6C9', 2),
    ('PVM\n(Interpreter)', '#FFCCBC', 2.5),
    ('Execution\nResult', '#E1BEE7', 2.5),
])

plt.tight_layout()
plt.show()
````

### 6.5 互动实验：体验语法高亮和错误检测

````python
import re

# Simple syntax highlighter concept
print('=== How Syntax Highlighting Works (语法高亮原理) ===')
print()

sample_code = [
    'def calculate_area(radius):',
    '    pi = 3.14159',
    '    area = pi * radius ** 2',
    '    return area',
    '',
    'result = calculate_area(5)',
    'print(result)'
]

# Define token types and their "colors" (using text labels)
keywords = {'def', 'return', 'if', 'else', 'for', 'while', 'import', 'class', 'print'}
builtins_set = {'print', 'range', 'len', 'int', 'str', 'float'}

print('Token Analysis (词法分析):')
print('-' * 60)
for line in sample_code:
    if not line.strip():
        print()
        continue
    tokens = re.findall(r'[a-zA-Z_]\w*|\d+\.?\d*|[^\s]', line)
    colored_tokens = []
    for token in tokens:
        if token in keywords:
            colored_tokens.append(f'[KEYWORD:{token}]')
        elif token in builtins_set:
            colored_tokens.append(f'[BUILTIN:{token}]')
        elif re.match(r'^\d+\.?\d*$', token):
            colored_tokens.append(f'[NUMBER:{token}]')
        elif token in ('(', ')', ':', '=', '*', '+', '-'):
            colored_tokens.append(f'[OPERATOR:{token}]')
        else:
            colored_tokens.append(f'[IDENTIFIER:{token}]')
    print('  ' + '  '.join(colored_tokens))

print('-' * 60)
print()
print('IDE editors use this analysis to color different token types!')
print('IDE编辑器使用这种分析来为不同类型的词法单元着色！')
print('Keywords: blue | Numbers: green | Strings: orange | etc.')
````

## 7. 编译流程 (The Compilation Pipeline)

编译器并不是简单地"翻译" —— 它经过多个精密的阶段 (*precise stages*)：

| # | 阶段 (Stage) | English | 任务 (Task) |
|---|------|------|------|
| 1 | **词法分析** | **Lexical Analysis (Lexing)** | 将源代码分解成 **词法单元 (tokens)** |
| 2 | **语法分析** | **Syntax Analysis (Parsing)** | 检查 tokens 是否符合 **语法 (grammar)**，构建 **抽象语法树 (Abstract Syntax Tree, AST)** |
| 3 | **语义分析** | **Semantic Analysis** | 检查含义是否正确（**类型检查 / Type Checking** 等） |
| 4 | **中间代码生成** | **Intermediate Code Generation** | 生成与平台无关的中间代码 |
| 5 | **代码优化** | **Code Optimization** | 让代码运行更快、占用更少内存 |
| 6 | **目标代码生成** | **Target Code Generation** | 生成目标 **机器码 (machine code)** |
| 7 | **链接** | **Linking** | 把多个目标文件和库合并为可执行文件 |

### 7.1 关键术语 (Key Terms)

- **词法单元 (Token)** — 代码中最小的有意义单位，如 `if`、`x`、`=`、`10`
- **词法分析器 / 扫描器 (Lexer / Scanner / Tokenizer)** — 执行词法分析的工具
- **语法分析器 (Parser)** — 执行语法分析的工具
- **抽象语法树 (Abstract Syntax Tree, AST)** — 表示程序结构的树形数据结构
- **符号表 (Symbol Table)** — 记录所有变量、函数名及其属性的数据结构
- **语义错误 (Semantic Error)** — 语法正确但意义错误（如用字符串和数字相加）
- **语法错误 (Syntax Error)** — 不符合语言语法规则的错误（如缺少分号）

> **类比 (Analogy)：** 就像翻译一本小说 —— 先把句子分成词 (**lexing**)，再理解句子结构 (**parsing**)，再理解含义 (**semantic analysis**)，再润色 (**optimisation**)，最后写出目标语言 (**code generation**)。

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(14, 8))

stages = [
    ('Source Code\n(源代码)\nx = a + b * 2', '#BBDEFB', 0),
    ('Lexical Analysis\n(词法分析)\nTokens: x, =, a, +, b, *, 2', '#B3E5FC', 1.5),
    ('Syntax Analysis\n(语法分析)\nParse Tree / AST', '#B2EBF2', 3.0),
    ('Semantic Analysis\n(语义分析)\nType checking', '#B2DFDB', 4.5),
    ('Code Optimization\n(代码优化)\nSimplify expressions', '#C8E6C9', 6.0),
    ('Code Generation\n(代码生成)\nMOV, ADD, MUL...', '#DCEDC8', 7.5),
    ('Machine Code\n(机器码)\n10110000...', '#F0F4C3', 9.0),
]

box_h = 1.2
box_w = 3.5
center_x = 5

for label, color, y_pos in stages:
    y = 9.0 - y_pos
    rect = patches.FancyBboxPatch((center_x - box_w/2, y - box_h/2), box_w, box_h,
                                   boxstyle='round,pad=0.15',
                                   facecolor=color, edgecolor='black', lw=2)
    ax.add_patch(rect)
    ax.text(center_x, y, label, ha='center', va='center', fontsize=9, fontweight='bold')

# Arrows
for i in range(len(stages) - 1):
    y_from = 9.0 - stages[i][2] - box_h/2
    y_to = 9.0 - stages[i+1][2] + box_h/2
    ax.annotate('', xy=(center_x, y_to), xytext=(center_x, y_from),
                arrowprops=dict(arrowstyle='->', lw=2, color='#333333'))

# Error feedback arrow
ax.annotate('Error\nFeedback', xy=(center_x + box_w/2 + 0.3, 9.0 - 1.5),
            xytext=(center_x + box_w/2 + 0.3, 9.0 - 4.5),
            arrowprops=dict(arrowstyle='->', lw=1.5, color='red', linestyle='dashed'),
            fontsize=9, color='red', ha='center')

ax.set_xlim(0, 10)
ax.set_ylim(-1, 10.5)
ax.axis('off')
ax.set_title('Compilation Pipeline (编译流程)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

### 互动实验：词法分析器 (Lexer) 模拟

````python
import re

def simple_lexer(code_text):
    """A very simple lexical analyzer (tokenizer)."""
    token_spec = [
        ('NUMBER',   r'\d+(\.\d+)?'),
        ('KEYWORD',  r'\b(if|else|while|for|def|return|print)\b'),
        ('IDENT',    r'[a-zA-Z_]\w*'),
        ('OP',       r'[+\-*/=<>!]+'),
        ('LPAREN',   r'\('),
        ('RPAREN',   r'\)'),
        ('COLON',    r':'),
        ('COMMA',    r','),
        ('STRING',   r'"[^"]*"'),
        ('NEWLINE',  r'\n'),
        ('SKIP',     r'[ \t]+'),
    ]
    
    combined = '|'.join(f'(?P<{name}>{pattern})' for name, pattern in token_spec)
    tokens = []
    
    for match in re.finditer(combined, code_text):
        kind = match.lastgroup
        value = match.group()
        if kind in ('SKIP', 'NEWLINE'):
            continue
        tokens.append((kind, value))
    
    return tokens

# Test with sample code
test_code = 'if x > 10:\n    print("hello")'

print('=== Simple Lexical Analyzer (简易词法分析器) ===')
print()
print(f'Input code: {repr(test_code)}')
print()
print('Tokens found:')
print(f'{"Type":<12} {"Value":<15}')
print('-' * 27)

tokens = simple_lexer(test_code)
for token_type, token_value in tokens:
    print(f'{token_type:<12} {token_value:<15}')

print()
print(f'Total tokens: {len(tokens)}')
print()
print('This is the FIRST step of compilation!')
print('这是编译的第一步 - 将代码文本分解成有意义的词法单元！')
````

### 综合对比：三种翻译器一览

````python
import matplotlib.pyplot as plt
import numpy as np

# Create a comparison radar chart
categories = ['Translation\nSpeed', 'Execution\nSpeed', 'Error\nDetection',
              'Portability', 'Ease of\nDebugging', 'Code\nSecurity']
N = len(categories)

# Scores (1-5) for each translator
compiler_scores =    [2, 5, 3, 2, 2, 5]
interpreter_scores = [5, 2, 4, 4, 5, 1]
hybrid_scores =      [3, 4, 3, 5, 4, 3]

# Close the radar chart
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]

compiler_scores += compiler_scores[:1]
interpreter_scores += interpreter_scores[:1]
hybrid_scores += hybrid_scores[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

ax.fill(angles, compiler_scores, alpha=0.15, color='blue')
ax.plot(angles, compiler_scores, 'o-', linewidth=2, label='Compiler (编译器)', color='blue')

ax.fill(angles, interpreter_scores, alpha=0.15, color='red')
ax.plot(angles, interpreter_scores, 's-', linewidth=2, label='Interpreter (解释器)', color='red')

ax.fill(angles, hybrid_scores, alpha=0.15, color='green')
ax.plot(angles, hybrid_scores, '^-', linewidth=2, label='Hybrid (混合型)', color='green')

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=10)
ax.set_ylim(0, 5.5)
ax.set_yticks([1, 2, 3, 4, 5])
ax.set_yticklabels(['1', '2', '3', '4', '5'], fontsize=8)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1), fontsize=11)
ax.set_title('Translator Comparison (翻译器对比)', fontsize=14, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()
````
