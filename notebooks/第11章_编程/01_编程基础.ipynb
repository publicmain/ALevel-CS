## 1. 什么是编程？ (What is Programming?)

### 从算法到可执行代码

编程就是将**算法 (algorithm)** 翻译成计算机能够理解和执行的**指令 (instructions)**。

想象一下：
- **算法**就像一份菜谱 —— 它告诉你做一道菜的步骤
- **编程**就是把这份菜谱翻译成厨房机器人能理解的指令
- **程序 (Program)**就是翻译后的、机器人能执行的指令集

### 从流程图/结构化英语到代码

在第9章中，我们学习了用**流程图 (flowchart)** 和**结构化英语 (structured English)** 来描述算法。
现在，我们要把它们变成真正可运行的代码！

```
算法设计（思考）  →  伪代码/流程图（描述）  →  编程语言代码（实现）  →  可执行程序（运行）
```

**为什么要经历这些步骤？**
- 直接写代码容易出错——先设计算法可以让我们专注于**逻辑**，而不被语法细节干扰
- 伪代码是**语言无关的** (language-independent)——同一个算法可以用 Python、Java、C++ 等任何语言实现

> **CS 拓展知识**：世界上有数百种编程语言，但它们的核心概念是相通的。学好基础概念比记住某种语言的语法更重要！

## 2. 常量 (Constants)

### 什么是常量？

**常量 (Constant)** 是在程序运行期间**值不会改变**的数据项。

**现实世界的例子：**
- 圆周率 $\pi \approx 3.14159$——这是一个数学常数，永远不会改变
- 增值税税率 (VAT Rate)——在一段时间内保持不变
- 重力加速度 $g \approx 9.81 m/s^2$

### 为什么使用常量？（三个核心原因）

| 原因 | 解释 | 例子 |
|------|------|------|
| **可读性 (Readability)** | 给"魔术数字"一个有意义的名字 | `TAX_RATE` 比 `0.2` 更容易理解 |
| **可维护性 (Maintainability)** | 修改一处，全局生效 | 税率变了？只改一行代码！ |
| **避免错误** | 防止意外修改 | 编译器/解释器会阻止你修改常量 |

### 什么是"魔术数字" (Magic Numbers)？

"魔术数字"是指代码中直接出现的、含义不明的数值。它们就像魔术一样——只有写代码的人知道它们的含义！

```python
# 坏代码 ❌ —— 0.2 是什么意思？
total = price * 0.2

# 好代码 ✅ —— 一目了然！
TAX_RATE = 0.2
total = price * TAX_RATE
```

### 伪代码声明
```
CONSTANT PI = 3.14159
CONSTANT TAX_RATE = 0.2
CONSTANT MAX_STUDENTS = 30
```

````python
# Python 中的常量
# Python 没有真正的常量关键字，但我们用 ALL_CAPS（全大写）命名来表示"这是常量，请不要修改"

PI = 3.14159
TAX_RATE = 0.2
MAX_STUDENTS = 30
GRAVITY = 9.81

# 使用常量计算
radius = 5
area = PI * radius ** 2
print(f"半径为 {radius} 的圆的面积 = {area:.2f}")

price = 100
tax = price * TAX_RATE
print(f"价格 {price} 元的商品，税额 = {tax} 元")
print(f"含税总价 = {price + tax} 元")
````

## 3. 变量 (Variables)

### 什么是变量？

**变量 (Variable)** 是内存中的一个**命名存储位置 (named storage location)**，其值在程序运行期间可以改变。

**形象类比：**
想象变量就是一个**带标签的盒子**：
- 盒子上的**标签** = 变量名 (variable name)
- 盒子里的**内容** = 变量的值 (value)
- 你可以随时打开盒子，**放入新内容**（赋值）或者**查看内容**（读取）

### 底层原理 —— 内存是什么？

计算机的内存 (RAM) 就像一排排小格子，每个格子都有一个**地址 (address)**。
变量名是地址的"别名"——编程语言帮我们把难记的地址（如 `0x7FFE42A0`）变成好记的名字（如 `score`）。

```
内存地址      变量名     值
0x7FFE42A0    score      95
0x7FFE42A4    name       "Alice"
0x7FFE42A8    passed     TRUE
```

### 声明与初始化

- **声明 (Declaration)**：告诉计算机"我要用一个叫 X 的变量"
- **初始化 (Initialization)**：给变量第一次赋值

**伪代码：**
```
DECLARE Score : INTEGER
Score ← 0        // 初始化
```

### 命名规范 (Naming Conventions)

| 规则 | 好的例子 | 坏的例子 |
|------|----------|----------|
| 使用有意义的名字 | `studentAge` | `x`, `a1` |
| camelCase（驼峰命名） | `firstName`, `totalScore` | `firstname` |
| snake_case（蛇形命名）| `first_name`, `total_score` | `FIRSTNAME` |
| 不以数字开头 | `score1` | `1score` |
| 不用保留字 | `my_input` | `input`, `print` |

> **Python 惯例**：变量用 `snake_case`，常量用 `ALL_CAPS`，类名用 `PascalCase`。

````python
# Python 中的变量：创建、赋值、重新赋值

# 创建并初始化变量
student_name = "小明"
student_age = 16
is_enrolled = True

print("=== 学生信息 ===")
print(f"姓名: {student_name}")
print(f"年龄: {student_age}")
print(f"已注册: {is_enrolled}")

# 重新赋值 —— 变量的值可以改变！
student_age = 17  # 生日了！
print(f"\n一年后，年龄变为: {student_age}")

# 变量的类型也可以改变（Python 是动态类型语言）
x = 10
print(f"\nx = {x}，类型: {type(x).__name__}")
x = "hello"
print(f"x = {x}，类型: {type(x).__name__}")
print("\n注意：虽然 Python 允许这样做，但在实际编程中不建议随意改变变量类型！")
````

### 作用域 (Scope) —— 简要介绍

变量的**作用域 (scope)** 决定了它在程序中**哪些地方可以被访问**。

- **局部变量 (Local Variable)**：在函数/过程内部定义，只能在该函数内使用
- **全局变量 (Global Variable)**：在函数外部定义，整个程序都能使用

我们将在第3节"结构化编程"中详细讨论作用域。现在只需要知道：

> **好的编程实践**：尽量使用局部变量，减少全局变量的使用。这就像"信息保密原则"——只有需要知道的人才应该知道。

## 4. 赋值 (Assignment)

### 赋值操作符

| 语言 | 赋值操作符 | 例子 |
|------|-----------|------|
| 伪代码 | `←` | `Score ← 100` |
| Python | `=` | `score = 100` |

### 重要提醒：`=` 不是"等于"！

在编程中，`=` 是**赋值操作符 (assignment operator)**，不是数学中的"等号"！

- 数学：$x = x + 1$ 是**不成立**的（没有数满足这个等式）
- 编程：`x = x + 1` 是**完全合法**的——意思是"取 x 的当前值，加 1，再存回 x"

**形象理解：**
把 `=` 读作"**变成**"或"**被赋值为**"，而不是"等于"：
- `x = 5` → "x **变成** 5"
- `x = x + 1` → "x **变成** x 的当前值加 1"

> **CS 拓展知识**：在 Python 中，`==` 才是"等于"比较运算符！新手最常见的 bug 之一就是把 `==` 写成 `=`。

````python
# 赋值操作演示

# 基本赋值
x = 10
print(f"x = {x}")

# x = x + 1 的含义
x = x + 1  # 等价于 x += 1
print(f"x = x + 1 之后，x = {x}")

# 多重赋值 (Multiple Assignment)
a, b, c = 1, 2, 3
print(f"\n多重赋值: a={a}, b={b}, c={c}")

# 交换两个变量的值 (Swap)
print(f"\n交换前: a={a}, b={b}")

# 方法1：使用临时变量（伪代码方式）
temp = a
a = b
b = temp
print(f"交换后（临时变量法）: a={a}, b={b}")

# 方法2：Python 的优雅写法
a, b = b, a
print(f"再次交换（Python 方式）: a={a}, b={b}")

# 复合赋值运算符
score = 100
score += 10   # score = score + 10
print(f"\nscore += 10 → {score}")
score -= 5    # score = score - 5
print(f"score -= 5  → {score}")
score *= 2    # score = score * 2
print(f"score *= 2  → {score}")
````

## 5. 表达式与运算符 (Expressions & Operators)

### 5.1 算术运算符 (Arithmetic Operators)

| 运算 | 伪代码 | Python | 例子 | 结果 |
|------|--------|--------|------|------|
| 加法 | `+` | `+` | `5 + 3` | `8` |
| 减法 | `-` | `-` | `10 - 4` | `6` |
| 乘法 | `*` | `*` | `3 * 7` | `21` |
| 除法 | `/` | `/` | `15 / 4` | `3.75` |
| 整数除法 | `DIV` | `//` | `15 // 4` | `3` |
| 取余 | `MOD` | `%` | `15 % 4` | `3` |
| 幂运算 | `^` | `**` | `2 ** 3` | `8` |

### 为什么 MOD 和 DIV 如此重要？

这两个运算符在编程中**用途极广**：

1. **判断奇偶**：`n % 2 == 0` → 偶数
2. **提取数字的各个位**：`1234 % 10 = 4`（个位），`1234 // 10 = 123`（去掉个位）
3. **时间计算**：`150 分钟 = 150 // 60` 小时 `150 % 60` 分钟 = 2小时30分
4. **循环计数**：每 N 次做一件事——`count % N == 0`
5. **哈希表**：用 MOD 来决定数据存放位置

### 运算符优先级 (Operator Precedence)

和数学一样，编程也有运算优先级——**BODMAS/PEMDAS** 规则：

```
1. Brackets   括号     ()
2. Orders     幂       **
3. Division   除法     / // %
   Multiplication 乘法 *
4. Addition   加法     +
   Subtraction 减法    -
```

> **提示**：不确定优先级时，**加括号**！这既能保证正确性，又能提高可读性。

````python
# 算术运算符演示
print("=== 基本算术 ===")
print(f"5 + 3  = {5 + 3}")
print(f"10 - 4 = {10 - 4}")
print(f"3 * 7  = {3 * 7}")
print(f"15 / 4 = {15 / 4}")     # 普通除法 → 浮点数
print(f"15 // 4 = {15 // 4}")   # 整数除法 DIV
print(f"15 % 4  = {15 % 4}")    # 取余 MOD
print(f"2 ** 3  = {2 ** 3}")    # 幂运算

print("\n=== MOD 和 DIV 的实用例子 ===")

# 判断奇偶
number = 17
if number % 2 == 0:
    print(f"{number} 是偶数")
else:
    print(f"{number} 是奇数")

# 提取数字的各个位
num = 1234
ones = num % 10          # 个位
tens = (num // 10) % 10  # 十位
hundreds = (num // 100) % 10  # 百位
print(f"\n{num} 的个位={ones}，十位={tens}，百位={hundreds}")

# 时间转换
total_minutes = 150
hours = total_minutes // 60
minutes = total_minutes % 60
print(f"\n{total_minutes} 分钟 = {hours} 小时 {minutes} 分钟")

# 运算优先级
print("\n=== 运算优先级 ===")
print(f"2 + 3 * 4   = {2 + 3 * 4}")      # 14, 不是 20
print(f"(2 + 3) * 4 = {(2 + 3) * 4}")    # 20
print(f"2 ** 3 ** 2  = {2 ** 3 ** 2}")    # 512, 幂运算右结合
````

### 5.2 比较运算符 (Comparison Operators)

| 含义 | 伪代码 | Python | 例子 |
|------|--------|--------|------|
| 等于 | `=` | `==` | `x == 5` |
| 不等于 | `<>` | `!=` | `x != 5` |

### 5.3 逻辑运算符 (Logical Operators)

| 运算 | 伪代码 | Python | 含义 |
|------|--------|--------|------|
| 与 | `AND` | `and` | 两个条件**都**为真 |
| 或 | `OR` | `or` | **至少一个**条件为真 |
| 非 | `NOT` | `not` | **取反** |

### 真值表 (Truth Tables)

**AND（与）—— 两者必须都为 True：**

| A | B | A AND B |
|---|---|--------|
| True | True | **True** |
| True | False | False |
| False | True | False |
| False | False | False |

**OR（或）—— 至少一个为 True：**

| A | B | A OR B |
|---|---|-------|
| True | True | True |
| True | False | True |
| False | True | True |
| False | False | **False** |

**NOT（非）—— 取反：**

| A | NOT A |
|---|------|
| True | False |
| False | True |

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 生成真值表可视化
operators = ['AND', 'OR', 'NOT']
truth_tables = [
    # AND
    [[True, True, True], [True, False, False], [False, True, False], [False, False, False]],
    # OR
    [[True, True, True], [True, False, True], [False, True, True], [False, False, False]],
    # NOT (only A column matters)
    [[True, None, False], [False, None, True]]
]

for idx, (ax, op, table) in enumerate(zip(axes, operators, truth_tables)):
    ax.set_xlim(0, 4 if idx < 2 else 3)
    ax.set_ylim(0, len(table) + 1.5)
    ax.axis('off')
    ax.set_title(f'{op} 运算', fontsize=14, fontweight='bold', pad=10)
    
    if idx < 2:  # AND, OR
        headers = ['A', 'B', f'A {op} B']
        col_x = [0.5, 1.5, 3.0]
    else:  # NOT
        headers = ['A', 'NOT A']
        col_x = [0.5, 2.0]
    
    y_top = len(table) + 0.8
    for i, h in enumerate(headers):
        ax.text(col_x[i], y_top, h, ha='center', va='center', fontsize=11, fontweight='bold')
    
    for row_idx, row in enumerate(table):
        y = len(table) - row_idx - 0.2
        vals = [row[0], row[1], row[2]] if idx < 2 else [row[0], row[2]]
        for col_idx, val in enumerate(vals):
            color = '#2ecc71' if val else '#e74c3c'
            label = 'T' if val else 'F'
            ax.text(col_x[col_idx], y, label, ha='center', va='center',
                    fontsize=12, fontweight='bold', color=color,
                    bbox=dict(boxstyle='round,pad=0.3', facecolor=color, alpha=0.15))

plt.tight_layout()
plt.savefig('truth_tables.png', dpi=100, bbox_inches='tight')
plt.show()
print("真值表可视化完成！")
````

### 5.4 字符串操作 (String Operations)

字符串是编程中最常用的数据类型之一。以下是常见操作：

| 操作 | 伪代码 | Python | 说明 |
|------|--------|--------|------|
| 连接 | `&` | `+` | 将两个字符串拼接在一起 |
| 长度 | `LENGTH(s)` | `len(s)` | 返回字符串长度 |
| 子串 | `SUBSTRING(s,start,len)` | `s[start:end]` | 提取一部分 |
| 大写 | `UCASE(s)` | `s.upper()` | 转为大写 |
| 小写 | `LCASE(s)` | `s.lower()` | 转为小写 |

````python
# 字符串操作演示
first_name = "John"
last_name = "Smith"

# 连接 (Concatenation)
full_name = first_name + " " + last_name
print(f"连接: {full_name}")

# 长度 (Length)
print(f"长度: {len(full_name)} 个字符")

# 子串 / 切片 (Substring / Slicing)
text = "Hello, World!"
print(f"前5个字符: {text[:5]}")
print(f"第7到11个字符: {text[7:12]}")

# 大小写
print(f"大写: {text.upper()}")
print(f"小写: {text.lower()}")

# 查找 (Find)
pos = text.find("World")
print(f"'World' 在位置: {pos}")

# 替换 (Replace)
new_text = text.replace("World", "Python")
print(f"替换后: {new_text}")
````

## 6. 输入与输出 (Input & Output)

### INPUT — 从键盘获取输入

| 伪代码 | Python |
|--------|--------|
| `INPUT Name` | `name = input()` |
| `OUTPUT "请输入名字"` 然后 `INPUT Name` | `name = input("请输入名字: ")` |

### 重要陷阱：`input()` 返回的是字符串！

这是新手最常遇到的问题之一：

```python
age = input("你的年龄: ")  # age 是字符串 "16"，不是数字 16！
age + 1  # 报错！不能字符串 + 数字
```

**解决方法：类型转换 (Type Conversion)**
```python
age = int(input("你的年龄: "))    # 转为整数
price = float(input("价格: "))    # 转为浮点数
```

### OUTPUT — 输出到屏幕

| 伪代码 | Python |
|--------|--------|
| `OUTPUT "Hello"` | `print("Hello")` |
| `OUTPUT "Score: ", Score` | `print("Score:", score)` |

### f-string 格式化输出

Python 的 f-string 是格式化输出的最优雅方式：

```python
name = "小明"
score = 95.678
print(f"姓名: {name}")         # 插入变量
print(f"分数: {score:.1f}")    # 保留1位小数 → 95.7
print(f"{'居中':^20}")         # 居中对齐，宽度20
```

````python
# 输入输出演示（这里使用预设值代替 input，方便运行）

# 模拟用户输入
name = "小明"      # 等效于 name = input("请输入名字: ")
age_str = "16"    # 等效于 age_str = input("请输入年龄: ")
score_str = "95.5" # 等效于 score_str = input("请输入分数: ")

# 类型转换
age = int(age_str)
score = float(score_str)

print(f"姓名: {name}")
print(f"年龄: {age}")
print(f"分数: {score}")

# 演示为什么类型转换很重要
print(f"\n没有转换: '16' + '1' = {'16' + '1'}")   # 字符串连接！
print(f"转换之后: 16 + 1 = {16 + 1}")              # 数学加法！

# f-string 格式化
print("\n=== 格式化输出 ===")
pi = 3.14159265
print(f"pi = {pi}")            # 默认
print(f"pi = {pi:.2f}")        # 2位小数
print(f"pi = {pi:.4f}")        # 4位小数

price = 42.5
print(f"\n价格: {price:>10.2f} 元")   # 右对齐，宽度10
print(f"价格: {price:<10.2f} 元")   # 左对齐
print(f"价格: {price:^10.2f} 元")   # 居中
````

## 7. 内置函数与库例程 (Built-in Functions & Library Routines)

### 什么是内置函数？

**内置函数 (Built-in Functions)** 是编程语言自带的、**预先编写好的代码块**。
你不需要知道它们内部是怎么工作的，只需要知道：
- **输入什么** (参数)
- **输出什么** (返回值)

**类比：** 内置函数就像家用电器——你不需要知道微波炉内部的电路原理，只需要知道：放入食物 → 按按钮 → 热食物出来。

### 为什么使用库？(Don't Reinvent the Wheel!)

"不要重新发明轮子"是编程的重要原则：
- 库函数经过**大量测试**，比你自己写的更可靠
- 节省**开发时间**
- 代码更**简洁**、更**易读**

### 常用内置函数

| 函数 | 用途 | 例子 |
|------|------|------|
| `len()` | 长度 | `len("hello")` → `5` |
| `int()` | 转为整数 | `int("42")` → `42` |
| `float()` | 转为浮点数 | `float("3.14")` → `3.14` |
| `str()` | 转为字符串 | `str(42)` → `"42"` |
| `round()` | 四舍五入 | `round(3.7)` → `4` |
| `abs()` | 绝对值 | `abs(-5)` → `5` |
| `max()` | 最大值 | `max(3, 7, 1)` → `7` |
| `min()` | 最小值 | `min(3, 7, 1)` → `1` |
| `type()` | 查看类型 | `type(42)` → `<class 'int'>` |

````python
# 内置函数探索
print("=== 类型转换函数 ===")
print(f"int('42')    = {int('42')}")
print(f"float('3.14')= {float('3.14')}")
print(f"str(42)      = '{str(42)}'")
print(f"int(3.9)     = {int(3.9)}")     # 截断，不是四舍五入！
print(f"round(3.9)   = {round(3.9)}")   # 四舍五入

print("\n=== 数学函数 ===")
print(f"abs(-7)      = {abs(-7)}")
print(f"round(3.14159, 2) = {round(3.14159, 2)}")
print(f"max(3, 7, 1) = {max(3, 7, 1)}")
print(f"min(3, 7, 1) = {min(3, 7, 1)}")
print(f"pow(2, 10)   = {pow(2, 10)}")

print("\n=== math 库 ===")
import math
print(f"math.sqrt(16)   = {math.sqrt(16)}")
print(f"math.pi         = {math.pi}")
print(f"math.floor(3.7) = {math.floor(3.7)}")
print(f"math.ceil(3.2)  = {math.ceil(3.2)}")

print("\n=== random 库 ===")
import random
print(f"random.randint(1, 6)  = {random.randint(1, 6)}")   # 模拟骰子
print(f"random.random()       = {random.random():.4f}")    # 0~1之间的随机数
print(f"random.choice(['a','b','c']) = {random.choice(['a','b','c'])}")

print("\n=== 字符串方法 ===")
text = "Hello, World!"
print(f"'{text}'.upper()   = '{text.upper()}'")
print(f"'{text}'.lower()   = '{text.lower()}'")
print(f"'{text}'.find('World') = {text.find('World')}")
print(f"'{text}'.replace('World','Python') = '{text.replace('World','Python')}'")
print(f"'{text}'.split(',') = {text.split(',')}")
print(f"' hello '.strip()  = '{'  hello  '.strip()}'")
````

## 8. 综合示例：从伪代码到 Python

### 示例1：计算圆的面积和周长

**伪代码：**
```
CONSTANT PI = 3.14159
OUTPUT "请输入圆的半径："
INPUT Radius
Area ← PI * Radius * Radius
Circumference ← 2 * PI * Radius
OUTPUT "面积 = ", Area
OUTPUT "周长 = ", Circumference
```

````python
# 示例1: 计算圆的面积和周长
PI = 3.14159

radius = 5  # 模拟输入
area = PI * radius * radius
circumference = 2 * PI * radius

print(f"半径 = {radius}")
print(f"面积 = {area:.2f}")
print(f"周长 = {circumference:.2f}")
````

### 示例2：温度转换器

**伪代码：**
```
OUTPUT "请输入摄氏温度："
INPUT Celsius
Fahrenheit ← Celsius * 9 / 5 + 32
OUTPUT Celsius, "°C = ", Fahrenheit, "°F"
```

````python
# 示例2: 温度转换器
celsius = 37  # 模拟输入：人体体温
fahrenheit = celsius * 9 / 5 + 32
print(f"{celsius}°C = {fahrenheit:.1f}°F")

# 批量转换
print("\n=== 温度对照表 ===")
print(f"{'摄氏(°C)':>10} | {'华氏(°F)':>10}")
print("-" * 25)
for c in range(0, 101, 10):
    f = c * 9 / 5 + 32
    print(f"{c:>10} | {f:>10.1f}")
````

### 练习4：综合应用
编写程序计算 BMI（身体质量指数）：
- 公式：BMI = 体重(kg) / 身高(m)^2
- 输出 BMI 值（保留1位小数）

试着先写伪代码，再翻译成 Python！

````python
# 练习2参考答案：秒数转换
total_seconds = 3661  # 模拟输入

hours = total_seconds // 3600
remaining = total_seconds % 3600
minutes = remaining // 60
seconds = remaining % 60

print(f"{total_seconds} 秒 = {hours} 小时 {minutes} 分钟 {seconds} 秒")
print(f"格式化: {hours:02d}:{minutes:02d}:{seconds:02d}")
````

````python
# 练习4参考答案：BMI 计算器
# 伪代码：
# INPUT Weight
# INPUT Height
# BMI <- Weight / (Height * Height)
# OUTPUT "BMI = ", BMI

weight = 70    # kg
height = 1.75  # m

bmi = weight / (height ** 2)
print(f"体重: {weight} kg")
print(f"身高: {height} m")
print(f"BMI = {bmi:.1f}")

# 额外：判断 BMI 类别
if bmi < 18.5:
    category = "偏瘦"
elif bmi < 24:
    category = "正常"
elif bmi < 28:
    category = "偏胖"
else:
    category = "肥胖"
print(f"类别: {category}")
````
