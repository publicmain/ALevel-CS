## 1. 什么是结构化编程？

### 意大利面条代码 vs 结构化代码

在编程的早期（1960年代），程序员使用 `GOTO` 语句随意跳转代码，导致程序逻辑混乱——
就像一盘**意大利面条 (Spaghetti Code)**，代码的执行路径缠绕在一起，几乎无法理解。

**结构化编程** 的出现改变了这一切。它的核心原则：

| 原则 | 解释 |
|------|------|
| **模块化 (Modularity)** | 将大问题分解为小模块，每个模块完成一个具体任务 |
| **单一入口/出口** | 每个模块只有一个入口和一个出口 |
| **自顶向下设计 (Top-Down Design)** | 从整体到细节，逐步细化 |
| **只用三种控制结构** | 顺序、选择、迭代 |

### 为什么结构化编程重要？

1. **可读性 (Readability)**：别人（包括未来的你）能看懂代码
2. **可维护性 (Maintainability)**：容易修改和更新
3. **可测试性 (Testability)**：每个模块可以独立测试
4. **可复用性 (Reusability)**：写一次，到处使用
5. **团队协作**：不同人可以同时编写不同模块

> **CS 拓展知识**：1968年，Edsger Dijkstra 发表了著名的文章 "Go To Statement Considered Harmful"，
> 推动了结构化编程的革命。这是计算机科学历史上最有影响力的文章之一！

## 2. 过程 (Procedures)

### 什么是过程？

**过程 (Procedure)** 是一个**命名的代码块**，执行特定的任务，但**不返回值**。

**类比：** 过程就像一个"做事的人"——你叫他做某件事，他就去做，但不会给你带什么东西回来。
比如你叫某人"去打扫房间"——他打扫了，但没有"返回"什么给你。

### 为什么使用过程？

1. **代码复用 (Reuse)**：写一次，调用多次
2. **可读性**：用有意义的名字描述一段代码做什么
3. **可维护性**：修改只需改一处
4. **模块化**：大问题 → 小任务

### 伪代码语法

```
PROCEDURE ProcedureName(parameter1 : DataType, parameter2 : DataType)
    statement(s)
ENDPROCEDURE

// 调用过程
CALL ProcedureName(argument1, argument2)
```

### Python 语法

```python
def procedure_name(parameter1, parameter2):
    statement(s)
    # 注意：过程没有 return 语句（或者 return 后面不跟值）

# 调用过程
procedure_name(argument1, argument2)
```

````python
# 过程示例

# 过程1：无参数的过程
def print_separator():
    """打印分隔线（无参数）"""
    print("=" * 40)

# 过程2：带参数的过程
def greet(name):
    """打招呼（有参数，无返回值）"""
    print(f"你好，{name}！欢迎学习编程！")

# 过程3：多参数
def print_student_info(name, age, grade):
    """打印学生信息"""
    print(f"  姓名: {name}")
    print(f"  年龄: {age}")
    print(f"  年级: {grade}")

# 调用过程
print_separator()
greet("小明")
print_separator()
greet("小红")
print_separator()

print("\n学生列表：")
students = [("小明", 16, "AS"), ("小红", 17, "A2"), ("小刚", 15, "IGCSE")]
for name, age, grade in students:
    print_student_info(name, age, grade)
    print()
````

## 3. 函数 (Functions)

### 什么是函数？

**函数 (Function)** 和过程类似，但**会返回一个值**。

**类比：** 函数就像一台"加工机器"——你放入原料（参数），它加工后吐出产品（返回值）。
比如一台榨汁机：放入水果 → 得到果汁。

### 关键区别：过程 vs 函数

| 特性 | 过程 (Procedure) | 函数 (Function) |
|------|-----------------|----------------|
| 返回值 | **没有** | **有** |
| 目的 | 执行一个动作 | 计算并返回结果 |
| 调用方式 | `CALL procedure()` | `result ← function()` |
| 可以赋值？ | 不可以 | 可以！ |
| 例子 | `print_line()` | `calculate_area()` |

### 伪代码语法

```
FUNCTION FunctionName(parameter : DataType) RETURNS DataType
    statement(s)
    RETURN value
ENDFUNCTION

// 调用
Result ← FunctionName(argument)
```

### Python 语法

```python
def function_name(parameter):
    statement(s)
    return value

# 调用
result = function_name(argument)
```

````python
# 函数示例

# 函数1：计算圆的面积
def calculate_area(radius):
    """计算圆的面积并返回结果"""
    PI = 3.14159
    area = PI * radius ** 2
    return area

# 函数2：判断是否为闰年
def is_leap_year(year):
    """判断是否为闰年，返回布尔值"""
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False

# 函数3：计算阶乘
def factorial(n):
    """计算 n 的阶乘"""
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# 调用函数并使用返回值
r = 5
area = calculate_area(r)
print(f"半径 {r} 的圆面积 = {area:.2f}")

print(f"\n2024 是闰年? {is_leap_year(2024)}")
print(f"2023 是闰年? {is_leap_year(2023)}")

for n in range(1, 8):
    print(f"{n}! = {factorial(n)}")
````

````python
# 过程 vs 函数 —— 直观对比

# 过程：只执行动作，不返回值
def display_result(value):
    """这是一个过程：只打印，不返回"""
    print(f"  结果是: {value}")
    # 没有 return

# 函数：计算并返回值
def square(x):
    """这是一个函数：计算并返回平方"""
    return x * x

print("=== 过程 vs 函数 ===")

# 过程的返回值是 None
proc_result = display_result(42)
print(f"  过程的返回值: {proc_result}")  # None

# 函数返回有意义的值
func_result = square(7)
print(f"  函数的返回值: {func_result}")  # 49

# 函数的返回值可以用在表达式中
total = square(3) + square(4)
print(f"\n  3^2 + 4^2 = {total}")  # 勾股定理！
print(f"  sqrt({total}) = {total**0.5}")  # 5.0
````

## 4. 参数与实参 (Parameters & Arguments)

### 术语区分

这两个概念经常被混淆，但它们是不同的：

| 术语 | 英文 | 含义 | 位置 |
|------|------|------|------|
| **形参** | Parameter | 函数定义中的**占位符** | 函数声明处 |
| **实参** | Argument | 调用函数时传入的**实际值** | 函数调用处 |

```python
def greet(name):     # name 是形参 (parameter)
    print(f"Hi {name}")

greet("Alice")       # "Alice" 是实参 (argument)
```

**记忆技巧：**
- **P**arameter = **P**laceholder（占位符）
- **A**rgument = **A**ctual（实际值）

## 5. 值传递 vs 引用传递

### 值传递 (Pass by Value)

传递参数的**副本**。函数内部对参数的修改**不会影响**原始变量。

**类比：** 给朋友一份文件的**复印件**。朋友在复印件上怎么涂改，你的原件不受影响。

### 引用传递 (Pass by Reference)

传递参数的**引用**（内存地址）。函数内部的修改**会影响**原始变量。

**类比：** 给朋友你的**原件**。朋友在上面写字，你拿回来时文件已经被改了。

### 伪代码表示

```
// 值传递（默认）
PROCEDURE double(BYVALUE x : INTEGER)
    x ← x * 2
ENDPROCEDURE

// 引用传递
PROCEDURE double(BYREF x : INTEGER)
    x ← x * 2
ENDPROCEDURE
```

### Python 的特殊情况

Python 使用的是 **"传递对象引用" (Pass by Object Reference)**，行为取决于对象是否可变：

| 类型 | 可变性 | 行为 | 类似于 |
|------|--------|------|--------|
| `int`, `float`, `str`, `tuple` | 不可变 (Immutable) | 类似值传递 | BYVALUE |
| `list`, `dict`, `set` | 可变 (Mutable) | 类似引用传递 | BYREF |

````python
# 值传递 vs 引用传递 演示

print("=== 不可变类型（类似值传递）===")
def try_to_modify_int(x):
    print(f"  函数内修改前: x = {x}")
    x = x * 2  # 创建了新对象，原始变量不受影响
    print(f"  函数内修改后: x = {x}")

num = 10
print(f"调用前: num = {num}")
try_to_modify_int(num)
print(f"调用后: num = {num}")  # 仍然是 10！

print("\n=== 可变类型（类似引用传递）===")
def modify_list(lst):
    print(f"  函数内修改前: lst = {lst}")
    lst.append(99)  # 修改原始列表！
    print(f"  函数内修改后: lst = {lst}")

my_list = [1, 2, 3]
print(f"调用前: my_list = {my_list}")
modify_list(my_list)
print(f"调用后: my_list = {my_list}")  # 变成了 [1, 2, 3, 99]！
````

## 6. 作用域 (Scope)：局部变量 vs 全局变量

### 局部变量 (Local Variable)

- 在函数/过程**内部**定义
- 只能在该函数/过程内部使用
- 函数执行完毕后，局部变量就被销毁
- **优点**：不会意外影响其他代码

### 全局变量 (Global Variable)

- 在函数/过程**外部**（最顶层）定义
- 整个程序都能访问
- **缺点**：任何地方都能修改，容易造成 bug

### 为什么优先使用局部变量？

| 局部变量 | 全局变量 |
|----------|----------|
| 安全，不会被意外修改 | 任何代码都能修改，难以追踪 |
| 不同函数可以用同名变量 | 命名冲突风险 |
| 函数独立，易于测试 | 函数依赖外部状态 |
| 内存自动释放 | 占用内存直到程序结束 |

> **黄金法则**：尽量使用局部变量 + 参数传递。只在真正需要时才用全局变量。

````python
# 作用域演示

# 全局变量
global_var = "我是全局变量"
counter = 0

def demonstrate_scope():
    # 局部变量
    local_var = "我是局部变量"
    print(f"  函数内部 - local_var: {local_var}")
    print(f"  函数内部 - global_var: {global_var}")  # 可以读取全局变量

demonstrate_scope()

print(f"\n函数外部 - global_var: {global_var}")
try:
    print(f"函数外部 - local_var: {local_var}")
except NameError as e:
    print(f"函数外部 - local_var: 错误！{e}")

print("\n=== 同名变量不同作用域 ===")
x = 100  # 全局

def func_a():
    x = 200  # 局部（不影响全局的 x）
    print(f"  func_a 内部: x = {x}")

def func_b():
    x = 300  # 另一个局部的 x
    print(f"  func_b 内部: x = {x}")

func_a()
func_b()
print(f"  全局: x = {x}")  # 仍然是 100
````

## 7. 术语总结 (Terminology)

| 术语 | 英文 | 含义 |
|------|------|------|
| 过程头 | Procedure Header | 过程的第一行（名字 + 参数列表） |
| 函数头 | Function Header | 函数的第一行（名字 + 参数 + 返回类型） |
| 接口 | Interface | 函数/过程与外界交互的方式（参数 + 返回值） |
| 参数列表 | Parameter List | 函数/过程接受的所有参数 |
| 返回值 | Return Value | 函数计算后返回的结果 |
| 调用 | Call | 执行一个过程/函数 |
| 形参 | Parameter | 函数定义中的占位符 |
| 实参 | Argument | 函数调用时的实际值 |
| 值传递 | Pass by Value | 传递参数的副本 |
| 引用传递 | Pass by Reference | 传递参数的引用 |
| 局部变量 | Local Variable | 函数内部定义，只在函数内有效 |
| 全局变量 | Global Variable | 函数外定义，全程序有效 |
| 作用域 | Scope | 变量可以被访问的范围 |

## 8. 编写高效代码的原则

### DRY 原则 (Don't Repeat Yourself)

如果你发现自己在**复制粘贴**代码，那就应该把它提取成一个函数/过程！

```python
# DRY 之前（重复代码）
print("学生1: " + name1 + ", 分数: " + str(score1))
print("学生2: " + name2 + ", 分数: " + str(score2))
print("学生3: " + name3 + ", 分数: " + str(score3))

# DRY 之后（使用函数）
def print_student(name, score):
    print(f"学生: {name}, 分数: {score}")
```

### 单一职责原则 (Single Responsibility Principle)

每个函数/过程只做**一件事**！

```python
# 不好：一个函数做太多事
def process_student(name, scores):
    # 计算平均分 + 判断等级 + 打印结果 + 保存到文件...

# 好：每个函数只做一件事
def calculate_average(scores): ...
def determine_grade(average): ...
def display_result(name, grade): ...
```

### 有意义的命名

- 函数名应该是**动词**或**动词短语**：`calculate_tax()`, `validate_input()`, `get_average()`
- 变量名应该是**名词**：`student_name`, `total_score`, `max_attempts`
- 布尔变量用 `is_`/`has_` 开头：`is_valid`, `has_permission`

## 9. 综合实例

### 实例1：模块化计算器程序

````python
# 实例1：模块化计算器

def add(a, b):
    """加法"""
    return a + b

def subtract(a, b):
    """减法"""
    return a - b

def multiply(a, b):
    """乘法"""
    return a * b

def divide(a, b):
    """除法（带错误处理）"""
    if b == 0:
        return "错误：除数不能为零！"
    return a / b

def display_menu():
    """显示菜单（过程）"""
    print("\n" + "=" * 25)
    print("    简易计算器")
    print("=" * 25)
    print("1. 加法 (+)")
    print("2. 减法 (-)")
    print("3. 乘法 (*)")
    print("4. 除法 (/)")
    print("5. 退出")

def calculate(choice, num1, num2):
    """根据选择执行运算（函数）"""
    if choice == 1:
        return add(num1, num2)
    elif choice == 2:
        return subtract(num1, num2)
    elif choice == 3:
        return multiply(num1, num2)
    elif choice == 4:
        return divide(num1, num2)
    else:
        return "无效选择"

# 模拟运行
operations = [(1, 15, 8), (3, 7, 6), (4, 20, 4), (4, 10, 0)]
op_symbols = {1: '+', 2: '-', 3: '*', 4: '/'}

display_menu()
for choice, a, b in operations:
    result = calculate(choice, a, b)
    symbol = op_symbols.get(choice, '?')
    print(f"\n  {a} {symbol} {b} = {result}")
````

### 实例2：学生成绩管理系统

````python
# 实例2：学生成绩管理系统

def calculate_average(scores):
    """计算平均分（函数）"""
    if len(scores) == 0:
        return 0
    return sum(scores) / len(scores)

def determine_grade(average):
    """根据平均分判断等级（函数）"""
    if average >= 90:
        return 'A*'
    elif average >= 80:
        return 'A'
    elif average >= 70:
        return 'B'
    elif average >= 60:
        return 'C'
    else:
        return 'U'

def display_report(name, scores):
    """显示学生成绩报告（过程）"""
    avg = calculate_average(scores)
    grade = determine_grade(avg)
    highest = max(scores)
    lowest = min(scores)
    print(f"  姓名: {name}")
    print(f"  各科成绩: {scores}")
    print(f"  平均分: {avg:.1f}")
    print(f"  最高分: {highest}  最低分: {lowest}")
    print(f"  等级: {grade}")

def generate_class_summary(students):
    """生成班级总结（过程）"""
    all_averages = []
    print("\n" + "=" * 40)
    print("      班级成绩报告")
    print("=" * 40)
    
    for name, scores in students:
        avg = calculate_average(scores)
        all_averages.append(avg)
        print()
        display_report(name, scores)
    
    class_avg = calculate_average(all_averages)
    print("\n" + "-" * 40)
    print(f"  班级平均分: {class_avg:.1f}")
    print(f"  最高平均分: {max(all_averages):.1f}")
    print(f"  最低平均分: {min(all_averages):.1f}")

# 测试数据
students = [
    ("小明", [92, 88, 95, 91]),
    ("小红", [78, 85, 72, 80]),
    ("小刚", [65, 70, 60, 68]),
    ("小丽", [95, 98, 92, 97]),
]

generate_class_summary(students)
````

### 实例3：温度转换器（展示函数的模块化）

````python
# 实例3：温度转换器

def celsius_to_fahrenheit(c):
    """摄氏转华氏"""
    return c * 9 / 5 + 32

def fahrenheit_to_celsius(f):
    """华氏转摄氏"""
    return (f - 32) * 5 / 9

def celsius_to_kelvin(c):
    """摄氏转开尔文"""
    return c + 273.15

def kelvin_to_celsius(k):
    """开尔文转摄氏"""
    return k - 273.15

def format_temp(value, unit):
    """格式化温度输出"""
    return f"{value:.1f}{unit}"

# 综合演示
print("=== 温度转换表 ===")
print(f"{'摄氏(C)':>10} | {'华氏(F)':>10} | {'开尔文(K)':>10}")
print("-" * 38)

for c in range(-20, 101, 10):
    f = celsius_to_fahrenheit(c)
    k = celsius_to_kelvin(c)
    print(f"{format_temp(c, ' C'):>10} | {format_temp(f, ' F'):>10} | {format_temp(k, ' K'):>10}")

# 特殊温度点
print("\n=== 重要温度点 ===")
special_temps = [
    ("绝对零度", -273.15),
    ("水的冰点", 0),
    ("人体体温", 37),
    ("水的沸点", 100),
]
for name, c in special_temps:
    print(f"  {name}: {c} C = {celsius_to_fahrenheit(c):.1f} F = {celsius_to_kelvin(c):.2f} K")
````

````python
# 练习1参考答案

def is_even(n):
    """判断偶数"""
    return n % 2 == 0

def is_prime(n):
    """判断质数"""
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def fibonacci(n):
    """返回第 n 个斐波那契数 (从 0 开始)"""
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

# 测试
print("偶数测试:")
for n in range(1, 11):
    print(f"  {n}: {'偶数' if is_even(n) else '奇数'}", end="")
print()

print("\n质数测试 (2-30):")
primes = [n for n in range(2, 31) if is_prime(n)]
print(f"  {primes}")

print("\n斐波那契数列 (前15个):")
fibs = [fibonacci(i) for i in range(15)]
print(f"  {fibs}")
````

````python
# 练习3参考答案：参数传递分析

def mystery(a, b):
    a = a + 10  # a 是 int（不可变），创建新对象，不影响外部 x
    b.append(10)  # b 是 list（可变），修改原始列表！

x = 5
y = [1, 2, 3]
print(f"调用前: x = {x}, y = {y}")

mystery(x, y)
print(f"调用后: x = {x}, y = {y}")

print("\n解释：")
print("  x 仍然是 5，因为 int 是不可变类型（值传递效果）")
print("  y 变成了 [1, 2, 3, 10]，因为 list 是可变类型（引用传递效果）")
````

## 本节总结

| 概念 | 关键要点 |
|------|----------|
| 结构化编程 | 模块化、自顶向下设计、单一入口出口 |
| 过程 (Procedure) | 命名代码块，执行任务，不返回值 |
| 函数 (Function) | 命名代码块，计算并返回值 |
| 形参 (Parameter) | 函数定义中的占位符 |
| 实参 (Argument) | 调用时传入的实际值 |
| 值传递 (By Value) | 传副本，原件不变 |
| 引用传递 (By Reference) | 传引用，可修改原件 |
| 局部变量 | 函数内定义，函数内使用 |
| 全局变量 | 函数外定义，全局可用（慎用） |
| DRY 原则 | 不要重复自己——提取公共代码为函数 |
| 单一职责 | 每个函数只做一件事 |
