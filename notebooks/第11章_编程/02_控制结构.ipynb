## 1. 三种基本控制结构

### 为什么只需要三种？

1965年，Bohm 和 Jacopini 证明了一个重要定理：**任何可计算的算法都可以仅用三种结构表达**：

| 结构 | 英文 | 含义 | 现实类比 |
|------|------|------|----------|
| 顺序 | Sequence | 按顺序逐行执行 | 按菜谱一步步做菜 |
| 选择 | Selection | 根据条件选择不同路径 | 到路口看红绿灯决定走哪边 |
| 迭代 | Iteration | 重复执行某段代码 | 跑圈——一圈一圈地跑 |

```
顺序：  A → B → C           （依次执行）

选择：  条件? ─ Yes → A      （二选一）
              └ No  → B

迭代：  ┌→ A ─ 条件? ─ Yes → 继续
        └────── No  ──┘       （重复）
```

## 2. 选择结构 (Selection)

### 2.1 IF...THEN...ENDIF（简单条件）

**伪代码：**
```
IF condition
  THEN
    statement(s)
ENDIF
```

**Python：**
```python
if condition:
    statement(s)
```

**理解要点：**
- 条件 (condition) 的结果必须是布尔值 (True/False)
- 只有条件为 True 时，THEN 后面的代码才会执行
- Python 使用**缩进 (indentation)** 来标识代码块，不需要 ENDIF

````python
# IF...THEN 示例
age = 18

if age >= 18:
    print("你已成年，可以投票！")

print("\n--- 另一个例子 ---")
temperature = 38.5

if temperature > 37.5:
    print(f"体温 {temperature}°C，发烧了！")
    print("请及时就医。")
````

### 2.2 IF...THEN...ELSE...ENDIF（二路分支）

**伪代码：**
```
IF condition
  THEN
    statement(s) for True
  ELSE
    statement(s) for False
ENDIF
```

**Python：**
```python
if condition:
    # True 分支
else:
    # False 分支
```

**现实类比：** 红绿灯——绿灯通行，红灯停下。只有两个选择。

````python
# IF...THEN...ELSE 示例
score = 55

if score >= 60:
    print(f"分数 {score}：通过考试！")
else:
    print(f"分数 {score}：未通过，需要补考。")

# 判断奇偶
number = 17
if number % 2 == 0:
    print(f"{number} 是偶数 (even)")
else:
    print(f"{number} 是奇数 (odd)")
````

### 2.3 嵌套 IF (Nested IF)

当需要检查多个条件时，可以在 IF 内部再写 IF：

**伪代码：**
```
IF score >= 90
  THEN OUTPUT "A"
  ELSE
    IF score >= 80
      THEN OUTPUT "B"
      ELSE
        IF score >= 70
          THEN OUTPUT "C"
          ELSE OUTPUT "F"
        ENDIF
    ENDIF
ENDIF
```

**Python 用 elif 简化：**
```python
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("F")
```

> **注意**：深层嵌套的 IF 会导致代码难以阅读和维护——这被称为"箭头代码 (Arrow Code)"。
> 当你发现嵌套超过3层，考虑使用 CASE 结构或重构代码。

````python
# 成绩等级判断 —— 使用 elif
def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

# 测试
test_scores = [95, 87, 73, 65, 42]
for s in test_scores:
    print(f"分数 {s} → 等级 {get_grade(s)}")
````

### 2.4 CASE 结构（多路分支）

当一个变量需要和多个特定值比较时，CASE 结构比多个 IF-ELIF 更清晰：

**伪代码：**
```
CASE OF Choice
  1 : OUTPUT "选项一"
  2 : OUTPUT "选项二"
  3 : OUTPUT "选项三"
  OTHERWISE OUTPUT "无效选项"
ENDCASE
```

**Python（传统方式 - if/elif）：**
```python
if choice == 1:
    print("选项一")
elif choice == 2:
    print("选项二")
elif choice == 3:
    print("选项三")
else:
    print("无效选项")
```

**Python 3.10+ match/case：**
```python
match choice:
    case 1:
        print("选项一")
    case 2:
        print("选项二")
    case 3:
        print("选项三")
    case _:
        print("无效选项")
```

````python
# CASE 示例：星期几
def day_of_week(day_num):
    days = {
        1: "星期一 (Monday)",
        2: "星期二 (Tuesday)",
        3: "星期三 (Wednesday)",
        4: "星期四 (Thursday)",
        5: "星期五 (Friday)",
        6: "星期六 (Saturday)",
        7: "星期日 (Sunday)"
    }
    return days.get(day_num, "无效的日期编号")

for d in range(1, 9):
    print(f"  {d} → {day_of_week(d)}")

# 使用 if/elif 实现 CASE（传统方式）
print("\n--- 计算器 CASE 示例 ---")
num1, num2, operator = 10, 3, '+'

if operator == '+':
    result = num1 + num2
elif operator == '-':
    result = num1 - num2
elif operator == '*':
    result = num1 * num2
elif operator == '/':
    result = num1 / num2 if num2 != 0 else '除数不能为零'
else:
    result = '无效运算符'

print(f"{num1} {operator} {num2} = {result}")
````

### 互动练习：成绩评级系统

下面用不同的选择结构实现同一个功能——比较它们的可读性：

````python
# 成绩评级系统 —— 三种实现方式对比
import matplotlib.pyplot as plt
import numpy as np

def grade_nested_if(score):
    """方法1: 嵌套 IF"""
    if score >= 90:
        return 'A*'
    else:
        if score >= 80:
            return 'A'
        else:
            if score >= 70:
                return 'B'
            else:
                if score >= 60:
                    return 'C'
                else:
                    return 'U'

def grade_elif(score):
    """方法2: elif（推荐）"""
    if score >= 90:
        return 'A*'
    elif score >= 80:
        return 'A'
    elif score >= 70:
        return 'B'
    elif score >= 60:
        return 'C'
    else:
        return 'U'

# 测试并可视化
scores = list(range(0, 101, 5))
grades = [grade_elif(s) for s in scores]

grade_colors = {'A*': '#2ecc71', 'A': '#27ae60', 'B': '#3498db', 'C': '#f39c12', 'U': '#e74c3c'}
colors = [grade_colors[g] for g in grades]

fig, ax = plt.subplots(figsize=(14, 4))
bars = ax.bar(scores, [1]*len(scores), width=4.5, color=colors, edgecolor='white')

# 添加等级标签
for s, g in zip(scores, grades):
    ax.text(s, 0.5, g, ha='center', va='center', fontsize=8, fontweight='bold', color='white')

ax.set_xlabel('分数 (Score)', fontsize=12)
ax.set_title('成绩等级分布 (Grade Distribution)', fontsize=14, fontweight='bold')
ax.set_yticks([])
ax.set_xticks(range(0, 101, 10))

# 图例
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=c, label=g) for g, c in grade_colors.items()]
ax.legend(handles=legend_elements, loc='upper left', ncol=5)

plt.tight_layout()
plt.savefig('grade_distribution.png', dpi=100, bbox_inches='tight')
plt.show()
````

## 3. 迭代结构 (Iteration)

### 3.1 计数控制循环：FOR...TO...NEXT

**何时使用：** 你**已知**需要重复多少次。

**伪代码：**
```
FOR Counter ← Start TO End
    statement(s)
NEXT Counter
```

**带步长 (STEP)：**
```
FOR Counter ← Start TO End STEP StepValue
    statement(s)
NEXT Counter
```

**Python：**
```python
for counter in range(start, end + 1):      # 注意：range 不包含 end+1
    statement(s)

for counter in range(start, end + 1, step): # 带步长
    statement(s)
```

> **重要区别**：伪代码的 `FOR i ← 1 TO 5` 包含 5（从1到5），但 Python 的 `range(1, 5)` 不包含 5！需要写 `range(1, 6)`。

````python
# FOR 循环示例

# 基本 FOR 循环：输出 1 到 5
print("=== 基本 FOR 循环 ===")
for i in range(1, 6):  # 伪代码: FOR i ← 1 TO 5
    print(f"  i = {i}")

# 带步长：偶数
print("\n=== 步长为 2：偶数 ===")
for i in range(2, 11, 2):  # STEP 2
    print(f"  {i}", end=" ")
print()

# 倒计数
print("\n=== 倒计数 ===")
for i in range(10, 0, -1):  # STEP -1
    print(f"  {i}...", end=" ")
print("发射！")

# 实际应用：九九乘法表（部分）
print("\n=== 乘法表 (1-5) ===")
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i}x{j}={i*j:2d}", end="  ")
    print()
````

### 3.2 前条件循环：WHILE...DO...ENDWHILE

**何时使用：** 你**不知道**需要重复多少次，且可能**一次都不执行**。

**伪代码：**
```
WHILE condition DO
    statement(s)
ENDWHILE
```

**Python：**
```python
while condition:
    statement(s)
```

**关键特点：**
- 条件在循环体**之前**检查（pre-condition）
- 如果条件一开始就是 False，循环体**一次都不执行**
- 必须确保循环体内有代码能让条件变为 False，否则会**无限循环 (infinite loop)**！

**现实类比：** "当还没吃饱的时候，继续吃"——如果你已经吃饱了，就不会再吃了（0次执行）。

````python
# WHILE 循环示例

# 示例1：累加到超过100
print("=== 累加直到超过 100 ===")
total = 0
count = 0
while total <= 100:
    count += 1
    total += count
    print(f"  加上 {count}，总计 = {total}")
print(f"最终：加了 {count} 个数，总计 {total}")

# 示例2：可能一次都不执行
print("\n=== 可能不执行的 WHILE ===")
x = 100
while x < 10:
    print("这行不会被打印！")
    x += 1
print(f"x 已经是 {x}，条件一开始就是 False，循环没有执行。")

# 示例3：猜数字（模拟）
print("\n=== 猜数字游戏（模拟） ===")
import random
secret = random.randint(1, 10)
guesses = [3, 7, 5, secret]  # 模拟用户输入
idx = 0
guess = guesses[idx]
while guess != secret:
    print(f"  猜 {guess}... 不对！")
    idx += 1
    guess = guesses[idx]
print(f"  猜 {guess}... 正确！答案就是 {secret}")
````

### 3.3 后条件循环：REPEAT...UNTIL

**何时使用：** 循环体**至少执行一次**（如用户输入验证、菜单）。

**伪代码：**
```
REPEAT
    statement(s)
UNTIL condition
```

**Python（没有原生 REPEAT...UNTIL，使用 while True + break）：**
```python
while True:
    statement(s)
    if condition:
        break
```

**关键特点：**
- 条件在循环体**之后**检查（post-condition）
- 循环体**至少执行一次**
- 注意：REPEAT...UNTIL 在条件为 **True** 时**停止**（和 WHILE 相反！）

**现实类比：** "先吃一口，再决定要不要继续吃"——至少会吃一口。

````python
# REPEAT...UNTIL 示例

# 示例1：输入验证（模拟）
print("=== 输入验证（模拟 REPEAT...UNTIL） ===")
test_inputs = [-5, 0, 150, 75]  # 模拟用户输入
idx = 0
while True:  # REPEAT
    score = test_inputs[idx]
    print(f"  输入: {score}")
    idx += 1
    if 0 <= score <= 100:  # UNTIL valid
        break
    print("    无效！分数必须在 0-100 之间。")
print(f"有效分数: {score}")

# 示例2：菜单系统（模拟）
print("\n=== 菜单系统 ===")
menu_choices = [1, 2, 3]  # 模拟用户选择
for choice in menu_choices:
    print(f"\n--- 菜单 ---")
    print("1. 查看余额")
    print("2. 存款")
    print("3. 退出")
    print(f"用户选择: {choice}")
    if choice == 1:
        print("  余额: $1000")
    elif choice == 2:
        print("  存款操作...")
    elif choice == 3:
        print("  再见！")
        break
````

### 三种循环对比

| 特性 | FOR | WHILE | REPEAT...UNTIL |
|------|-----|-------|----------------|
| 重复次数 | **已知** | **未知** | **未知** |
| 条件检查 | 自动计数 | 循环**前** | 循环**后** |
| 最少执行 | 0次或指定次数 | **0次** | **1次** |
| 典型场景 | 遍历数组、计数 | 搜索、文件读取 | 输入验证、菜单 |
| 伪代码 | `FOR...NEXT` | `WHILE...ENDWHILE` | `REPEAT...UNTIL` |
| Python | `for...in range()` | `while condition:` | `while True: ...break` |

**选择规则：**
1. 知道重复几次？→ **FOR**
2. 不知道次数，可能不执行？→ **WHILE**
3. 不知道次数，至少执行一次？→ **REPEAT...UNTIL**

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, axes = plt.subplots(1, 3, figsize=(18, 8))

def draw_box(ax, x, y, w, h, text, color='#3498db'):
    rect = patches.FancyBboxPatch((x, y), w, h, boxstyle='round,pad=0.1',
                                   facecolor=color, edgecolor='#2c3e50', linewidth=2)
    ax.add_patch(rect)
    ax.text(x+w/2, y+h/2, text, ha='center', va='center', fontsize=9, color='white', fontweight='bold')

def draw_diamond(ax, x, y, text, color='#f39c12'):
    diamond = patches.RegularPolygon((x, y), 4, radius=0.7, orientation=0,
                                      facecolor=color, edgecolor='#2c3e50', linewidth=2)
    ax.add_patch(diamond)
    ax.text(x, y, text, ha='center', va='center', fontsize=8, fontweight='bold')

# --- FOR Loop ---
ax = axes[0]
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 10)
ax.axis('off')
ax.set_title('FOR 循环', fontsize=14, fontweight='bold')

draw_box(ax, 0.5, 8.5, 3, 0.8, 'i = start', '#27ae60')
draw_diamond(ax, 2, 7, 'i <= end?')
draw_box(ax, 0.5, 5, 3, 0.8, '执行循环体', '#3498db')
draw_box(ax, 0.5, 3.5, 3, 0.8, 'i = i + step', '#9b59b6')
draw_box(ax, 0.5, 1, 3, 0.8, '继续执行', '#95a5a6')

ax.annotate('', xy=(2, 7.7), xytext=(2, 8.5), arrowprops=dict(arrowstyle='->', lw=1.5))
ax.annotate('', xy=(2, 5.8), xytext=(2, 6.3), arrowprops=dict(arrowstyle='->', lw=1.5))
ax.text(2.3, 6.5, 'Yes', fontsize=9, color='#2ecc71')
ax.annotate('', xy=(2, 4.3), xytext=(2, 5), arrowprops=dict(arrowstyle='->', lw=1.5))
ax.annotate('', xy=(0.3, 7), xytext=(0.3, 3.9),
           arrowprops=dict(arrowstyle='->', lw=1.5, connectionstyle='arc3,rad=-0.5'))
ax.annotate('', xy=(2, 1.8), xytext=(4, 7),
           arrowprops=dict(arrowstyle='->', lw=1.5, connectionstyle='arc3,rad=-0.3', color='#e74c3c'))
ax.text(4.1, 7.3, 'No', fontsize=9, color='#e74c3c')

# --- WHILE Loop ---
ax = axes[1]
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 10)
ax.axis('off')
ax.set_title('WHILE 循环 (前条件)', fontsize=14, fontweight='bold')

draw_diamond(ax, 2, 8, '条件?')
draw_box(ax, 0.5, 5.5, 3, 0.8, '执行循环体', '#3498db')
draw_box(ax, 0.5, 2, 3, 0.8, '继续执行', '#95a5a6')

ax.annotate('', xy=(2, 6.3), xytext=(2, 7.3), arrowprops=dict(arrowstyle='->', lw=1.5))
ax.text(2.3, 6.9, 'Yes', fontsize=9, color='#2ecc71')
ax.annotate('', xy=(0.3, 8), xytext=(0.3, 5.5),
           arrowprops=dict(arrowstyle='->', lw=1.5, connectionstyle='arc3,rad=-0.5'))
ax.annotate('', xy=(2, 2.8), xytext=(4, 8),
           arrowprops=dict(arrowstyle='->', lw=1.5, connectionstyle='arc3,rad=-0.3', color='#e74c3c'))
ax.text(4.1, 8.3, 'No', fontsize=9, color='#e74c3c')
ax.text(2, 9.3, '先检查条件', fontsize=10, ha='center', style='italic', color='#e67e22')
ax.text(2, 0.8, '可能 0 次执行!', fontsize=10, ha='center', color='#e74c3c', fontweight='bold')

# --- REPEAT...UNTIL Loop ---
ax = axes[2]
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 10)
ax.axis('off')
ax.set_title('REPEAT...UNTIL 循环 (后条件)', fontsize=14, fontweight='bold')

draw_box(ax, 0.5, 7.5, 3, 0.8, '执行循环体', '#3498db')
draw_diamond(ax, 2, 5.5, '条件?')
draw_box(ax, 0.5, 2, 3, 0.8, '继续执行', '#95a5a6')

ax.annotate('', xy=(2, 6.2), xytext=(2, 7.5), arrowprops=dict(arrowstyle='->', lw=1.5))
ax.annotate('', xy=(2, 2.8), xytext=(2, 4.8), arrowprops=dict(arrowstyle='->', lw=1.5))
ax.text(2.3, 3.9, 'Yes(结束)', fontsize=9, color='#2ecc71')
ax.annotate('', xy=(4, 7.9), xytext=(4, 5.5),
           arrowprops=dict(arrowstyle='->', lw=1.5, connectionstyle='arc3,rad=-0.5', color='#e74c3c'))
ax.text(4.2, 6.7, 'No(继续)', fontsize=9, color='#e74c3c')
ax.text(2, 9.3, '先执行一次', fontsize=10, ha='center', style='italic', color='#e67e22')
ax.text(2, 0.8, '至少执行 1 次!', fontsize=10, ha='center', color='#2ecc71', fontweight='bold')

plt.tight_layout()
plt.savefig('loop_flowcharts.png', dpi=100, bbox_inches='tight')
plt.show()
print("三种循环流程图可视化完成！")
````

### 互动：三种循环对比演示

下面用同一个任务（求 1+2+...+N 的和）展示三种循环的写法：

````python
# 用三种循环计算 1+2+3+...+10 的和
N = 10

# 方法1: FOR 循环
total_for = 0
for i in range(1, N + 1):
    total_for += i
print(f"FOR 循环:    1+2+...+{N} = {total_for}")

# 方法2: WHILE 循环
total_while = 0
i = 1
while i <= N:
    total_while += i
    i += 1
print(f"WHILE 循环:  1+2+...+{N} = {total_while}")

# 方法3: REPEAT...UNTIL（while True + break）
total_repeat = 0
i = 1
while True:
    total_repeat += i
    i += 1
    if i > N:
        break
print(f"REPEAT 循环: 1+2+...+{N} = {total_repeat}")

# 验证：数学公式
formula_result = N * (N + 1) // 2
print(f"\n数学公式 N*(N+1)/2 = {formula_result}")
print(f"结果一致: {total_for == total_while == total_repeat == formula_result}")
````

## 4. 嵌套 (Nesting)

控制结构可以**嵌套**——循环里面放循环，条件里面放条件，循环里面放条件……

**形象理解：** 嵌套就像俄罗斯套娃——一个结构里面包含另一个结构。

````python
# 嵌套示例1：循环中嵌套条件
print("=== 1-20 中的偶数和奇数 ===")
for num in range(1, 21):
    if num % 2 == 0:
        print(f"  {num:2d} → 偶数", end="")
    else:
        print(f"  {num:2d} → 奇数", end="")
    if num % 5 == 0:
        print("  (5的倍数)", end="")
    print()

# 嵌套示例2：循环中嵌套循环
print("\n=== 图案打印 ===")
rows = 5
for i in range(1, rows + 1):
    for j in range(i):
        print("* ", end="")
    print()

print()
for i in range(rows, 0, -1):
    print("  " * (rows - i), end="")
    for j in range(i):
        print("* ", end="")
    print()
````

## 5. 常见编程模式 (Common Patterns)

以下是考试中经常出现的编程模式，务必掌握：

````python
# 模式1：输入验证循环 (Input Validation Loop)
print("=== 模式1：输入验证 ===")
# 伪代码:
# REPEAT
#   INPUT Score
# UNTIL Score >= 0 AND Score <= 100

test_inputs = [-1, 200, 85]  # 模拟输入
idx = 0
while True:
    score = test_inputs[idx]
    print(f"  输入分数: {score}")
    idx += 1
    if 0 <= score <= 100:
        break
    print("    错误：分数必须在 0~100 之间！")
print(f"  有效分数: {score}")
````

````python
# 模式2：累加器/计数器 (Running Total / Counter)
print("=== 模式2：累加与计数 ===")
numbers = [85, 92, 78, 95, 60, 73, 88]

total = 0      # 累加器 (accumulator)
count = 0      # 计数器 (counter)
pass_count = 0 # 通过人数

for num in numbers:
    total += num
    count += 1
    if num >= 70:
        pass_count += 1

average = total / count
print(f"  总分: {total}")
print(f"  人数: {count}")
print(f"  平均分: {average:.1f}")
print(f"  通过人数 (>=70): {pass_count}")
print(f"  通过率: {pass_count/count*100:.1f}%")
````

````python
# 模式3：哨兵值循环 (Sentinel Value Loop)
print("=== 模式3：哨兵值循环 ===")
# 用一个特殊值（如 -1）来标记输入结束
# 伪代码:
# INPUT Number
# WHILE Number <> -1 DO
#   Total ← Total + Number
#   INPUT Number
# ENDWHILE

data = [10, 20, 30, 40, -1]  # -1 是哨兵值
idx = 0
total = 0
count = 0

number = data[idx]
idx += 1
while number != -1:  # -1 是哨兵值 (sentinel)
    total += number
    count += 1
    print(f"  读入: {number}，当前总和: {total}")
    number = data[idx]
    idx += 1

print(f"  遇到哨兵值 -1，停止。")
print(f"  总和: {total}，共 {count} 个数")
````

````python
# 模式4：菜单驱动程序 (Menu-Driven Program)
print("=== 模式4：菜单驱动程序 ===")

def show_menu():
    print("\n" + "=" * 30)
    print("   简易计算器")
    print("=" * 30)
    print("1. 加法")
    print("2. 减法")
    print("3. 乘法")
    print("4. 退出")
    print("=" * 30)

# 模拟用户选择
user_choices = [(1, 10, 5), (3, 7, 8), (4, 0, 0)]

for choice, a, b in user_choices:
    show_menu()
    print(f"用户选择: {choice}")
    if choice == 1:
        print(f"  {a} + {b} = {a + b}")
    elif choice == 2:
        print(f"  {a} - {b} = {a - b}")
    elif choice == 3:
        print(f"  {a} * {b} = {a * b}")
    elif choice == 4:
        print("  感谢使用，再见！")
        break
    else:
        print("  无效选择！")
````

````python
# 练习1参考答案：闰年判断
def is_leap_year(year):
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False

test_years = [2000, 1900, 2024, 2023, 2100, 2400]
for y in test_years:
    result = "闰年" if is_leap_year(y) else "平年"
    print(f"  {y} → {result}")
````

````python
# 练习4参考答案：密码验证
CORRECT_PASSWORD = "cs9618"
MAX_ATTEMPTS = 3

# 模拟用户输入
user_inputs = ["wrong1", "wrong2", "cs9618"]

attempts = 0
locked = False

while True:
    password = user_inputs[attempts]
    attempts += 1
    print(f"尝试 {attempts}: 输入密码 '{password}'")
    
    if password == CORRECT_PASSWORD:
        print("  登录成功！")
        break
    else:
        remaining = MAX_ATTEMPTS - attempts
        if remaining > 0:
            print(f"  密码错误！剩余 {remaining} 次机会")
        else:
            print("  密码错误！账户已锁定！")
            locked = True
            break
````
