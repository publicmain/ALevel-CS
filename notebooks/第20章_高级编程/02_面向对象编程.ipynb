## 1. OOP的基本思想：为什么需要面向对象？

### 1.1 过程式编程的局限

想象你正在开发一个**校园管理系统**。用过程式编程，你会写：

```python
student_name = "张三"
student_age = 17
student_grade = "A"

teacher_name = "李老师"
teacher_subject = "物理"
```

问题来了：
- 如果有1000个学生怎么办？变量爆炸！
- 学生和教师的数据混在一起，容易搞混
- 修改学生结构需要改很多地方

### 1.2 OOP的核心思想

**面向对象编程 (Object-Oriented Programming, OOP)** 的核心思想是：

> **将现实世界的事物抽象为"对象"(Object)，每个对象有自己的属性(数据)和行为(方法)。**

就像现实世界中，"学生"是一类事物，每个学生都有姓名、年龄、成绩等**属性**，也有学习、考试等**行为**。

### 1.3 四大支柱 (Four Pillars of OOP)

1. **抽象 (Abstraction)** - 只展示必要信息，隐藏复杂细节
2. **封装 (Encapsulation)** - 将数据和方法包装在一起，保护数据
3. **继承 (Inheritance)** - 子类继承父类的特征
4. **多态 (Polymorphism)** - 同一方法在不同对象上有不同表现

## 2. 类 (Class) 与对象 (Object)

### 2.1 类与对象的关系

- **类 (Class)** = 蓝图/模板 (Blueprint/Template)
- **对象 (Object)** = 根据蓝图创建的实例 (Instance)

> **类比**："学生"是一个类（概念），"张三"是一个对象（具体的学生）。"汽车"是一个类，你家的那辆白色特斯拉是一个对象。

### 2.2 属性 (Attributes) 与方法 (Methods)

- **属性 (Attributes/Properties)** = 对象的数据/特征
- **方法 (Methods)** = 对象的行为/功能

````python
# 第一个OOP示例：定义一个学生类

class Student:
    """
    学生类 (Student Class)
    这是一个蓝图，定义了"学生"应该有什么属性和行为
    """
    
    # 构造函数 (Constructor) - 创建对象时自动调用
    def __init__(self, name, age, student_id):
        """
        __init__ 是Python的构造函数
        self 代表当前对象自身
        """
        self.name = name           # 属性：姓名
        self.age = age             # 属性：年龄
        self.student_id = student_id  # 属性：学号
        self.grades = {}           # 属性：成绩字典
        print(f'  [构造] 学生对象 "{name}" 已创建')
    
    # 方法 (Method) - 对象的行为
    def add_grade(self, subject, score):
        """添加成绩"""
        self.grades[subject] = score
    
    def get_average(self):
        """计算平均分"""
        if not self.grades:
            return 0
        return sum(self.grades.values()) / len(self.grades)
    
    def introduce(self):
        """自我介绍"""
        avg = self.get_average()
        return f'我是{self.name}，学号{self.student_id}，今年{self.age}岁，平均分{avg:.1f}'
    
    # 析构函数 (Destructor) - 对象被销毁时自动调用
    def __del__(self):
        print(f'  [析构] 学生对象 "{self.name}" 已销毁')


# ============ 创建对象 (实例化 Instantiation) ============
print('=== 创建学生对象 ===')
s1 = Student('张三', 17, 'S001')
s2 = Student('李四', 16, 'S002')
print()

# 使用方法
s1.add_grade('数学', 92)
s1.add_grade('物理', 88)
s1.add_grade('化学', 95)

s2.add_grade('数学', 78)
s2.add_grade('物理', 85)
s2.add_grade('化学', 72)

print('=== 学生自我介绍 ===')
print(s1.introduce())
print(s2.introduce())
print()

# 类与对象的关系
print('=== 类与对象的关系 ===')
print(f's1的类型: {type(s1)}')
print(f's1是Student的实例吗? {isinstance(s1, Student)}')
print(f's1和s2是同一个对象吗? {s1 is s2}')
````

## 3. 封装 (Encapsulation)

### 3.1 什么是封装？

**封装 (Encapsulation)** 是将数据（属性）和操作数据的方法**绑定在一起**，并**限制外部对数据的直接访问**。

### 3.2 为什么需要封装？

想象一个银行账户：
- 你可以查看余额（通过柜台/ATM）
- 你可以存钱取钱（通过正规渠道）
- 但你**不能**直接修改数据库里的余额数字！

封装就是这种"保护机制"。

### 3.3 访问修饰符 (Access Modifiers)

| 修饰符 | Python表示 | 含义 |
|--------|-----------|------|
| Public | `self.name` | 公有，任何地方都能访问 |
| Private | `self.__name` | 私有，只能在类内部访问 |
| Protected | `self._name` | 受保护，类和子类可以访问（Python约定） |

**注意**：Cambridge考试中使用 `PRIVATE` 和 `PUBLIC` 关键字（伪代码）。Python用双下划线 `__` 表示私有。

````python
# 封装示例：银行账户类

class BankAccount:
    """
    银行账户类 - 演示封装
    余额是私有的，只能通过方法来操作
    """
    
    def __init__(self, owner, initial_balance=0):
        self.owner = owner              # public: 公有属性
        self.__balance = initial_balance # private: 私有属性（双下划线）
        self.__transaction_log = []     # private: 交易记录
    
    # Getter方法 - 安全地获取私有数据
    def get_balance(self):
        """获取余额 - 只读访问"""
        return self.__balance
    
    def deposit(self, amount):
        """存款 - 有验证的写操作"""
        if amount <= 0:
            print(f'  错误：存款金额必须为正数！')
            return False
        self.__balance += amount
        self.__transaction_log.append(f'存款 +{amount}')
        print(f'  存款成功：+{amount}，余额：{self.__balance}')
        return True
    
    def withdraw(self, amount):
        """取款 - 有验证的写操作"""
        if amount <= 0:
            print(f'  错误：取款金额必须为正数！')
            return False
        if amount > self.__balance:
            print(f'  错误：余额不足！当前余额：{self.__balance}')
            return False
        self.__balance -= amount
        self.__transaction_log.append(f'取款 -{amount}')
        print(f'  取款成功：-{amount}，余额：{self.__balance}')
        return True
    
    def get_statement(self):
        """获取账户报告"""
        print(f'\n=== 账户报告：{self.owner} ===')
        for record in self.__transaction_log:
            print(f'  {record}')
        print(f'  当前余额：{self.__balance}')
        print('=' * 30)


# ============ 演示封装 ============
print('=== 封装 (Encapsulation) 演示 ===')
print()

account = BankAccount('张三', 1000)
print(f'账户所有者（public）: {account.owner}')
print(f'账户余额（通过getter）: {account.get_balance()}')
print()

# 正常操作
print('--- 通过方法操作（正确方式）---')
account.deposit(500)
account.withdraw(200)
account.withdraw(2000)  # 余额不足
account.deposit(-100)   # 无效金额
print()

# 尝试直接访问私有属性
print('--- 尝试直接访问私有属性 ---')
try:
    print(account.__balance)  # 会报错！
except AttributeError as e:
    print(f'  错误：{e}')
    print('  私有属性无法从外部直接访问！这就是封装的保护作用。')
print()

# 账户报告
account.get_statement()
print()
print('封装的意义：')
print('  1. 数据保护：防止外部代码随意修改余额')
print('  2. 数据验证：存取款前检查金额是否有效')
````

## 4. 继承 (Inheritance)

### 4.1 什么是继承？

**继承 (Inheritance)** 允许一个类（子类/派生类）从另一个类（父类/基类）继承属性和方法。

- **父类 / 基类 (Parent/Base Class)** = 超类 (Superclass)
- **子类 / 派生类 (Child/Derived Class)** = 子类 (Subclass)

### 4.2 为什么需要继承？

1. **代码复用 (Code Reuse)**：不用重复写相同的代码
2. **层次结构 (Hierarchy)**：反映现实世界的分类关系
3. **扩展性 (Extensibility)**：在不修改原有代码的情况下添加新功能

> **类比**："动物"是父类，"猫"和"狗"是子类。猫和狗都会"吃"和"睡"（继承自动物），但猫会"喵"而狗会"汪"（各自特有的行为）。

````python
# 继承示例：动物园管理系统

class Animal:
    """动物基类 (Base Class)"""
    
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self.age = age
        self.__health = 100  # 私有属性
    
    def eat(self, food):
        print(f'  {self.name}({self.species})正在吃{food}')
    
    def sleep(self):
        print(f'  {self.name}正在睡觉... zzz')
    
    def get_health(self):
        return self.__health
    
    def get_info(self):
        return f'{self.name} - {self.species}, {self.age}岁, 健康值:{self.__health}'


class Mammal(Animal):
    """哺乳动物类 - 继承自Animal"""
    
    def __init__(self, name, species, age, fur_color):
        super().__init__(name, species, age)  # 调用父类构造函数
        self.fur_color = fur_color  # 子类新增属性
    
    def nurse(self):
        """哺乳 - 子类新增方法"""
        print(f'  {self.name}正在哺乳幼崽')


class Dog(Mammal):
    """狗类 - 继承自Mammal"""
    
    def __init__(self, name, age, fur_color, breed):
        super().__init__(name, '犬科', age, fur_color)
        self.breed = breed  # 品种
    
    def bark(self):
        """汪汪叫 - Dog特有方法"""
        print(f'  {self.name}：汪汪汪！！')
    
    def fetch(self, item):
        print(f'  {self.name}捡回了{item}')


class Cat(Mammal):
    """猫类 - 继承自Mammal"""
    
    def __init__(self, name, age, fur_color, is_indoor):
        super().__init__(name, '猫科', age, fur_color)
        self.is_indoor = is_indoor
    
    def meow(self):
        print(f'  {self.name}：喵喵喵～')
    
    def purr(self):
        print(f'  {self.name}发出呼噜声...')


class Bird(Animal):
    """鸟类 - 继承自Animal"""
    
    def __init__(self, name, species, age, can_fly):
        super().__init__(name, species, age)
        self.can_fly = can_fly
    
    def sing(self):
        print(f'  {self.name}在唱歌：啾啾啾～')


# ============ 演示继承 ============
print('=== 继承 (Inheritance) 演示 ===')
print()

dog = Dog('旺财', 3, '金色', '金毛寻回犬')
cat = Cat('咪咪', 2, '白色', True)
bird = Bird('小黄', '金丝雀', 1, True)

print('--- 狗的行为 ---')
print(dog.get_info())        # 继承自Animal
dog.eat('骨头')              # 继承自Animal
dog.nurse()                  # 继承自Mammal
dog.bark()                   # Dog特有
dog.fetch('飞盘')            # Dog特有
print()

print('--- 猫的行为 ---')
print(cat.get_info())
cat.eat('鱼')               # 继承自Animal
cat.meow()                  # Cat特有
cat.purr()                  # Cat特有
print()

print('--- 继承关系验证 ---')
print(f'Dog是Animal的子类吗？ {issubclass(Dog, Animal)}')
print(f'Dog是Mammal的子类吗？ {issubclass(Dog, Mammal)}')
print(f'Bird是Mammal的子类吗？ {issubclass(Bird, Mammal)}')
````

## 5. 多态 (Polymorphism)

### 5.1 什么是多态？

**多态 (Polymorphism)** 的字面意思是"多种形态"。在OOP中，它指的是：

> **同一个方法名，在不同的类中可以有不同的实现。**

### 5.2 方法重写 (Override)

子类可以**重写 (Override)** 父类的方法，提供自己的实现。

### 5.3 为什么多态重要？

多态让我们可以用**统一的接口**处理不同类型的对象，而不需要知道对象的具体类型。这大大简化了代码。

````python
# 多态示例：图形计算系统
import math

class Shape:
    """图形基类"""
    def __init__(self, name):
        self.name = name
    
    def area(self):
        """计算面积 - 子类必须重写此方法"""
        raise NotImplementedError('子类必须实现area()方法')
    
    def perimeter(self):
        """计算周长 - 子类必须重写此方法"""
        raise NotImplementedError('子类必须实现perimeter()方法')
    
    def describe(self):
        return f'{self.name}: 面积={self.area():.2f}, 周长={self.perimeter():.2f}'


class Circle(Shape):
    def __init__(self, radius):
        super().__init__('圆形')
        self.radius = radius
    
    def area(self):          # 重写 (Override)
        return math.pi * self.radius ** 2
    
    def perimeter(self):     # 重写 (Override)
        return 2 * math.pi * self.radius


class Rectangle(Shape):
    def __init__(self, width, height):
        super().__init__('矩形')
        self.width = width
        self.height = height
    
    def area(self):          # 重写
        return self.width * self.height
    
    def perimeter(self):     # 重写
        return 2 * (self.width + self.height)


class Triangle(Shape):
    def __init__(self, a, b, c):
        super().__init__('三角形')
        self.a, self.b, self.c = a, b, c
    
    def area(self):          # 重写 - 海伦公式
        s = (self.a + self.b + self.c) / 2
        return math.sqrt(s * (s-self.a) * (s-self.b) * (s-self.c))
    
    def perimeter(self):     # 重写
        return self.a + self.b + self.c


# ============ 多态演示 ============
print('=== 多态 (Polymorphism) 演示 ===')
print()

# 创建不同类型的图形
shapes = [
    Circle(5),
    Rectangle(4, 6),
    Triangle(3, 4, 5),
    Circle(3),
    Rectangle(10, 2),
]

# 多态的威力：用同一个接口处理不同类型的对象！
print('用统一接口处理不同类型的对象：')
for shape in shapes:
    print(f'  {shape.describe()}')

print()
print('多态的威力：')
print('  所有图形都有 area() 和 perimeter() 方法')
print('  但每种图形的计算方式完全不同！')
print('  调用者不需要知道具体是什么图形。')
print()

# 计算总面积
total_area = sum(shape.area() for shape in shapes)
print(f'所有图形的总面积: {total_area:.2f}')
````

## 6. 组合与聚合 (Containment/Aggregation)

### 6.1 概念

除了继承(IS-A关系)，对象之间还有另一种重要的关系：**组合/包含(HAS-A关系)**。

- **组合 (Composition)**：强关系 - 部分离开整体就没有意义。如：心脏和人体
- **聚合 (Aggregation)**：弱关系 - 部分可以独立存在。如：学生和班级

### 6.2 继承 vs 组合

| | 继承 (Inheritance) | 组合 (Composition) |
|--|-------------------|-------------------|
| 关系 | IS-A (是一个) | HAS-A (有一个) |
| 例子 | Dog IS-A Animal | Car HAS-A Engine |
| 耦合度 | 高 | 低 |
| 灵活性 | 较低 | 较高 |

````python
# 组合与聚合示例：学校系统

class Address:
    """地址类"""
    def __init__(self, city, street, number):
        self.city = city
        self.street = street
        self.number = number
    
    def __str__(self):
        return f'{self.city}{self.street}{self.number}号'


class Subject:
    """科目类"""
    def __init__(self, name, code):
        self.name = name
        self.code = code
    
    def __str__(self):
        return f'{self.name}({self.code})'


class Teacher:
    """教师类 - 聚合了Subject（教师可以教多门课）"""
    def __init__(self, name, employee_id):
        self.name = name
        self.employee_id = employee_id
        self.subjects = []  # 聚合：教师HAS subjects
    
    def add_subject(self, subject):
        self.subjects.append(subject)
    
    def __str__(self):
        subs = ', '.join(str(s) for s in self.subjects)
        return f'{self.name}(工号:{self.employee_id}) - 教授: {subs}'


class Classroom:
    """教室类 - 组合了Address（教室离开学校没意义）"""
    def __init__(self, room_number, capacity, address):
        self.room_number = room_number
        self.capacity = capacity
        self.address = address  # 组合：教室HAS-A address
    
    def __str__(self):
        return f'教室{self.room_number}（容量:{self.capacity}人）- {self.address}'


class School:
    """学校类 - 聚合了Teacher和Classroom"""
    def __init__(self, name):
        self.name = name
        self.teachers = []     # 聚合
        self.classrooms = []   # 组合
    
    def add_teacher(self, teacher):
        self.teachers.append(teacher)
    
    def add_classroom(self, classroom):
        self.classrooms.append(classroom)
    
    def show_info(self):
        print(f'\n=== {self.name} ===')
        print(f'教师 ({len(self.teachers)}人):')
        for t in self.teachers:
            print(f'  {t}')
        print(f'教室 ({len(self.classrooms)}间):')
        for c in self.classrooms:
            print(f'  {c}')


# ============ 演示 ============
print('=== 组合与聚合 (Composition & Aggregation) 演示 ===')

# 创建科目
math = Subject('数学', 'MATH101')
physics = Subject('物理', 'PHYS101')
cs = Subject('计算机科学', 'CS101')

# 创建教师（聚合科目）
t1 = Teacher('李老师', 'T001')
t1.add_subject(math)
t1.add_subject(physics)

t2 = Teacher('王老师', 'T002')
t2.add_subject(cs)

# 创建地址和教室（组合关系）
addr = Address('北京市', '学院路', 35)
room1 = Classroom('A101', 40, addr)
room2 = Classroom('B203', 30, addr)

# 创建学校（聚合教师和教室）
school = School('星辰国际学校')
school.add_teacher(t1)
school.add_teacher(t2)
school.add_classroom(room1)
school.add_classroom(room2)

school.show_info()

print()
print('关系分析：')
print('  School HAS Teachers (聚合 - 教师可以离职去别的学校)')
print('  School HAS Classrooms (组合 - 教室属于学校)')
print('  Teacher HAS Subjects (聚合 - 科目独立存在)')
````

## 7. 综合示例：游戏角色系统

让我们用一个完整的游戏角色系统来综合展示OOP的所有概念。

````python
# 综合OOP示例：RPG游戏角色系统
import random

class Weapon:
    """武器类 (用于组合)"""
    def __init__(self, name, damage, weapon_type):
        self.name = name
        self.damage = damage
        self.weapon_type = weapon_type
    
    def __str__(self):
        return f'{self.name}(伤害:{self.damage})'


class Character:
    """游戏角色基类"""
    
    def __init__(self, name, hp, attack, defense):
        self.name = name
        self.__hp = hp             # 私有：当前生命值
        self.__max_hp = hp         # 私有：最大生命值
        self.attack = attack
        self.defense = defense
        self.weapon = None         # 组合：角色HAS-A武器
        self.is_alive = True
    
    def get_hp(self):
        return self.__hp
    
    def get_max_hp(self):
        return self.__max_hp
    
    def equip_weapon(self, weapon):
        self.weapon = weapon
        print(f'  {self.name} 装备了 {weapon}')
    
    def take_damage(self, damage):
        actual_damage = max(1, damage - self.defense)
        self.__hp = max(0, self.__hp - actual_damage)
        if self.__hp == 0:
            self.is_alive = False
        return actual_damage
    
    def heal(self, amount):
        self.__hp = min(self.__max_hp, self.__hp + amount)
    
    def basic_attack(self, target):
        """基本攻击 - 可被子类重写"""
        weapon_dmg = self.weapon.damage if self.weapon else 0
        total_dmg = self.attack + weapon_dmg
        actual = target.take_damage(total_dmg)
        print(f'  {self.name} 攻击 {target.name}，造成 {actual} 伤害 (HP: {target.get_hp()}/{target.get_max_hp()})')
    
    def special_ability(self):
        """特殊技能 - 子类必须重写 (多态)"""
        raise NotImplementedError
    
    def status(self):
        weapon_info = str(self.weapon) if self.weapon else '无'
        return f'{self.name} | HP:{self.__hp}/{self.__max_hp} | ATK:{self.attack} | DEF:{self.defense} | 武器:{weapon_info}'


class Warrior(Character):
    """战士类 - 继承自Character"""
    def __init__(self, name):
        super().__init__(name, hp=150, attack=20, defense=15)
        self.rage = 0
    
    def special_ability(self):  # 重写 (Override) - 多态
        """狂暴打击：消耗怒气，造成双倍伤害"""
        self.rage += 30
        bonus = self.rage // 10
        print(f'  {self.name} 发动【狂暴打击】！怒气值:{self.rage}，额外伤害+{bonus}')
        return self.attack * 2 + bonus


class Mage(Character):
    """法师类 - 继承自Character"""
    def __init__(self, name):
        super().__init__(name, hp=80, attack=10, defense=5)
        self.__mana = 100
    
    def special_ability(self):  # 重写 (Override) - 多态
        """火球术：消耗法力，造成高魔法伤害"""
        if self.__mana >= 30:
            self.__mana -= 30
            damage = 50
            print(f'  {self.name} 释放【火球术】！剩余法力:{self.__mana}')
            return damage
        else:
            print(f'  {self.name} 法力不足！')
            return 0


class Healer(Character):
    """治疗师类 - 继承自Character"""
    def __init__(self, name):
        super().__init__(name, hp=100, attack=8, defense=8)
        self.__faith = 100
    
    def special_ability(self):  # 重写 (Override) - 多态
        """治愈之光：恢复自身生命"""
        if self.__faith >= 20:
            self.__faith -= 20
            heal_amount = 40
            self.heal(heal_amount)
            print(f'  {self.name} 使用【治愈之光】！恢复{heal_amount}HP (HP:{self.get_hp()}/{self.get_max_hp()})')
            return 0
        else:
            print(f'  {self.name} 信仰不足！')
            return 0


# ============ 游戏演示 ============
print('=== RPG游戏角色系统 - OOP综合演示 ===')
print()

# 创建角色
warrior = Warrior('亚瑟')
mage = Mage('梅林')
healer = Healer('天使')

# 创建武器（组合关系）
sword = Weapon('圣剑', 15, '近战')
staff = Weapon('魔杖', 8, '法杖')
wand = Weapon('神圣权杖', 5, '法杖')

# 装备武器
warrior.equip_weapon(sword)
mage.equip_weapon(staff)
healer.equip_weapon(wand)
print()

# 显示状态
print('--- 角色状态 ---')
team = [warrior, mage, healer]
for char in team:
    print(f'  {char.status()}')
print()

# 模拟战斗 - 多态的威力！
print('--- 模拟战斗 ---')
enemy = Warrior('暗影骑士')
enemy.equip_weapon(Weapon('暗影之刃', 12, '近战'))
print(f'  敌人: {enemy.status()}')
print()

# 每个角色使用特殊技能（同一方法，不同行为 = 多态）
print('所有角色使用特殊技能（多态 Polymorphism）：')
for char in team:
    dmg = char.special_ability()
    if dmg > 0:
        actual = enemy.take_damage(dmg)
        print(f'    -> 对敌人造成 {actual} 伤害')

print()
````

## 8. 深入：为什么大型项目都用OOP？

### 8.1 OOP解决的核心问题

1. **代码组织**：类将相关的数据和行为打包在一起，代码结构清晰
2. **代码复用**：继承和组合避免了重复代码
3. **可维护性**：封装隐藏了实现细节，修改内部不影响外部
4. **可扩展性**：新增功能只需添加新类，不需修改现有代码
5. **团队协作**：不同开发者可以开发不同的类，通过接口交互

### 8.2 现实世界到对象的映射

| 现实世界 | OOP映射 |
|----------|--------|
| 一类事物 | 类 (Class) |
| 具体事物 | 对象 (Object) |
| 事物的特征 | 属性 (Attributes) |
| 事物的行为 | 方法 (Methods) |
| 事物的分类关系 | 继承 (Inheritance) |
| 事物包含其他事物 | 组合 (Composition) |

## 10. 本节小结 (Summary)

### 核心知识点

| 概念 | 英文 | 关键理解 |
|------|------|----------|
| 类 | Class | 对象的蓝图/模板 |
| 对象 | Object/Instance | 类的具体实例 |
| 属性 | Attributes | 对象的数据 |
| 方法 | Methods | 对象的行为 |
| 构造函数 | Constructor | 创建对象时自动调用 (__init__) |
| 析构函数 | Destructor | 对象销毁时自动调用 (__del__) |
| 封装 | Encapsulation | 隐藏内部数据，通过方法访问 |
| 继承 | Inheritance | 子类获得父类的属性和方法 |
| 多态 | Polymorphism | 同一方法在不同类中有不同实现 |
| 组合 | Composition | 强HAS-A关系 |
| 聚合 | Aggregation | 弱HAS-A关系 |
