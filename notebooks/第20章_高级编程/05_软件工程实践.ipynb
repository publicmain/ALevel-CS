## 一、软件开发生命周期模型 (Software Development Lifecycle Models)

### 1.1 为什么需要开发模型？

想象你要盖一栋大楼。你不会随便拿些砖头就开始砌墙——你需要一个**计划**：
先设计蓝图，再打地基，再建框架，再装修……

软件开发也是一样。**SDLC（Software Development Lifecycle）** 定义了软件从构思到退役的整个过程。

不同的项目适合不同的开发模型，就像不同类型的建筑需要不同的施工方案。

### 1.2 瀑布模型 (Waterfall Model)

瀑布模型是最经典、最直观的开发模型。它像瀑布一样，**从上到下**，每个阶段完成后才进入下一个阶段。

```
需求分析 (Requirements) 
    ↓
系统设计 (Design)
    ↓
编码实现 (Implementation)
    ↓
测试 (Testing)
    ↓
部署 (Deployment)
    ↓
维护 (Maintenance)
```

| 优点 | 缺点 |
|------|------|
| 结构清晰，易于管理 | 不灵活，难以应对需求变更 |
| 每阶段有明确的文档 | 用户要等很久才能看到成果 |
| 适合需求明确的项目 | 发现错误时修改成本高 |

**适用场景：** 需求非常明确且不太可能变化的项目，如航天系统、医疗设备软件。

**真实案例：** NASA 的航天飞机软件就是用类似瀑布的方法开发的——因为需求非常明确，且不允许出错。

````python
# ============================================
# 可视化：瀑布模型流程图
# ============================================
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis('off')
ax.set_title('瀑布模型 (Waterfall Model)', fontsize=18, fontweight='bold', pad=20)

stages = [
    ('需求分析\nRequirements', '#3498db'),
    ('系统设计\nDesign', '#2ecc71'),
    ('编码实现\nImplementation', '#e67e22'),
    ('测试\nTesting', '#e74c3c'),
    ('部署\nDeployment', '#9b59b6'),
    ('维护\nMaintenance', '#1abc9c'),
]

for i, (label, color) in enumerate(stages):
    x = 1 + i * 0.8
    y = 8.5 - i * 1.4
    rect = mpatches.FancyBboxPatch((x, y - 0.5), 3, 1, boxstyle="round,pad=0.1",
                                   facecolor=color, edgecolor='white', alpha=0.85)
    ax.add_patch(rect)
    ax.text(x + 1.5, y, label, ha='center', va='center',
            fontsize=11, fontweight='bold', color='white')
    if i < len(stages) - 1:
        ax.annotate('', xy=(1 + (i+1)*0.8 + 1.5, 8.5 - (i+1)*1.4 + 0.5),
                    xytext=(x + 1.5, y - 0.5),
                    arrowprops=dict(arrowstyle='->', color='#555', lw=2))

ax.text(8.5, 1, '特点: 线性、顺序、\n不可回溯', fontsize=11,
        ha='center', va='center',
        bbox=dict(boxstyle='round,pad=0.5', facecolor='#f0f0f0', edgecolor='#ccc'))

plt.tight_layout()
plt.show()
````

### 1.3 敏捷开发 / Scrum (Agile / Scrum)

敏捷开发（Agile）是对瀑布模型的**反思**。它的核心理念是：

> **「拥抱变化，快速迭代，持续交付」**

Agile 不是一种具体的方法，而是一组**价值观和原则**（参见 Agile Manifesto，敏捷宣言）。
Scrum 是 Agile 最流行的实践框架。

**Scrum 的核心概念：**

| 概念 | 英文 | 解释 |
|------|------|------|
| 冲刺 | Sprint | 一个固定时间的开发周期（通常2-4周） |
| 产品待办列表 | Product Backlog | 所有需要做的功能列表 |
| 冲刺待办列表 | Sprint Backlog | 本次冲刺要完成的任务 |
| 每日站会 | Daily Standup | 每天15分钟的简短会议 |
| 冲刺评审 | Sprint Review | 展示本次冲刺的成果 |
| 冲刺回顾 | Sprint Retrospective | 反思如何改进 |

| 优点 | 缺点 |
|------|------|
| 快速响应需求变化 | 需要高度团队协作 |
| 用户能尽早看到成果 | 文档可能不够充分 |
| 持续改进 | 范围容易蔓延（scope creep） |

**适用场景：** 需求不确定或经常变化的项目，如互联网产品、手机App。

**真实案例：** Spotify、微信等互联网产品都采用敏捷开发——因为用户需求在不断变化。

### 1.4 螺旋模型 (Spiral Model)

螺旋模型由 Barry Boehm 在1986年提出，它结合了瀑布模型的系统性和原型法的迭代性，
**特别强调风险分析（Risk Analysis）**。

每一圈螺旋包含四个阶段：
1. **确定目标** (Determine objectives)
2. **风险分析** (Risk analysis) ← 这是螺旋模型的核心特征
3. **开发和测试** (Development & testing)
4. **评审和计划** (Review & plan next iteration)

| 优点 | 缺点 |
|------|------|
| 强调风险管理 | 复杂、成本高 |
| 适合大型复杂项目 | 需要风险评估专家 |
| 灵活、可迭代 | 不适合小项目 |

**适用场景：** 高风险、大规模的项目，如银行系统、国防项目。

### 1.5 V-模型 (V-Model)

V-模型是瀑布模型的扩展，左边是开发阶段，右边是对应的**测试阶段**。
它强调：**每个开发阶段都有对应的测试阶段**。

```
需求分析  ←→  验收测试 (Acceptance Testing)
  系统设计  ←→  系统测试 (System Testing)
    详细设计  ←→  集成测试 (Integration Testing)
      编码  ←→  单元测试 (Unit Testing)
```

**关键思想：** 测试不是开发结束后才做的事情，而是**从一开始就要规划**。

### 1.6 快速应用开发 (RAD - Rapid Application Development)

RAD 强调**快速构建原型**，通过用户反馈不断完善。

四个阶段：需求规划 → 用户设计 → 快速构建 → 切换部署

| 优点 | 缺点 |
|------|------|
| 开发速度快 | 不适合大型系统 |
| 用户参与度高 | 需要充足的人力资源 |
| 灵活性强 | 可能牺牲代码质量 |

### 1.7 原型法 (Prototyping)

原型法是先快速构建一个**原型（Prototype）**，让用户体验并反馈，然后反复修改。

两种类型：
- **抛弃式原型（Throwaway Prototyping）**：原型只用于了解需求，最终会被丢弃
- **演化式原型（Evolutionary Prototyping）**：原型逐步演化为最终产品

````python
# ============================================
# 可视化：SDLC 模型对比图
# ============================================
fig, ax = plt.subplots(figsize=(14, 7))

models = ['Waterfall\n瀑布', 'Agile/Scrum\n敏捷', 'Spiral\n螺旋',
          'V-Model\nV模型', 'RAD\n快速开发', 'Prototyping\n原型法']

# 各模型在不同维度上的评分（1-5）
categories = ['灵活性', '文档完整性', '风险管理', '用户参与', '适合大项目', '开发速度']
scores = {
    'Waterfall': [1, 5, 2, 1, 3, 2],
    'Agile':     [5, 2, 3, 5, 3, 5],
    'Spiral':    [4, 4, 5, 3, 5, 2],
    'V-Model':   [2, 5, 3, 2, 4, 2],
    'RAD':       [4, 2, 2, 4, 2, 5],
    'Prototype': [4, 2, 2, 5, 2, 4],
}

x = np.arange(len(categories))
width = 0.13
colors = ['#3498db', '#2ecc71', '#e74c3c', '#9b59b6', '#e67e22', '#1abc9c']
names = list(scores.keys())

for i, (name, vals) in enumerate(scores.items()):
    offset = (i - len(scores)/2 + 0.5) * width
    bars = ax.bar(x + offset, vals, width, label=f'{models[i]}',
                  color=colors[i], alpha=0.85, edgecolor='white')

ax.set_xlabel('评估维度', fontsize=13)
ax.set_ylabel('评分 (1-5)', fontsize=13)
ax.set_title('SDLC 模型多维度对比', fontsize=16, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=11)
ax.set_ylim(0, 6)
ax.legend(loc='upper right', fontsize=9, ncol=2)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\n📊 关键结论:")
print("• 没有'最好'的模型，只有'最适合'的模型")
print("• 瀑布模型适合需求明确的项目")
print("• 敏捷适合需求变化快的项目")
print("• 螺旋模型适合高风险大项目")
````

## 二、版本控制 (Version Control)

### 2.1 为什么需要版本控制？

你有没有这样的经历？

```
报告_v1.docx
报告_v2.docx
报告_v2_修改.docx
报告_v2_修改_最终版.docx
报告_v2_修改_最终版_真的最终版.docx
```

这就是**没有版本控制**的痛苦！在软件开发中，代码文件可能有成百上千个，
多个开发者同时修改——如果没有版本控制，后果不堪设想。

**版本控制系统（Version Control System, VCS）** 帮我们：

1. **记录每次变更的历史** - 谁在什么时候改了什么
2. **支持多人协作** - 不同人可以同时修改不同部分
3. **允许回退** - 发现问题可以恢复到之前的版本
4. **支持分支开发** - 可以在不影响主线的情况下开发新功能

### 2.2 Git 基础概念

**Git** 是目前最流行的分布式版本控制系统，由 Linux 之父 Linus Torvalds 创建。

| 概念 | 英文 | 解释 |
|------|------|------|
| 仓库 | Repository (repo) | 存放代码和版本历史的地方 |
| 提交 | Commit | 保存一次代码快照，附带说明信息 |
| 分支 | Branch | 从主线分出的独立开发线 |
| 合并 | Merge | 将分支的修改合并回主线 |
| 暂存区 | Staging Area | 准备提交的文件的临时区域 |
| 克隆 | Clone | 复制远程仓库到本地 |
| 推送/拉取 | Push / Pull | 将本地修改推送到远程 / 从远程获取更新 |

### 2.3 Git 的工作流程

```
工作区 (Working Directory)
    ↓  git add
暂存区 (Staging Area)
    ↓  git commit
本地仓库 (Local Repository)
    ↓  git push
远程仓库 (Remote Repository)
```

````python
# ============================================
# 演示：模拟 Git 版本控制的基本操作
# ============================================

class SimpleGit:
    """一个简化的 Git 模拟器，帮助理解版本控制概念"""

    def __init__(self, repo_name):
        self.repo_name = repo_name
        self.commits = []          # 提交历史
        self.staging = []           # 暂存区
        self.branches = {'main': []}  # 分支
        self.current_branch = 'main'
        print(f"初始化仓库: {repo_name}")
        print(f"当前分支: {self.current_branch}")

    def add(self, filename):
        """git add - 将文件添加到暂存区"""
        self.staging.append(filename)
        print(f"  [git add] '{filename}' 已添加到暂存区")

    def commit(self, message):
        """git commit - 提交暂存区的更改"""
        if not self.staging:
            print("  [error] 暂存区为空，没有可提交的文件")
            return
        commit_id = f"c{len(self.commits) + 1:03d}"
        commit_info = {
            'id': commit_id,
            'message': message,
            'files': self.staging.copy(),
            'branch': self.current_branch
        }
        self.commits.append(commit_info)
        self.branches[self.current_branch].append(commit_id)
        self.staging = []
        print(f"  [git commit] {commit_id}: '{message}'")
        print(f"    包含文件: {commit_info['files']}")

    def branch(self, branch_name):
        """git branch - 创建新分支"""
        self.branches[branch_name] = self.branches[self.current_branch].copy()
        print(f"  [git branch] 创建分支 '{branch_name}'")

    def checkout(self, branch_name):
        """git checkout - 切换分支"""
        if branch_name in self.branches:
            self.current_branch = branch_name
            print(f"  [git checkout] 切换到分支 '{branch_name}'")
        else:
            print(f"  [error] 分支 '{branch_name}' 不存在")

    def merge(self, branch_name):
        """git merge - 合并分支"""
        if branch_name in self.branches:
            new_commits = [c for c in self.branches[branch_name]
                          if c not in self.branches[self.current_branch]]
            self.branches[self.current_branch].extend(new_commits)
            print(f"  [git merge] 将 '{branch_name}' 合并到 '{self.current_branch}'")
            print(f"    合并了 {len(new_commits)} 个新提交")
        else:
            print(f"  [error] 分支 '{branch_name}' 不存在")

    def log(self):
        """git log - 显示提交历史"""
        print(f"\n--- 提交历史 (分支: {self.current_branch}) ---")
        branch_commits = self.branches[self.current_branch]
        for commit in self.commits:
            if commit['id'] in branch_commits:
                print(f"  {commit['id']} | {commit['branch']:10s} | {commit['message']}")
        print("---")


# 实际演示
print("=" * 50)
print("模拟 Git 工作流程")
print("=" * 50)

repo = SimpleGit("my_project")
print()

# 在 main 分支上进行初始开发
repo.add("main.py")
repo.add("utils.py")
repo.commit("初始提交：添加主程序和工具模块")
print()

# 创建功能分支
repo.branch("feature/login")
repo.checkout("feature/login")
print()

# 在功能分支上开发
repo.add("login.py")
repo.commit("添加登录功能")
repo.add("auth.py")
repo.commit("添加身份验证模块")
print()

# 切回 main 分支并合并
repo.checkout("main")
repo.merge("feature/login")

# 查看历史
repo.log()
````

## 三、代码文档 (Code Documentation)

### 3.1 为什么代码文档很重要？

> **「代码是写给人读的，顺便让机器执行。」** —— Harold Abelson

好的文档可以：
- 帮助**其他开发者**理解你的代码
- 帮助**未来的你**回忆当初的设计思路
- 帮助**用户**学会使用你的代码

### 3.2 文档的三个层次

| 层次 | 英文 | 说明 | 读者 |
|------|------|------|------|
| 行内注释 | Inline Comments | 解释复杂的代码逻辑 | 开发者 |
| 文档字符串 | Docstrings | 描述函数/类的用途和接口 | 开发者和使用者 |
| API 文档 | API Documentation | 完整的接口说明和使用指南 | 使用者 |

````python
# ============================================
# 代码文档示范
# ============================================

# --- 1. 行内注释 (Inline Comments) ---
# 好的注释解释 WHY（为什么），而不是 WHAT（做什么）

# 不好的注释（解释了显而易见的事情）:
x = x + 1  # x 加 1

# 好的注释（解释了为什么这样做）:
x = x + 1  # 补偿数组从0开始的偏移量

print("示范1: 行内注释 - 解释 WHY 而不是 WHAT")
print()


# --- 2. 文档字符串 (Docstrings) ---
def calculate_bmi(weight_kg: float, height_m: float) -> float:
    """
    计算身体质量指数 (Body Mass Index)。

    BMI 是国际通用的衡量人体胖瘦程度的标准。
    计算公式: BMI = 体重(kg) / 身高(m)²

    Args:
        weight_kg (float): 体重，单位为千克，必须为正数
        height_m (float): 身高，单位为米，必须为正数

    Returns:
        float: BMI 值，保留一位小数

    Raises:
        ValueError: 如果体重或身高不是正数

    Examples:
        >>> calculate_bmi(70, 1.75)
        22.9
        >>> calculate_bmi(55, 1.60)
        21.5
    """
    if weight_kg <= 0 or height_m <= 0:
        raise ValueError("体重和身高必须为正数")
    return round(weight_kg / (height_m ** 2), 1)


# 展示 docstring 的使用
print("示范2: 文档字符串 (Docstring)")
print("---")
print(f"函数名: {calculate_bmi.__name__}")
print(f"文档:\n{calculate_bmi.__doc__}")
print(f"\n调用结果: BMI = {calculate_bmi(70, 1.75)}")
````

````python
# ============================================
# 互动演示：简单的文档生成器
# ============================================
import inspect

def generate_doc(func):
    """自动从函数生成文档摘要"""
    doc = inspect.getdoc(func) or "(无文档)"
    sig = inspect.signature(func)
    source_lines = inspect.getsource(func).count('\n')

    print(f"{'=' * 50}")
    print(f"函数文档报告")
    print(f"{'=' * 50}")
    print(f"函数名: {func.__name__}")
    print(f"参数: {sig}")
    print(f"代码行数: ~{source_lines}")
    print(f"有文档: {'是' if func.__doc__ else '否'}")
    print(f"\n文档内容:")
    print(f"  {doc[:200]}..." if len(doc) > 200 else f"  {doc}")
    print(f"{'=' * 50}")


# 一个有文档的函数
def binary_search(arr: list, target: int) -> int:
    """
    在已排序数组中使用二分查找定位目标元素。

    时间复杂度: O(log n)
    空间复杂度: O(1)

    Args:
        arr: 已升序排列的整数列表
        target: 要查找的目标值

    Returns:
        目标元素的索引，如果不存在则返回 -1
    """
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1


# 一个没有文档的函数（反面教材）
def f(a, b):
    return a * b + a - b


print("有文档的函数:")
generate_doc(binary_search)
print()
print("没有文档的函数（反面教材）:")
generate_doc(f)
print()
print("对比之下，哪个更容易理解？")
print("好的文档是给未来的自己最好的礼物！")
````

## 四、测试方法论 (Testing Methodologies)

### 4.1 为什么测试如此重要？

软件测试的目的不是**证明代码没有bug**，而是**尽可能找出bug**。

> **「程序测试能展示 bug 的存在，但不能证明 bug 的不存在。」** —— Edsger Dijkstra

不做测试的后果可能非常严重：
- **1996年 Ariane 5 火箭爆炸**：一个整数溢出 bug 导致价值3.7亿美元的火箭在发射后39秒爆炸
- **2012年 Knight Capital 事件**：一个软件部署错误导致公司在45分钟内损失4.4亿美元

### 4.2 测试金字塔 (Testing Pyramid)

```
        /  \          ← 端到端测试 (E2E) - 少量但关键
       /    \
      /      \        ← 集成测试 (Integration) - 适量
     /        \
    /          \      ← 单元测试 (Unit) - 大量
   /____________\
```

| 测试类型 | 英文 | 测试什么 | 数量 | 速度 |
|----------|------|----------|------|------|
| 单元测试 | Unit Testing | 单个函数/方法 | 最多 | 最快 |
| 集成测试 | Integration Testing | 模块间的交互 | 适中 | 中等 |

````python
# ============================================
# 单元测试示范：使用 unittest 模块
# ============================================
import unittest

# --- 被测试的代码 ---
class Calculator:
    """一个简单的计算器类"""

    def add(self, a, b):
        """加法"""
        return a + b

    def subtract(self, a, b):
        """减法"""
        return a - b

    def multiply(self, a, b):
        """乘法"""
        return a * b

    def divide(self, a, b):
        """除法，除数不能为0"""
        if b == 0:
            raise ValueError("除数不能为零")
        return a / b


# --- 测试代码 ---
class TestCalculator(unittest.TestCase):
    """Calculator 类的单元测试"""

    def setUp(self):
        """每个测试方法执行前都会调用，用于初始化"""
        self.calc = Calculator()

    def test_add(self):
        """测试加法"""
        self.assertEqual(self.calc.add(2, 3), 5)
        self.assertEqual(self.calc.add(-1, 1), 0)
        self.assertEqual(self.calc.add(0, 0), 0)

    def test_subtract(self):
        """测试减法"""
        self.assertEqual(self.calc.subtract(5, 3), 2)
        self.assertEqual(self.calc.subtract(3, 5), -2)

    def test_multiply(self):
        """测试乘法"""
        self.assertEqual(self.calc.multiply(3, 4), 12)
        self.assertEqual(self.calc.multiply(0, 100), 0)
        self.assertEqual(self.calc.multiply(-2, 3), -6)

    def test_divide(self):
        """测试除法"""
        self.assertEqual(self.calc.divide(10, 2), 5)
        self.assertAlmostEqual(self.calc.divide(1, 3), 0.3333, places=3)

    def test_divide_by_zero(self):
        """测试除以零应该抛出异常"""
        with self.assertRaises(ValueError):
            self.calc.divide(10, 0)


# 在 Jupyter 中运行测试
print("运行单元测试...")
print("=" * 50)

# 创建测试套件并运行
suite = unittest.TestLoader().loadTestsFromTestCase(TestCalculator)
runner = unittest.TextTestRunner(verbosity=2)
result = runner.run(suite)

print(f"\n测试总结:")
print(f"  运行: {result.testsRun} 个测试")
print(f"  成功: {result.testsRun - len(result.failures) - len(result.errors)}")
print(f"  失败: {len(result.failures)}")
print(f"  错误: {len(result.errors)}")
````

````python
# ============================================
# pytest 风格的测试（更简洁的语法）
# ============================================

# pytest 使用普通的 assert 语句，更加直观
# 注意：在 Jupyter 中我们手动模拟 pytest 的风格

def is_palindrome(s: str) -> bool:
    """检查字符串是否为回文（忽略大小写和空格）"""
    cleaned = s.lower().replace(' ', '')
    return cleaned == cleaned[::-1]


# pytest 风格的测试函数
def test_is_palindrome():
    """测试回文检测函数"""
    test_cases = [
        # (输入, 期望结果, 说明)
        ("racecar", True, "标准回文"),
        ("hello", False, "非回文"),
        ("A man a plan a canal Panama", True, "带空格的回文"),
        ("", True, "空字符串"),
        ("a", True, "单字符"),
        ("Race Car", True, "大小写混合的回文"),
    ]

    passed = 0
    failed = 0

    for input_str, expected, description in test_cases:
        result = is_palindrome(input_str)
        status = "PASS" if result == expected else "FAIL"
        if status == "PASS":
            passed += 1
        else:
            failed += 1
        symbol = "[OK]" if status == "PASS" else "[X]"
        print(f"  {symbol} {description}: is_palindrome('{input_str}') = {result}")

    print(f"\n结果: {passed} 通过, {failed} 失败")
    return failed == 0


print("pytest 风格的测试演示")
print("=" * 50)
all_passed = test_is_palindrome()
print(f"\n所有测试{'通过' if all_passed else '未全部通过'}!")
````

### 4.4 测试驱动开发 (TDD - Test-Driven Development)

TDD 的核心理念是：**先写测试，再写代码**。

这听起来很奇怪——代码都还没写，怎么写测试？但这恰恰是 TDD 的精妙之处：
**测试定义了代码应该做什么**，相当于一个精确的「规格说明」。

**TDD 的三个步骤（红-绿-重构循环）：**

1. **红色 (Red)**：写一个会失败的测试 —— 因为功能还没实现
2. **绿色 (Green)**：写最少的代码让测试通过
3. **重构 (Refactor)**：改善代码质量，同时确保测试仍然通过

```
Red → Green → Refactor → Red → Green → Refactor → ...
```

**为什么 TDD 有效？**
- 强迫你在写代码之前**思考清楚**需求
- 自动产生**完整的测试覆盖**
- 每次修改后都能**快速验证**是否破坏了已有功能
- 代码天然具有**可测试性**（因为是为测试而写的）

````python
# ============================================
# TDD 实战演示：开发一个密码强度检测器
# ============================================

print("TDD 实战：开发密码强度检测器")
print("=" * 50)

# === 第一步 (Red): 先写测试 ===
print("\n--- 第一步: 编写测试（此时函数还不存在）---")

def run_password_tests(check_func):
    """密码强度检测器的测试套件"""
    tests = [
        # (密码, 期望强度, 说明)
        ("123", "弱", "太短的密码"),
        ("abcdefgh", "弱", "只有小写字母"),
        ("Abcdef1", "中", "有大小写和数字但较短"),
        ("Abcdef1!", "强", "有大小写、数字和特殊字符"),
        ("MyP@ssw0rd123", "强", "长密码包含所有类型"),
        ("", "弱", "空密码"),
    ]

    passed = 0
    for password, expected, desc in tests:
        result = check_func(password)
        status = "PASS" if result == expected else "FAIL"
        if status == "PASS":
            passed += 1
        symbol = "[OK]" if status == "PASS" else "[X]"
        print(f"  {symbol} {desc}: '{password}' -> 期望'{expected}', 实际'{result}'")

    return passed, len(tests)


# === 第二步 (Green): 实现最少的代码让测试通过 ===
print("\n--- 第二步: 实现功能 ---")

def check_password_strength(password: str) -> str:
    """
    检查密码强度。

    规则:
    - 弱: 长度<6 或 只包含一种字符类型
    - 中: 长度>=6 且 包含2-3种字符类型
    - 强: 长度>=8 且 包含4种字符类型（大写、小写、数字、特殊字符）
    """
    if len(password) < 6:
        return "弱"

    has_upper = any(c.isupper() for c in password)
    has_lower = any(c.islower() for c in password)
    has_digit = any(c.isdigit() for c in password)
    has_special = any(not c.isalnum() for c in password)

    types_count = sum([has_upper, has_lower, has_digit, has_special])

    if types_count >= 4 and len(password) >= 8:
        return "强"
    elif types_count >= 2:
        return "中"
    else:
        return "弱"


# === 第三步: 运行测试 ===
print("\n--- 第三步: 运行测试 ---")
passed, total = run_password_tests(check_password_strength)
print(f"\n结果: {passed}/{total} 通过")

if passed == total:
    print("所有测试通过！TDD 循环完成。")
    print("\n下一步将是 Refactor（重构）—— 改善代码质量而不改变行为。")
````

## 五、代码重构 (Code Refactoring)

### 5.1 什么是重构？

**重构（Refactoring）** 是指在**不改变代码外部行为**的前提下，改善代码的内部结构。

类比：重新装修房子——房子的功能不变（住人），但内部更整洁、更高效。

### 5.2 为什么要重构？

- **消除重复代码** (Eliminate duplication)
- **提高可读性** (Improve readability)
- **降低复杂度** (Reduce complexity)
- **便于后续修改** (Easier to modify)

### 5.3 常见的重构技术

| 技术 | 英文 | 说明 |
|------|------|------|
| 提取方法 | Extract Method | 把一段代码抽取为独立函数 |
| 重命名 | Rename | 给变量/函数取更有意义的名字 |
| 简化条件 | Simplify Conditionals | 简化复杂的 if/else 逻辑 |
| 移除魔法数字 | Remove Magic Numbers | 用命名常量替代硬编码的数字 |
| 内联临时变量 | Inline Temp | 去掉不必要的临时变量 |

````python
# ============================================
# 重构示例：从糟糕的代码到优雅的代码
# ============================================

print("重构示例：计算学生成绩等级")
print("=" * 60)

# ---- 重构前：充满问题的代码 ----
print("\n【重构前】—— 有什么问题？")
print("-" * 40)

before_code = '''
def calc(n, s):
    t = 0
    for i in s:
        t = t + i
    a = t / len(s)
    if a >= 90:
        g = "A*"
    elif a >= 80:
        g = "A"
    elif a >= 70:
        g = "B"
    elif a >= 60:
        g = "C"
    elif a >= 50:
        g = "D"
    else:
        g = "U"
    print(n + " got " + g + " with average " + str(a))
    return g
'''
print(before_code)

print("问题分析:")
print("  1. 变量名无意义 (n, s, t, a, g)")
print("  2. 函数做了太多事情（计算+判断+打印）")
print("  3. 魔法数字 (90, 80, 70...)")
print("  4. 没有文档和类型提示")


# ---- 重构后：清晰优雅的代码 ----
print("\n\n【重构后】—— 应用多种重构技术")
print("-" * 40)

after_code = '''
# 提取常量（移除魔法数字）
GRADE_BOUNDARIES = [
    (90, "A*"),
    (80, "A"),
    (70, "B"),
    (60, "C"),
    (50, "D"),
]
DEFAULT_GRADE = "U"

def calculate_average(scores: list[float]) -> float:
    """计算平均分"""
    return sum(scores) / len(scores)

def determine_grade(average: float) -> str:
    """根据平均分确定等级"""
    for boundary, grade in GRADE_BOUNDARIES:
        if average >= boundary:
            return grade
    return DEFAULT_GRADE

def get_student_result(student_name: str,
                       scores: list[float]) -> dict:
    """获取学生成绩结果"""
    average = calculate_average(scores)
    grade = determine_grade(average)
    return {
        "name": student_name,
        "average": round(average, 1),
        "grade": grade
    }
'''
print(after_code)

print("改进总结:")
print("  1. [重命名] 变量名有意义 (student_name, scores, average)")
print("  2. [提取方法] 一个函数只做一件事")
print("  3. [移除魔法数字] 使用命名常量 GRADE_BOUNDARIES")
print("  4. [添加文档] 类型提示 + docstring")
print("  5. [简化条件] 用循环替代冗长的 if-elif")
````

````python
# ============================================
# 重构后的代码实际运行
# ============================================

GRADE_BOUNDARIES = [
    (90, "A*"),
    (80, "A"),
    (70, "B"),
    (60, "C"),
    (50, "D"),
]
DEFAULT_GRADE = "U"

def calculate_average(scores: list) -> float:
    """计算平均分"""
    if not scores:
        return 0.0
    return sum(scores) / len(scores)

def determine_grade(average: float) -> str:
    """根据平均分确定等级"""
    for boundary, grade in GRADE_BOUNDARIES:
        if average >= boundary:
            return grade
    return DEFAULT_GRADE

def get_student_result(student_name: str, scores: list) -> dict:
    """获取学生成绩结果"""
    average = calculate_average(scores)
    grade = determine_grade(average)
    return {
        "name": student_name,
        "average": round(average, 1),
        "grade": grade
    }


# 测试重构后的代码
students = [
    ("小明", [95, 88, 92, 87]),
    ("小红", [78, 82, 75, 80]),
    ("小华", [65, 58, 62, 55]),
    ("小李", [45, 50, 42, 48]),
]

print("学生成绩报告")
print("=" * 45)
print(f"{'姓名':<8} {'平均分':<10} {'等级':<6}")
print("-" * 45)

for name, scores in students:
    result = get_student_result(name, scores)
    print(f"{result['name']:<8} {result['average']:<10} {result['grade']:<6}")
````

## 六、软件度量 (Software Metrics)

### 6.1 为什么需要软件度量？

> **「你无法管理你无法衡量的东西。」** —— Peter Drucker

软件度量帮助我们用**客观数据**而不是主观感觉来评估代码质量。

### 6.2 常见的软件度量指标

| 指标 | 英文 | 衡量什么 |
|------|------|----------|
| 代码行数 | Lines of Code (LOC) | 代码量（粗略度量） |
| 圈复杂度 | Cyclomatic Complexity | 代码逻辑的复杂程度 |
| 代码覆盖率 | Code Coverage | 测试覆盖了多少代码 |
| 注释比率 | Comment Ratio | 注释占代码的比例 |
| 耦合度 | Coupling | 模块之间的依赖程度 |

### 6.3 圈复杂度 (Cyclomatic Complexity)

圈复杂度是由 Thomas McCabe 在1976年提出的，它衡量代码中**独立路径的数量**。

简单计算方法：**从1开始，每遇到一个判断/分支（if, elif, while, for, and, or），复杂度 +1**

| 复杂度 | 风险等级 | 建议 |
|--------|----------|------|
| 1-10 | 低风险 | 简单，容易测试 |
| 11-20 | 中等风险 | 较复杂，需要更多测试 |
| 21-50 | 高风险 | 需要重构 |
| >50 | 极高风险 | 必须重构 |

````python
# ============================================
# 互动演示：圈复杂度计算器
# ============================================
import ast
import textwrap

def calculate_cyclomatic_complexity(source_code: str) -> int:
    """
    计算 Python 函数的圈复杂度。

    规则: 从1开始，每个决策点（if, elif, for, while, and, or,
          except, with, assert, 列表推导中的 if）加1。
    """
    tree = ast.parse(textwrap.dedent(source_code))
    complexity = 1  # 基础复杂度

    for node in ast.walk(tree):
        if isinstance(node, (ast.If, ast.While, ast.For)):
            complexity += 1
        elif isinstance(node, ast.ExceptHandler):
            complexity += 1
        elif isinstance(node, ast.BoolOp):
            # and/or 运算符：每个额外操作数加1
            complexity += len(node.values) - 1
        elif isinstance(node, ast.Assert):
            complexity += 1

    return complexity


# 测试不同复杂度的函数
examples = [
    ("简单函数（无分支）", """
def greet(name):
    return f"Hello, {name}"
"""),
    ("中等复杂度", """
def check_age(age):
    if age < 0:
        return "无效"
    elif age < 18:
        return "未成年"
    elif age < 65:
        return "成年"
    else:
        return "老年"
"""),
    ("较高复杂度", """
def validate_input(data):
    if not data:
        return False
    if len(data) < 3 or len(data) > 100:
        return False
    for char in data:
        if char.isdigit() and int(char) > 5:
            return False
        elif not char.isalnum():
            return False
    return True
"""),
]

print("圈复杂度分析")
print("=" * 50)

for name, code in examples:
    cc = calculate_cyclomatic_complexity(code)
    if cc <= 10:
        risk = "低风险 (简单)"
    elif cc <= 20:
        risk = "中等风险"
    else:
        risk = "高风险 (需重构)"
    print(f"\n{name}:")
    print(f"  圈复杂度 = {cc}")
    print(f"  风险等级: {risk}")
````

## 七、设计原则 (Design Principles)

### 7.1 DRY - Don't Repeat Yourself（不要重复自己）

**核心思想：** 每一条知识/逻辑在系统中应该只有**一个明确的、权威的表示**。

如果相同的代码出现在两个地方，将来修改时你可能只改了一处而忘记另一处——这就会产生 bug。

```python
# 违反 DRY:
def calculate_circle_area(radius):
    return 3.14159 * radius * radius

def calculate_circle_perimeter(radius):
    return 2 * 3.14159 * radius  # 3.14159 重复了！

# 遵守 DRY:
import math
def calculate_circle_area(radius):
    return math.pi * radius ** 2

def calculate_circle_perimeter(radius):
    return 2 * math.pi * radius
```

### 7.2 KISS - Keep It Simple, Stupid（保持简单）

**核心思想：** 代码应该尽可能简单。复杂的解决方案不等于好的解决方案。

```python
# 过于复杂:
def is_even(n):
    return not bool(n - (n // 2) * 2)

# KISS:
def is_even(n):
    return n % 2 == 0
```

### 7.3 SOLID 原则（简介）

SOLID 是面向对象设计的五大原则，由 Robert C. Martin（Uncle Bob）提出：

| 字母 | 原则 | 英文 | 核心含义 |
|------|------|------|----------|
| **S** | 单一职责 | Single Responsibility | 一个类只做一件事 |
| **O** | 开放封闭 | Open/Closed | 对扩展开放，对修改封闭 |
| **L** | 里氏替换 | Liskov Substitution | 子类可以替换父类 |
| **I** | 接口隔离 | Interface Segregation | 不要强迫使用不需要的接口 |
| **D** | 依赖反转 | Dependency Inversion | 依赖抽象而非具体实现 |

这些原则在 A-Level 考试中通常只需要**了解概念**，不需要深入细节。

````python
# ============================================
# 设计原则示范：DRY 和 KISS
# ============================================

print("设计原则实战演示")
print("=" * 50)

# --- DRY 示例 ---
print("\n【DRY 原则】消除重复")
print("-" * 40)

# 违反 DRY（重复的验证逻辑）
print("\n违反 DRY:")
print('''  def register_student(name, age):
      if not name or len(name) < 2:
          raise ValueError("名字无效")
      if age < 0 or age > 150:
          raise ValueError("年龄无效")
      # ... 注册逻辑

  def update_student(name, age):
      if not name or len(name) < 2:     # 重复！
          raise ValueError("名字无效")
      if age < 0 or age > 150:           # 重复！
          raise ValueError("年龄无效")
      # ... 更新逻辑''')

# 遵守 DRY
print("\n遵守 DRY:")

def validate_student_data(name: str, age: int) -> bool:
    """统一的学生数据验证（只写一次）"""
    if not name or len(name) < 2:
        raise ValueError("名字无效")
    if age < 0 or age > 150:
        raise ValueError("年龄无效")
    return True

def register_student(name: str, age: int):
    validate_student_data(name, age)  # 复用验证逻辑
    return f"注册成功: {name}, {age}岁"

def update_student(name: str, age: int):
    validate_student_data(name, age)  # 复用验证逻辑
    return f"更新成功: {name}, {age}岁"

print(f"  {register_student('小明', 17)}")
print(f"  {update_student('小红', 18)}")

# --- KISS 示例 ---
print("\n\n【KISS 原则】保持简单")
print("-" * 40)

# 过度复杂
def get_max_complex(lst):
    """过于复杂的求最大值"""
    result = lst[0]
    for i in range(1, len(lst)):
        if lst[i] > result:
            result = lst[i]
    return result

# KISS 方式
def get_max_simple(lst):
    """简单直接的求最大值"""
    return max(lst)

test_list = [3, 7, 1, 9, 4, 6]
print(f"复杂方式: {get_max_complex(test_list)}")
print(f"KISS方式: {get_max_simple(test_list)}")
print("两种方式结果相同，但KISS版本更清晰、更不容易出错！")

# --- 单一职责原则 ---
print("\n\n【单一职责原则 (SRP)】")
print("-" * 40)
print("一个类/函数应该只有一个改变的理由。")
print("")
print("违反 SRP: 一个函数负责 读取数据 + 处理数据 + 保存结果")
print("遵守 SRP: 拆分为 read_data() + process_data() + save_results()")
````

## 八、黑盒测试 vs 白盒测试（详细演示）

### 8.1 黑盒测试 (Black Box Testing)

黑盒测试只关注**输入和输出**，不关心代码内部如何实现。

常用的黑盒测试技术：

- **等价划分（Equivalence Partitioning）**：将输入分成若干等价类，每类取一个代表测试
- **边界值分析（Boundary Value Analysis）**：测试边界值，因为边界处最容易出错

### 8.2 白盒测试 (White Box Testing)

白盒测试关注代码的**内部逻辑**，确保每条路径都被测试到。

常用的白盒测试技术：

- **语句覆盖（Statement Coverage）**：每条语句至少执行一次
- **分支覆盖（Branch Coverage）**：每个分支都被测试
- **路径覆盖（Path Coverage）**：所有可能的执行路径都被测试

````python
# ============================================
# 黑盒测试 vs 白盒测试 完整演示
# ============================================

def classify_triangle(a: float, b: float, c: float) -> str:
    """
    根据三条边的长度判断三角形类型。

    Returns:
        "不是三角形", "等边三角形", "等腰三角形", "普通三角形"
    """
    # 检查是否能构成三角形
    if a <= 0 or b <= 0 or c <= 0:
        return "不是三角形"
    if a + b <= c or a + c <= b or b + c <= a:
        return "不是三角形"

    # 判断类型
    if a == b == c:
        return "等边三角形"
    elif a == b or b == c or a == c:
        return "等腰三角形"
    else:
        return "普通三角形"


print("黑盒测试 (Black Box Testing)")
print("=" * 60)
print("不看代码实现，只根据需求设计测试用例")
print()

# 等价划分 + 边界值分析
black_box_tests = [
    # 等价类: 有效等边三角形
    ((3, 3, 3), "等边三角形", "等价类: 等边三角形"),
    # 等价类: 有效等腰三角形
    ((3, 3, 4), "等腰三角形", "等价类: 等腰三角形"),
    ((5, 3, 5), "等腰三角形", "等价类: 等腰三角形(不同位置)"),
    # 等价类: 普通三角形
    ((3, 4, 5), "普通三角形", "等价类: 普通三角形"),
    # 边界值: 刚好不能构成三角形
    ((1, 2, 3), "不是三角形", "边界值: a+b=c"),
    # 边界值: 零和负数
    ((0, 1, 1), "不是三角形", "边界值: 边长为0"),
    ((-1, 2, 3), "不是三角形", "边界值: 负数边长"),
    # 边界值: 很大的数
    ((1000000, 1000000, 1000000), "等边三角形", "边界值: 大数"),
]

for (a, b, c), expected, desc in black_box_tests:
    result = classify_triangle(a, b, c)
    status = "[OK]" if result == expected else "[X]"
    print(f"  {status} {desc}: ({a},{b},{c}) -> {result}")

print("\n")
print("白盒测试 (White Box Testing)")
print("=" * 60)
print("看代码实现，确保每个分支都被覆盖")
print()

white_box_tests = [
    # 分支1: a <= 0
    ((-1, 3, 3), "不是三角形", "分支: a<=0"),
    # 分支2: a + b <= c
    ((1, 1, 5), "不是三角形", "分支: a+b<=c"),
    # 分支3: a + c <= b
    ((1, 5, 1), "不是三角形", "分支: a+c<=b"),
    # 分支4: b + c <= a
    ((5, 1, 1), "不是三角形", "分支: b+c<=a"),
    # 分支5: a == b == c
    ((4, 4, 4), "等边三角形", "分支: a==b==c"),
    # 分支6: a == b (but not c)
    ((4, 4, 3), "等腰三角形", "分支: a==b"),
    # 分支7: b == c (but not a)
    ((3, 4, 4), "等腰三角形", "分支: b==c"),
    # 分支8: a == c (but not b)
    ((4, 3, 4), "等腰三角形", "分支: a==c"),
    # 分支9: else (普通三角形)
    ((3, 4, 5), "普通三角形", "分支: else"),
]

for (a, b, c), expected, desc in white_box_tests:
    result = classify_triangle(a, b, c)
    status = "[OK]" if result == expected else "[X]"
    print(f"  {status} {desc}: ({a},{b},{c}) -> {result}")

print("\n关键区别:")
print("  黑盒: 从用户角度出发，按等价类和边界值设计")
print("  白盒: 从代码角度出发，确保每个分支都被执行")
````

````python
# ============================================
# 互动演示：简易测试运行器 (Test Runner)
# ============================================

class SimpleTestRunner:
    """一个简易的测试框架，展示测试运行器的基本原理"""

    def __init__(self, name="测试套件"):
        self.name = name
        self.tests = []
        self.results = {'passed': 0, 'failed': 0, 'errors': 0}

    def add_test(self, test_name, test_func):
        """添加一个测试"""
        self.tests.append((test_name, test_func))

    def run(self):
        """运行所有测试"""
        print(f"\n{'=' * 50}")
        print(f"运行: {self.name} ({len(self.tests)} 个测试)")
        print(f"{'=' * 50}")

        for test_name, test_func in self.tests:
            try:
                test_func()
                self.results['passed'] += 1
                print(f"  [PASS] {test_name}")
            except AssertionError as e:
                self.results['failed'] += 1
                print(f"  [FAIL] {test_name}: {e}")
            except AssertionError:
                self.results['failed'] += 1
                print(f"  [FAIL] {test_name}")
            except Exception as e:
                self.results['errors'] += 1
                print(f"  [ERR]  {test_name}: {type(e).__name__}: {e}")

        self._print_summary()

    def _print_summary(self):
        """打印测试总结"""
        total = sum(self.results.values())
        print(f"\n--- 测试总结 ---")
        print(f"  总计: {total}")
        print(f"  通过: {self.results['passed']}")
        print(f"  失败: {self.results['failed']}")
        print(f"  错误: {self.results['errors']}")
        rate = (self.results['passed'] / total * 100) if total > 0 else 0
        print(f"  通过率: {rate:.1f}%")


# 被测试的函数
def fibonacci(n):
    """返回第n个斐波那契数 (从0开始)"""
    if n < 0:
        raise ValueError("n 不能为负数")
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b


# 创建测试
runner = SimpleTestRunner("Fibonacci 函数测试")

runner.add_test("fib(0) = 0", lambda: None if fibonacci(0) == 0 else (_ for _ in ()).throw(AssertionError))
runner.add_test("fib(1) = 1", lambda: None if fibonacci(1) == 1 else (_ for _ in ()).throw(Exception("expected 1")))
runner.add_test("fib(5) = 5", lambda: None if fibonacci(5) == 5 else (_ for _ in ()).throw(Exception("expected 5")))
runner.add_test("fib(10) = 55", lambda: None if fibonacci(10) == 55 else (_ for _ in ()).throw(Exception("expected 55")))
runner.add_test("fib(-1) raises ValueError",
                lambda: (_ for _ in ()).throw(Exception("Should have raised")) if True else None)

# 更简洁的写法
runner2 = SimpleTestRunner("Fibonacci 函数测试 (简洁版)")

def make_test(n, expected):
    def test():
        result = fibonacci(n)
        assert result == expected, f"fibonacci({n}) = {result}, 期望 {expected}"
    return test

runner2.add_test("fib(0) = 0", make_test(0, 0))
runner2.add_test("fib(1) = 1", make_test(1, 1))
runner2.add_test("fib(5) = 5", make_test(5, 5))
runner2.add_test("fib(10) = 55", make_test(10, 55))
runner2.add_test("fib(20) = 6765", make_test(20, 6765))

def test_negative():
    try:
        fibonacci(-1)
        assert False, "应该抛出 ValueError"
    except ValueError:
        pass  # 期望的行为

runner2.add_test("fib(-1) 抛出 ValueError", test_negative)

runner2.run()
````

````python
# ============================================
# 练习 3 参考答案：代码重构
# ============================================

print("练习 3 参考答案 - 代码重构")
print("=" * 50)

# 重构后的代码
MAX_ITEM_PRICE = 100  # 移除魔法数字

DISCOUNT_TIERS = [    # 移除魔法数字
    (1000, 0.10),
    (500, 0.05),
]


def cap_price(price: float) -> float:
    """将单价限制在最大值以内（提取方法）"""
    return min(price, MAX_ITEM_PRICE)


def calculate_subtotal(prices: list) -> float:
    """计算小计（提取方法 + 重命名）"""
    return sum(cap_price(p) for p in prices if p > 0)


def calculate_discount(subtotal: float) -> float:
    """计算折扣金额（提取方法 + 简化条件）"""
    for threshold, rate in DISCOUNT_TIERS:
        if subtotal > threshold:
            return subtotal * rate
    return 0


def calculate_total_price(prices: list) -> float:
    """计算最终价格（重命名 + 单一职责）"""
    subtotal = calculate_subtotal(prices)
    discount = calculate_discount(subtotal)
    return subtotal - discount


# 测试
test_prices = [150, 80, 30, -10, 200, 50]
result = calculate_total_price(test_prices)
print(f"\n测试数据: {test_prices}")
print(f"小计: {calculate_subtotal(test_prices)}")
print(f"折扣: {calculate_discount(calculate_subtotal(test_prices))}")
print(f"最终价格: {result}")

print("\n应用的重构技术:")
print("  1. [重命名] p->calculate_total_price, l->prices, t->subtotal")
print("  2. [提取方法] cap_price(), calculate_subtotal(), calculate_discount()")
print("  3. [移除魔法数字] 100->MAX_ITEM_PRICE, 1000/0.1->DISCOUNT_TIERS")
print("  4. [简化条件] 嵌套if -> min() 和 列表推导")
print("  5. [添加文档] 每个函数都有 docstring")
````

````python
# ============================================
# 练习 4 参考答案：TDD 开发 Stack
# ============================================

print("练习 4 参考答案 - TDD 开发 Stack")
print("=" * 50)

# 第一步 (Red): 先写测试
print("\n第一步: 编写测试")


# 第二步 (Green): 实现 Stack
class Stack:
    """使用列表实现的栈 (LIFO)"""

    def __init__(self):
        self._items = []

    def push(self, item):
        """将元素压入栈顶"""
        self._items.append(item)

    def pop(self):
        """弹出栈顶元素"""
        if self.is_empty():
            raise IndexError("栈为空，无法弹出")
        return self._items.pop()

    def peek(self):
        """查看栈顶元素（不弹出）"""
        if self.is_empty():
            raise IndexError("栈为空，无法查看")
        return self._items[-1]

    def is_empty(self) -> bool:
        """判断栈是否为空"""
        return len(self._items) == 0

    def size(self) -> int:
        """返回栈的大小"""
        return len(self._items)


# 第三步: 运行测试
print("\n第二步: 实现 Stack 类 (已完成)")
print("\n第三步: 运行测试")


class TestStack(unittest.TestCase):
    def setUp(self):
        self.stack = Stack()

    def test_new_stack_is_empty(self):
        self.assertTrue(self.stack.is_empty())
        self.assertEqual(self.stack.size(), 0)

    def test_push_makes_non_empty(self):
        self.stack.push(1)
        self.assertFalse(self.stack.is_empty())
        self.assertEqual(self.stack.size(), 1)

    def test_push_pop_returns_correct(self):
        self.stack.push(42)
        self.assertEqual(self.stack.pop(), 42)
        self.assertTrue(self.stack.is_empty())

    def test_lifo_order(self):
        for i in [1, 2, 3]:
            self.stack.push(i)
        self.assertEqual(self.stack.pop(), 3)
        self.assertEqual(self.stack.pop(), 2)
        self.assertEqual(self.stack.pop(), 1)

    def test_peek_does_not_remove(self):
        self.stack.push("hello")
        self.assertEqual(self.stack.peek(), "hello")
        self.assertEqual(self.stack.size(), 1)

    def test_pop_empty_raises(self):
        with self.assertRaises(IndexError):
            self.stack.pop()

    def test_peek_empty_raises(self):
        with self.assertRaises(IndexError):
            self.stack.peek()


suite = unittest.TestLoader().loadTestsFromTestCase(TestStack)
runner = unittest.TextTestRunner(verbosity=2)
result = runner.run(suite)

print(f"\nTDD 完成！所有 {result.testsRun} 个测试通过。")
print("\nTDD 流程回顾:")
print("  Red   -> 先写测试（定义期望行为）")
print("  Green -> 写最少代码让测试通过")
print("  Refactor -> 改善代码质量，确保测试仍通过")
````

## 本节总结

### 核心知识点

1. **SDLC 模型**：没有最好的模型，只有最适合项目的模型。考试中需要根据场景选择并说明理由。

2. **版本控制 (Git)**：理解 commit、branch、merge 的概念。版本控制是现代软件开发的基础设施。

3. **代码文档**：好的代码 = 有意义的命名 + 适当的注释 + 完整的 docstring。

4. **测试方法论**：
   - 测试金字塔：单元测试最多，E2E 测试最少
   - TDD：Red → Green → Refactor
   - 黑盒看输入输出，白盒看内部逻辑

5. **代码重构**：在不改变行为的前提下改善代码结构。常用技术包括提取方法、重命名、移除魔法数字。

6. **软件度量**：圈复杂度衡量代码复杂程度，代码覆盖率衡量测试完整性。

7. **设计原则**：DRY（不重复）、KISS（保持简单）、SOLID（面向对象五原则）。

### 考试建议

- SDLC 模型选择题：**先分析场景特征**（需求是否明确？风险高不高？项目大不大？），再匹配模型
- 测试相关题：区分**测试类型**（unit/integration/system/acceptance）和**测试方法**（black box/white box）
- 重构题：**指出代码的问题**（code smell），然后**说明重构方法**和**重构后的改进**
